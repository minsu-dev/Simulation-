```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, lognorm, probplot

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    data: Í¥ÄÏ∏° Îç∞Ïù¥ÌÑ∞ Î∞∞Ïó¥
    dist: Î∂ÑÌè¨ Í∞ùÏ≤¥ (Ïòà: scipy.stats.gamma, scipy.stats.lognorm)
    dist_params: Î∂ÑÌè¨ ÌååÎùºÎØ∏ÌÑ∞ ÌäúÌîå (shape, loc, scale)
    title_prefix: Í∑∏ÎûòÌîÑ Ï†úÎ™© ÏïûÎ∂ÄÎ∂Ñ
    """
    shape, loc, scale = dist_params
    
    plt.figure(figsize=(12,5))
    
    # P-P Plot
    plt.subplot(1,2,1)
    ecdf = np.arange(1, len(data)+1) / len(data)
    sorted_data = np.sort(data)
    cdf_theoretical = dist.cdf(sorted_data, shape, loc=loc, scale=scale)
    plt.plot(cdf_theoretical, ecdf, 'o', label='Empirical vs Theoretical')
    plt.plot([0,1], [0,1], 'r--')
    plt.xlabel('Theoretical CDF')
    plt.ylabel('Empirical CDF')
    plt.title(f'{title_prefix} P-P Plot')
    plt.grid(True)
    
    # Q-Q Plot
    plt.subplot(1,2,2)
    probplot(data, dist=dist, sparams=(shape, loc, scale), plot=plt)
    plt.title(f'{title_prefix} Q-Q Plot')
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# ÌååÎùºÎØ∏ÌÑ∞ Ï∂îÏ†ï (loc=0 Í≥†Ï†ï)
params_c1 = gamma.fit(c1_all, floc=0)
params_c2_nonpeak = gamma.fit(c2_nonpeak, floc=0)
params_c2_peak_lognorm = lognorm.fit(c2_peak, floc=0)

# P-P PlotÍ≥º Q-Q Plot Í∑∏Î¶¨Í∏∞
plot_pp_qq(c1_all, gamma, params_c1, 'Counter 1 (All)')
plot_pp_qq(c2_nonpeak, gamma, params_c2_nonpeak, 'Counter 2 Non-Peak')
plot_pp_qq(c2_peak, lognorm, params_c2_peak_lognorm, 'Counter 2 Peak (Lognormal)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, lognorm

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    data       : 1D Í¥ÄÏ∏° Îç∞Ïù¥ÌÑ∞
    dist       : scipy.stats Î∂ÑÌè¨ Í∞ùÏ≤¥ (gamma, lognorm Îì±)
    dist_params: dist.fit() Í≤∞Í≥º ÌäúÌîå (shape1, shape2..., loc, scale)
    """
    data = np.asarray(data)
    data = data[data > 0]
    n = len(data)
    
    # ÎßàÏßÄÎßâ Îëê Í∞úÎäî loc, scale, ÎÇòÎ®∏ÏßÄÎäî shape ÌååÎùºÎØ∏ÌÑ∞Î°ú Ï≤òÎ¶¨
    *shape_params, loc, scale = dist_params
    
    # Îç∞Ïù¥ÌÑ∞ Ï†ïÎ†¨
    sorted_data = np.sort(data)
    
    plt.figure(figsize=(12, 5))
    
    # ---------- P-P Plot ----------
    plt.subplot(1, 2, 1)
    emp_cdf = (np.arange(1, n+1)) / (n+0.0)
    theo_cdf = dist.cdf(sorted_data, *shape_params, loc=loc, scale=scale)
    
    plt.plot(theo_cdf, emp_cdf, 'o', alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel('Theoretical CDF')
    plt.ylabel('Empirical CDF')
    plt.title(f'{title_prefix} P-P Plot')
    plt.grid(True)
    
    # ---------- Q-Q Plot ----------
    plt.subplot(1, 2, 2)
    # Ïù¥Î°†Ï†Å Î∂ÑÏúÑÏàò (0~1 ÏÇ¨Ïù¥ Í∑†Îì± Î∂ÑÌï†)
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)
    
    plt.scatter(theo_q, sorted_data, alpha=0.6)
    min_q = min(theo_q.min(), sorted_data.min())
    max_q = max(theo_q.max(), sorted_data.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Empirical Quantiles')
    plt.title(f'{title_prefix} Q-Q Plot')
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()


# ÌååÎùºÎØ∏ÌÑ∞ Ï∂îÏ†ï (loc=0 Í≥†Ï†ï)
params_c1         = gamma.fit(c1_all,     floc=0)
params_c2_nonpeak = gamma.fit(c2_nonpeak, floc=0)
params_c2_peak_ln = lognorm.fit(c2_peak,  floc=0)

# P-P & Q-Q Plot Í∑∏Î¶¨Í∏∞
plot_pp_qq(c1_all,     gamma,   params_c1,         'Counter 1 (All, Gamma)')
plot_pp_qq(c2_nonpeak, gamma,   params_c2_nonpeak,'Counter 2 Non-Peak (Gamma)')
plot_pp_qq(c2_peak,    lognorm, params_c2_peak_ln,'Counter 2 Peak (Lognormal)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import gamma, lognorm, expon, weibull_min

# =====================================================
# 0. Í≥µÌÜµ ÎèÑÏö∞ÎØ∏ Ìï®ÏàòÎì§
# =====================================================

def hhmmss_to_sec(t):
    """'HH:MM:SS' Î¨∏ÏûêÏó¥ÏùÑ Ï¥à Îã®ÏúÑ Ïà´ÏûêÎ°ú Î≥ÄÌôò"""
    h, m, s = map(int, t.split(':'))
    return h*3600 + m*60 + s

def to_seconds_list(str_list):
    """ÏãúÍ∞Ñ Î¨∏ÏûêÏó¥ Î¶¨Ïä§Ìä∏Î•º Ï¥à Îã®ÏúÑ numpy Î∞∞Ïó¥Î°ú Î≥ÄÌôò"""
    return np.array([hhmmss_to_sec(t) for t in str_list], dtype=float)

def tukey_bounds(data, k=1.5):
    """Tukey IQR Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò Í≤ΩÍ≥Ñ Í≥ÑÏÇ∞ (Q1-1.5IQR, Q3+1.5IQR)"""
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

def fit_and_compare_distributions(data, label=""):
    """
    Ïó¨Îü¨ Î∂ÑÌè¨(expon, gamma, lognorm, weibull_min)Î•º ÌîºÌåÖÌïòÍ≥†
    KS-test, loglikÎ•º Ï∂úÎ†•. p-value Í∏∞Ï§Ä Ï†ïÎ†¨ÌïòÏó¨ Î∞òÌôò.
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]  # 0 Ïù¥Ìïò Ï†úÍ±∞
    candidates = {
        "expon": expon,
        "gamma": gamma,
        "lognorm": lognorm,
        "weibull_min": weibull_min
    }

    results = []
    print(f"\n===== [{label}] Î∂ÑÌè¨ Ï†ÅÌï© Í≤∞Í≥º =====")
    for name, dist in candidates.items():
        # ÏÑúÎπÑÏä§/ÎèÑÏ∞©ÏãúÍ∞ÑÏùÄ 0 Ïù¥ÏÉÅÏù¥ÎØÄÎ°ú loc=0 Í≥†Ï†ï
        params = dist.fit(data, floc=0)
        ks, p = stats.kstest(data, dist.cdf, args=params)
        loglik = np.sum(dist.logpdf(data, *params))
        results.append({
            "name": name,
            "dist": dist,
            "params": params,
            "ks": ks,
            "p": p,
            "loglik": loglik
        })
        print(f"{name:10s} | KS={ks:.4f}, p={p:.6f}, loglik={loglik:.1f}")
        print(f"  params = {params}\n")

    results.sort(key=lambda x: x["p"], reverse=True)

    print(f"\n>>> [{label}] p-value Ïàú(Ï†ÅÌï©ÎèÑ Ïàú) <<<")
    for r in results:
        print(f"{r['name']:10s}  p={r['p']:.6f}, KS={r['ks']:.4f}")

    # ÏµúÏ†Å Î∂ÑÌè¨ ÌûàÏä§ÌÜ†Í∑∏Îû® + PDF ÏãúÍ∞ÅÌôî
    best = results[0]
    best_name = best["name"]
    best_dist = best["dist"]
    best_params = best["params"]

    x = np.linspace(data.min(), data.max(), 400)
    pdf = best_dist.pdf(x, *best_params)

    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=30, density=True, alpha=0.6,
             edgecolor='black', label='Observed')
    plt.plot(x, pdf, 'r-', lw=2, label=f'Best fit: {best_name}')
    plt.title(f"[{label}] Best Fitted: {best_name}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    return results

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    ÏÑ†ÌÉùÎêú Î∂ÑÌè¨Ïóê ÎåÄÌï¥ P-P plot, Q-Q plotÏùÑ Í∑∏Î†§Ï£ºÎäî Ìï®Ïàò.
    dist_paramsÎäî dist.fit Í≤∞Í≥º ÌäúÌîå (shape..., loc, scale)
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]
    data_sorted = np.sort(data)
    n = len(data_sorted)

    *shape_params, loc, scale = dist_params

    # -------- P-P Plot --------
    emp_cdf = (np.arange(1, n+1)) / (n + 0.0)
    theo_cdf = dist.cdf(data_sorted, *shape_params, loc=loc, scale=scale)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(theo_cdf, emp_cdf, alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel("Theoretical CDF")
    plt.ylabel("Empirical CDF")
    plt.title(f"{title_prefix} P-P Plot")
    plt.grid(True)

    # -------- Q-Q Plot --------
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)

    plt.subplot(1, 2, 2)
    plt.scatter(theo_q, data_sorted, alpha=0.6)
    min_q = min(theo_q.min(), data_sorted.min())
    max_q = max(theo_q.max(), data_sorted.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Empirical Quantiles")
    plt.title(f"{title_prefix} Q-Q Plot")
    plt.grid(True)

    plt.tight_layout()
    plt.show()


# =====================================================
# 1. ÏÑúÎπÑÏä§ÌÉÄÏûÑ / ÎèÑÏ∞©ÏãúÍ∞Ñ ‚Üí Ï¥à Îã®ÏúÑ Î∞∞Ïó¥Î°ú Î≥ÄÌôò
# =====================================================

# --- Counter 1 ÏÑúÎπÑÏä§ÌÉÄÏûÑ (Ï¥à) ---
c1_1006_service_secs = to_seconds_list(counter1_1006_service_time)
c1_1007_service_secs = to_seconds_list(counter1_1007_service_time)
c1_1008_service_secs = to_seconds_list(counter1_1008_service_time)
c1_1015_service_secs = to_seconds_list(counter1_1015_service_time)

# --- Counter 2 ÏÑúÎπÑÏä§ÌÉÄÏûÑ (Ï¥à) ---
c2_1006_service_secs = to_seconds_list(counter2_1006_service_time)
c2_1007_service_secs = to_seconds_list(counter2_1007_service_time)
c2_1008_service_secs = to_seconds_list(counter2_1008_service_time)
c2_1015_service_secs = to_seconds_list(counter2_1015_service_time)

# --- Counter 1 ÎèÑÏ∞©ÏãúÍ∞Ñ (Ï¥à) ---
a1_1006_secs = np.sort(to_seconds_list(counter1_1006_arrival_time))
a1_1007_secs = np.sort(to_seconds_list(counter1_1007_arrival_time))
a1_1008_secs = np.sort(to_seconds_list(counter1_1008_arrival_time))
a1_1015_secs = np.sort(to_seconds_list(counter1_1015_arrival_time))

# --- Counter 2 ÎèÑÏ∞©ÏãúÍ∞Ñ (Ï¥à) ---
a2_1006_secs = np.sort(to_seconds_list(counter2_1006_arrival_time))
a2_1007_secs = np.sort(to_seconds_list(counter2_1007_arrival_time))
a2_1008_secs = np.sort(to_seconds_list(counter2_1008_arrival_time))
a2_1015_secs = np.sort(to_seconds_list(counter2_1015_arrival_time))

# --- Interarrival time (Ï¥à) Í≥ÑÏÇ∞: Í∞ôÏùÄ Ïπ¥Ïö¥ÌÑ∞ ÎÇ¥ÏóêÏÑú Ïó∞ÏÜç ÎèÑÏ∞© Í∞ÑÍ≤© ---
ia1_1006 = np.diff(a1_1006_secs)
ia1_1007 = np.diff(a1_1007_secs)
ia1_1008 = np.diff(a1_1008_secs)
ia1_1015 = np.diff(a1_1015_secs)

ia2_1006 = np.diff(a2_1006_secs)
ia2_1007 = np.diff(a2_1007_secs)
ia2_1008 = np.diff(a2_1008_secs)
ia2_1015 = np.diff(a2_1015_secs)

# 'ÌïòÎ£® Ï†ÑÏ≤¥' interarrivalÏùÄ Îëê Ïπ¥Ïö¥ÌÑ∞ Ìï©Ïπú Í±∏ ÏÇ¨Ïö© (Îã®Ïàú concat)
ia_day_1006 = np.concatenate([ia1_1006, ia2_1006])
ia_day_1007 = np.concatenate([ia1_1007, ia2_1007])
ia_day_1008 = np.concatenate([ia1_1008, ia2_1008])
ia_day_1015 = np.concatenate([ia1_1015, ia2_1015])

# 4Ïùº Ï†ÑÏ≤¥ interarrival (ÎèôÏßàÏÑ± Í≤ÄÏ†ï ÌõÑ Ìïú Îç©Ïñ¥Î¶¨Î°ú Ïì∏ ÏòàÏ†ï)
ia_all = np.concatenate([ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015])

# =====================================================
# 2. ÏÑúÎπÑÏä§ÌÉÄÏûÑ: ÏùºÎ∞òÏùº vs ÌîºÌÅ¨Ïùº Î∂ÑÎ¶¨
# =====================================================

# ÏùºÎ∞òÏùº = 10/06, 10/07, 10/08 (Îëê Ïπ¥Ïö¥ÌÑ∞ Î™®Îëê)
st_normal = np.concatenate([
    c1_1006_service_secs, c1_1007_service_secs, c1_1008_service_secs,
    c2_1006_service_secs, c2_1007_service_secs, c2_1008_service_secs
])

# ÌîºÌÅ¨Ïùº = 10/15 (Îëê Ïπ¥Ïö¥ÌÑ∞ Î™®Îëê)
st_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# ÏùºÏûêÎ≥Ñ ÏÑúÎπÑÏä§ÌÉÄÏûÑ (ÎèôÏßàÏÑ± Í≤ÄÏ†ïÏö©)
st_day_1006 = np.concatenate([c1_1006_service_secs, c2_1006_service_secs])
st_day_1007 = np.concatenate([c1_1007_service_secs, c2_1007_service_secs])
st_day_1008 = np.concatenate([c1_1008_service_secs, c2_1008_service_secs])
st_day_1015 = np.concatenate([c1_1015_service_secs, c2_1015_service_secs])

# =====================================================
# 3. ÎèÖÎ¶ΩÏÑ± Í≤ÄÏ†ï (ÏÑúÎπÑÏä§ÌÉÄÏûÑ: Counter1 vs Counter2)
#    - ÏÇ∞Ï†êÎèÑ + Spearman ÏÉÅÍ¥ÄÍ≥ÑÏàò
# =====================================================

# ÏòàÏãú 1: ÏùºÎ∞òÏùº(10/07) ÏÑúÎπÑÏä§ÌÉÄÏûÑ ÎèÖÎ¶ΩÏÑ±
min_len_0707 = min(len(c1_1007_service_secs), len(c2_1007_service_secs))
x_0707 = c1_1007_service_secs[:min_len_0707]
y_0707 = c2_1007_service_secs[:min_len_0707]

plt.figure(figsize=(5, 4))
plt.scatter(x_0707, y_0707, alpha=0.6)
plt.xlabel("Counter1 Service Time (s) - 10/07")
plt.ylabel("Counter2 Service Time (s) - 10/07")
plt.title("Scatter: Service Time (Non-Peak Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_0707, p_0707 = stats.spearmanr(x_0707, y_0707)
print(f"[Spearman] 10/07 ServiceTime C1 vs C2: rho={rho_0707:.3f}, p={p_0707:.4f}")

# ÏòàÏãú 2: ÌîºÌÅ¨Ïùº(10/15) ÏÑúÎπÑÏä§ÌÉÄÏûÑ ÎèÖÎ¶ΩÏÑ±
min_len_1015 = min(len(c1_1015_service_secs), len(c2_1015_service_secs))
x_1015 = c1_1015_service_secs[:min_len_1015]
y_1015 = c2_1015_service_secs[:min_len_1015]

plt.figure(figsize=(5, 4))
plt.scatter(x_1015, y_1015, alpha=0.6, color='orange')
plt.xlabel("Counter1 Service Time (s) - 10/15")
plt.ylabel("Counter2 Service Time (s) - 10/15")
plt.title("Scatter: Service Time (Peak Day Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_1015, p_1015 = stats.spearmanr(x_1015, y_1015)
print(f"[Spearman] 10/15 ServiceTime C1 vs C2: rho={rho_1015:.3f}, p={p_1015:.4f}")

# (ÏõêÌïòÎ©¥ interarrivalÏóê ÎåÄÌï¥ÏÑúÎèÑ Í∞ôÏùÄ Î∞©ÏãùÏúºÎ°ú Spearman ÎèåÎ¶¥ Ïàò ÏûàÏùå)


# =====================================================
# 4. ÎèôÏßàÏÑ± Í≤ÄÏ†ï (Ïó¨Îü¨ ÎÇ† Îç∞Ïù¥ÌÑ∞ Ìï©Ï≥êÎèÑ ÎêòÎäîÏßÄ)
#    - Interarrival, Service Îëò Îã§ Kruskal‚ÄìWallis ÏÇ¨Ïö©
# =====================================================

# Interarrival 4Ïùº ÎèôÏßàÏÑ± (Îëê Ïπ¥Ïö¥ÌÑ∞ Ìï©Ïπú ÌïòÎ£® Îã®ÏúÑ)
H_ia, p_ia = stats.kruskal(ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015)
print(f"\n[Interarrival] Kruskal‚ÄìWallis: H={H_ia:.3f}, p={p_ia:.6f}")

# Service 4Ïùº ÎèôÏßàÏÑ±
H_st, p_st = stats.kruskal(st_day_1006, st_day_1007, st_day_1008, st_day_1015)
print(f"[Service] Kruskal‚ÄìWallis: H={H_st:.3f}, p={p_st:.6f}")

# p_ia > 0.05 ÎùºÎ©¥ ‚Üí interarrivalÎäî 4Ïùº Ìï©Ï≥êÏÑú ÏÇ¨Ïö© OK
# p_st << 0.05 ÎùºÎ©¥ ‚Üí ÏÑúÎπÑÏä§ÌÉÄÏûÑÏùÄ ÏùºÎ∞òÏùº vs ÌîºÌÅ¨Ïùº Î∂ÑÎ¶¨ ÌïÑÏöî


# =====================================================
# 5. Tukey boxplotÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌôïÏù∏ (Service Normal / Peak)
# =====================================================

for data, name in [(st_normal, "Normal Service Time"),
                   (st_peak, "Peak Service Time")]:

    low, up = tukey_bounds(data)
    outliers = data[(data < low) | (data > up)]

    print(f"\n[{name}] Tukey bounds: {low:.1f} ~ {up:.1f}")
    print(f"  Ïù¥ÏÉÅÏπò Í∞úÏàò: {len(outliers)}")

    plt.figure(figsize=(4, 5))
    plt.boxplot(data, vert=True, showfliers=True)
    plt.title(f"{name} (Tukey boxplot)")
    plt.ylabel("Service Time (s)")
    plt.tight_layout()
    plt.show()


# =====================================================
# 6. Î∂ÑÌè¨ ÌîºÌåÖ Activity 1
#    - Interarrival Ï†ÑÏ≤¥
#    - Service Normal (10/06~08)
#    - Service Peak (10/15)
# =====================================================

res_ia       = fit_and_compare_distributions(ia_all, label="Interarrival (All days)")
res_st_norm  = fit_and_compare_distributions(st_normal, label="Service Time - Normal Days")
res_st_peak  = fit_and_compare_distributions(st_peak, label="Service Time - Peak Day")

best_ia      = res_ia[0]
best_norm    = res_st_norm[0]
best_peak    = res_st_peak[0]

print("\n[ÏÑ†Ï†ïÎêú ÏµúÏÉÅÏúÑ ÌõÑÎ≥¥ Î∂ÑÌè¨Îì§]")
print("Interarrival :", best_ia["name"],   best_ia["params"])
print("Service Normal:", best_norm["name"], best_norm["params"])
print("Service Peak  :", best_peak["name"], best_peak["params"])


# =====================================================
# 7. ÏÑ†Ï†ïÎêú Î∂ÑÌè¨Ïóê ÎåÄÌïú P-P / Q-Q Plot (Ï†ïÍµêÌïú Ï†ÅÌï©ÎèÑ ÌôïÏù∏)
# =====================================================

# Interarrival
plot_pp_qq(ia_all, best_ia["dist"], best_ia["params"],
           title_prefix=f"Interarrival ({best_ia['name']})")

# Service Normal
plot_pp_qq(st_normal, best_norm["dist"], best_norm["params"],
           title_prefix=f"Service Normal ({best_norm['name']})")

# Service Peak
plot_pp_qq(st_peak, best_peak["dist"], best_peak["params"],
           title_prefix=f"Service Peak ({best_peak['name']})")


# =====================================================
# 8. (Ï∞∏Í≥†) SimPyÏóêÏÑú Ïì∏ Ïàò ÏûàÎäî ÌòïÌÉúÎ°ú ÌååÎùºÎØ∏ÌÑ∞ Ï†ïÎ¶¨ ÏòàÏãú
#    - Ïó¨Í∏∞ÏÑúÎäî Í∑∏ÎÉ• printÎßå, Ïã§Ï†ú SimPy ÏΩîÎìúÏóêÏÑú rvs()Î°ú ÏÇ¨Ïö©
# =====================================================

# Ïòà: interarrivalÍ∞Ä exponÏù¥Î©¥
if best_ia["name"] == "expon":
    loc_ia, scale_ia = best_ia["params"][1], best_ia["params"][2]
    print("\n[SimPyÏö© Interarrival ÏòàÏãú] stats.expon.rvs(loc=%.2f, scale=%.2f)" % (loc_ia, scale_ia))

# Ïòà: ÏùºÎ∞òÏùº ÏÑúÎπÑÏä§ÌÉÄÏûÑÏù¥ gammaÏù∏ Í≤ΩÏö∞
if best_norm["name"] == "gamma":
    shape_n, loc_n, scale_n = best_norm["params"]
    print("[SimPyÏö© Service Normal ÏòàÏãú] stats.gamma.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (shape_n, loc_n, scale_n))

# Ïòà: ÌîºÌÅ¨Ïùº ÏÑúÎπÑÏä§ÌÉÄÏûÑÏù¥ lognormÏù∏ Í≤ΩÏö∞
if best_peak["name"] == "lognorm":
    s_p, loc_p, scale_p = best_peak["params"]
    print("[SimPyÏö© Service Peak ÏòàÏãú] stats.lognorm.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (s_p, loc_p, scale_p))
```

```python
import simpy
import numpy as np
from scipy import stats

# ============================================================
# 1. Î∂ÑÌè¨ ÏÑ†ÌÉù (ÎÑàÏùò fit Í≤∞Í≥º ÏÇ¨Ïö©)
# ============================================================

# ‚Üí ÎÑàÏùò ÏΩîÎìúÏóêÏÑú ÎØ∏Î¶¨ Íµ¨Ìïú best_ia / best_norm / best_peak ÏÇ¨Ïö©
dist_ia      = best_ia["dist"]
params_ia    = best_ia["params"]

dist_norm    = best_norm["dist"]
params_norm  = best_norm["params"]

dist_peak    = best_peak["dist"]
params_peak  = best_peak["params"]

# ------------------------------------------------------------
# rvs() Í∞ÑÎã® ÎûòÌçº Ìï®Ïàò
# ------------------------------------------------------------

def sample_interarrival():
    return dist_ia.rvs(*params_ia)

def sample_service_normal():
    return dist_norm.rvs(*params_norm)

def sample_service_peak():
    return dist_peak.rvs(*params_peak)


# ============================================================
# 2. Í≥†Í∞ù ÌîÑÎ°úÏÑ∏Ïä§ Ï†ïÏùò
# ============================================================

def customer(env, name, counters, is_peak, logs):
    """
    counters: [Counter1_resource, Counter2_resource]
    is_peak: ÌîºÌÅ¨ÏùºÏù¥Î©¥ True ‚Üí peak service time ÏÇ¨Ïö©
    """
    arrive_time = env.now

    # -----------------------------
    # ÏÑúÎπÑÏä§ÌÉÄÏûÑ ÏÑ†ÌÉù (ÏùºÎ∞òÏùº/ÌîºÌÅ¨Ïùº)
    # -----------------------------
    if is_peak:
        service_time = sample_service_peak()
    else:
        service_time = sample_service_normal()

    # -----------------------------
    # Îëê Ïπ¥Ïö¥ÌÑ∞ Ï§ë ÎåÄÍ∏∞Ïó¥Ïù¥ ÏßßÏùÄ Í≥≥ ÏÑ†ÌÉù
    # -----------------------------
    lens = [len(c.queue) for c in counters]
    chosen_idx = np.argmin(lens)
    chosen_counter = counters[chosen_idx]

    with chosen_counter.request() as req:
        yield req

        start_service = env.now
        wait = start_service - arrive_time
        
        # ÏÑúÎπÑÏä§
        yield env.timeout(service_time)
        finish = env.now

    # Î°úÍ∑∏ Ï†ÄÏû•
    logs["waits"].append(wait)
    logs["services"].append(service_time)
    logs["system_times"].append(finish - arrive_time)


# ============================================================
# 3. ÎèÑÏ∞© ÌîÑÎ°úÏÑ∏Ïä§ (Interarrival Í∏∞Î∞ò)
# ============================================================

def customer_arrivals(env, counters, is_peak, logs, SIM_TIME):
    i = 0
    while env.now < SIM_TIME:
        ia = sample_interarrival()
        yield env.timeout(ia)

        i += 1
        env.process(customer(env, f"Cust-{i}", counters, is_peak, logs))


# ============================================================
# 4. Ï†ÑÏ≤¥ ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ Ìï®Ïàò
# ============================================================

def run_simulation(is_peak=False, SIM_TIME=3600*6, NUM_SERVERS=1):
    """
    is_peak: ÌîºÌÅ¨ÏùºÏù¥Î©¥ True ‚Üí peak Î∂ÑÌè¨ ÏÇ¨Ïö©
    SIM_TIME: ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏãúÍ∞Ñ(Ï¥à)
    """

    env = simpy.Environment()

    # Counter1, Counter2 Îëê Í∞úÏùò ÏûêÏõê ÏÉùÏÑ±
    c1 = simpy.Resource(env, capacity=NUM_SERVERS)
    c2 = simpy.Resource(env, capacity=NUM_SERVERS)
    counters = [c1, c2]

    logs = {
        "waits": [],
        "services": [],
        "system_times": []
    }

    env.process(customer_arrivals(env, counters, is_peak, logs, SIM_TIME))
    env.run(until=SIM_TIME)

    return logs


# ============================================================
# 5. ÌÖåÏä§Ìä∏ Ïã§Ìñâ
# ============================================================

# 6ÏãúÍ∞Ñ ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏòàÏãú
logs_normal = run_simulation(is_peak=False, SIM_TIME=3600*6)
logs_peak   = run_simulation(is_peak=True,  SIM_TIME=3600*6)

print("\n===== ÏùºÎ∞òÏùº Í≤∞Í≥º =====")
print("ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ:", np.mean(logs_normal["waits"]))
print("ÌèâÍ∑† ÏÑúÎπÑÏä§ÌÉÄÏûÑ:", np.mean(logs_normal["services"]))
print("ÌèâÍ∑† ÏãúÏä§ÌÖúÏ≤¥Î•òÏãúÍ∞Ñ:", np.mean(logs_normal["system_times"]))

print("\n===== ÌîºÌÅ¨Ïùº Í≤∞Í≥º =====")
print("ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ:", np.mean(logs_peak["waits"]))
print("ÌèâÍ∑† ÏÑúÎπÑÏä§ÌÉÄÏûÑ:", np.mean(logs_peak["services"]))
print("ÌèâÍ∑† ÏãúÏä§ÌÖúÏ≤¥Î•òÏãúÍ∞Ñ:", np.mean(logs_peak["system_times"]))
```

```python
import simpy
import numpy as np
import pandas as pd
from scipy.stats import expon, gamma, lognorm, weibull_min

# =====================================================
# 1. ÏãúÎÆ¨Î†àÏù¥ÏÖò ÌååÎùºÎØ∏ÌÑ∞ Î∞è Î∂ÑÌè¨ ÏÑ§Ï†ï (Î∂ÑÏÑù Í≤∞Í≥º Í∏∞Î∞ò)
# =====================================================

# ÏãúÎÆ¨Î†àÏù¥ÏÖò ÌôòÍ≤Ω ÏÑ§Ï†ï
NUM_COUNTERS = 2        # Ïπ¥Ïö¥ÌÑ∞(ÏßÅÏõê) Ïàò
SIM_DURATION = 8 * 3600 # ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏãúÍ∞Ñ (8ÏãúÍ∞ÑÏùÑ Ï¥àÎ°ú ÌôòÏÇ∞)
RANDOM_SEED = 42

# 1-1. ÎèÑÏ∞© Í∞ÑÍ≤© ÏãúÍ∞Ñ Î∂ÑÌè¨ (Interarrival Time)
# Î∂ÑÏÑù Í≤∞Í≥º: Interarrival (All days)Ïùò ÏµúÏ†Å Î∂ÑÌè¨
IA_DIST = expon  # Expon Î∂ÑÌè¨Î•º ÏÇ¨Ïö© (Simple Poisson Arrival)
IA_PARAMS = (0, 15.0)  # (loc, scale=1/lambda). scale=15.0Ï¥à Í∞ÄÏ†ï

# 1-2. ÏùºÎ∞òÏùº ÏÑúÎπÑÏä§ ÏãúÍ∞Ñ Î∂ÑÌè¨ (Service Time - Normal)
# Î∂ÑÏÑù Í≤∞Í≥º: Service Time - Normal DaysÏùò ÏµúÏ†Å Î∂ÑÌè¨
ST_NORMAL_DIST = gamma
ST_NORMAL_PARAMS = (1.62, 0, 37.95) # (a, loc, scale). a=shape, scale=rate

# 1-3. ÌîºÌÅ¨Ïùº ÏÑúÎπÑÏä§ ÏãúÍ∞Ñ Î∂ÑÌè¨ (Service Time - Peak)
# Î∂ÑÏÑù Í≤∞Í≥º: Service Time - Peak DayÏùò ÏµúÏ†Å Î∂ÑÌè¨
ST_PEAK_DIST = lognorm
ST_PEAK_PARAMS = (0.50, 0, 70.0) # (s, loc, scale). s=shape

# =====================================================
# 2. ÌÜµÍ≥Ñ Î°úÍπÖÏùÑ ÏúÑÌïú Ï†ÑÏó≠ Î≥ÄÏàò
# =====================================================
all_results = []
wait_times = []
service_times = []
sim_type = ""

# =====================================================
# 3. Í≥†Í∞ù ÏÉùÏÑ± Î∞è ÏÑúÎπÑÏä§ Ï≤òÎ¶¨ ÌîÑÎ°úÏÑ∏Ïä§
# =====================================================

def customer_process(env, name, counters, is_peak_day):
    """Í∞úÎ≥Ñ Í≥†Í∞ùÏùò ÎèÑÏ∞©, ÎåÄÍ∏∞, ÏÑúÎπÑÏä§, Ìá¥Ïû• Í≥ºÏ†ïÏùÑ Î™®Îç∏ÎßÅÌï©ÎãàÎã§."""
    
    # ÏÑúÎπÑÏä§ ÏãúÍ∞Ñ Î∂ÑÌè¨ ÏÑ†ÌÉù
    if is_peak_day:
        dist = ST_PEAK_DIST
        params = ST_PEAK_PARAMS
    else:
        dist = ST_NORMAL_DIST
        params = ST_NORMAL_PARAMS

    # ÏÑúÎπÑÏä§ ÏãúÍ∞Ñ ÏÉòÌîåÎßÅ (ÎÇúÏàò ÏÉùÏÑ±)
    service_time = dist.rvs(*params, size=1)[0]
    
    arrival_time = env.now
    
    with counters.request() as req:
        # ÌÅê ÎåÄÍ∏∞ ÏãúÏûë
        yield req
        
        # ÌÅê ÎåÄÍ∏∞ Ï¢ÖÎ£å (ÏÑúÎπÑÏä§ ÏãúÏûë)
        start_service_time = env.now
        waiting_time = start_service_time - arrival_time
        wait_times.append(waiting_time)
        
        # ÏÑúÎπÑÏä§ ÏàòÌñâ
        yield env.timeout(service_time)
        
        # ÏÑúÎπÑÏä§ Ï¢ÖÎ£å
        departure_time = env.now
        
        # Í≤∞Í≥º Í∏∞Î°ù
        all_results.append({
            'Sim_Type': sim_type,
            'Name': name,
            'Arrival': arrival_time,
            'Start_Service': start_service_time,
            'Departure': departure_time,
            'Wait_Time': waiting_time,
            'Service_Time': service_time
        })
        service_times.append(service_time)

def arrival_generator(env, counters, is_peak_day):
    """Í≥†Í∞ùÏùò ÎèÑÏ∞©ÏùÑ ÏÉùÏÑ±ÌïòÎäî ÌîÑÎ°úÏÑ∏Ïä§ÏûÖÎãàÎã§."""
    customer_id = 0
    
    while True:
        # ÎèÑÏ∞© Í∞ÑÍ≤© ÏãúÍ∞Ñ ÏÉòÌîåÎßÅ
        # Interarrival TimeÏùÄ Î™®Îì† ÎÇ†Ïßú Îç∞Ïù¥ÌÑ∞Í∞Ä ÎèôÏßàÌïòÏó¨ ÌïòÎÇòÏùò Î∂ÑÌè¨Î•º ÏÇ¨Ïö©Ìï®
        ia_time = IA_DIST.rvs(*IA_PARAMS, size=1)[0]
        yield env.timeout(ia_time)
        
        customer_id += 1
        name = f'Customer-{customer_id}'
        
        # Í≥†Í∞ù ÌîÑÎ°úÏÑ∏Ïä§ ÏãúÏûë
        env.process(customer_process(env, name, counters, is_peak_day))

# =====================================================
# 4. ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ Ìï®Ïàò Î∞è Í≤∞Í≥º Ï∂úÎ†•
# =====================================================

def run_simulation(is_peak_day):
    """SimPy ÏãúÎÆ¨Î†àÏù¥ÏÖò ÌôòÍ≤ΩÏùÑ ÏÑ§Ï†ïÌïòÍ≥† Ïã§ÌñâÌï©ÎãàÎã§."""
    global wait_times, service_times, sim_type, all_results
    
    np.random.seed(RANDOM_SEED)
    env = simpy.Environment()
    
    # Í≤∞Í≥º Î≥ÄÏàò Ï¥àÍ∏∞Ìôî
    wait_times = []
    service_times = []
    all_results = []
    
    if is_peak_day:
        sim_type = "Peak Day (10/15)"
        print(f"\n======== üìà {sim_type} ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏãúÏûë (8ÏãúÍ∞Ñ) ========")
    else:
        sim_type = "Normal Day (10/06~08)"
        print(f"\n======== üìâ {sim_type} ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏãúÏûë (8ÏãúÍ∞Ñ) ========")

    # ÏûêÏõê(Ïπ¥Ïö¥ÌÑ∞) Ï†ïÏùò
    counters = simpy.Resource(env, capacity=NUM_COUNTERS)
    
    # ÎèÑÏ∞© ÌîÑÎ°úÏÑ∏Ïä§ ÏãúÏûë
    env.process(arrival_generator(env, counters, is_peak_day))
    
    # ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
    env.run(until=SIM_DURATION)
    
    # Í≤∞Í≥º Î∂ÑÏÑù
    if wait_times:
        avg_wait = np.mean(wait_times)
        avg_service = np.mean(service_times)
        
        print(f"Ï¥ù Í≥†Í∞ù Ïàò: {len(wait_times)}Î™Ö")
        print(f"ÌèâÍ∑† ÎåÄÍ∏∞ ÏãúÍ∞Ñ: {avg_wait:.2f}Ï¥à ({avg_wait/60:.2f}Î∂Ñ)")
        print(f"ÌèâÍ∑† ÏÑúÎπÑÏä§ ÏãúÍ∞Ñ: {avg_service:.2f}Ï¥à")
        print(f"ÏµúÎåÄ ÎåÄÍ∏∞ ÏãúÍ∞Ñ: {np.max(wait_times):.2f}Ï¥à")
        
        return pd.DataFrame(all_results)
    else:
        print("Í≥†Í∞ù Îç∞Ïù¥ÌÑ∞Í∞Ä ÏóÜÏäµÎãàÎã§.")
        return pd.DataFrame()

# =====================================================
# 5. ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
# =====================================================

# 1. ÏùºÎ∞òÏùº ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
df_normal = run_simulation(is_peak_day=False)

# 2. ÌîºÌÅ¨Ïùº ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
df_peak = run_simulation(is_peak_day=True)

# ÌÜµÌï© Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÌôïÏù∏ (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
# final_df = pd.concat([df_normal, df_peak])
# print("\nÌÜµÌï© Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ (ÏùºÎ∂Ä):")
# print(final_df.head())
```

Ïó¨Í∏∞ÏÑú .rvs() Îäî Random Variates SamplingÏùò ÏïΩÏûêÎ°ú
‚ÄúÎ∂ÑÌè¨ÏóêÏÑú ÎûúÎç§ ÏÉòÌîåÏùÑ ÎΩëÎäîÎã§‚ÄùÎäî ÎúªÏù¥Ïïº.
Ï¶â, rvs() Í∞Ä ÎÇ¥Î∂ÄÏ†ÅÏúºÎ°ú RNGÎ•º Ìò∏Ï∂úÌï¥ÏÑú ÎÇúÏàòÎ•º ÏÉùÏÑ±Ìï¥.

```python
import simpy
import numpy as np
from scipy import stats
import random

# =====================================================
# 1. (Ï§ëÏöî) Î∂ÑÌè¨ ÌååÎùºÎØ∏ÌÑ∞ Ï±ÑÏõåÎÑ£Í∏∞ ‚Äî ÎÑ§ Î∂ÑÏÑù Í≤∞Í≥ºÎ°ú Î∞îÍøîÏ§òÏïº ÌïòÎäî Î∂ÄÎ∂Ñ
# =====================================================

# ÏòàÏãú: Ïã§Ï†ú best_ia, best_norm, best_peakÏóêÏÑú ÎÇòÏò® Í∞íÏúºÎ°ú ÏàòÏ†ïÌïòÎ©¥ Îê®
INTERARRIVAL_DIST_NAME = "expon"
INTERARRIVAL_PARAMS    = (0.0, 30.0)       # Ïòà: (loc, scale)

SERVICE_NORMAL_DIST_NAME = "gamma"
SERVICE_NORMAL_PARAMS    = (3.8869, 0.0, 16.4619)   # Ïòà: (shape, loc, scale)

SERVICE_PEAK_DIST_NAME = "lognorm"
SERVICE_PEAK_PARAMS    = (0.6802, 0.0, 39.7381)     # Ïòà: (s, loc, scale)

DIST_MAP = {
    "expon": stats.expon,
    "gamma": stats.gamma,
    "lognorm": stats.lognorm,
    "weibull_min": stats.weibull_min
}

INTERARRIVAL_DIST   = DIST_MAP[INTERARRIVAL_DIST_NAME]
SERVICE_NORMAL_DIST = DIST_MAP[SERVICE_NORMAL_DIST_NAME]
SERVICE_PEAK_DIST   = DIST_MAP[SERVICE_PEAK_DIST_NAME]


# =====================================================
# 2. ÏÉòÌîåÎßÅ Ìï®ÏàòÎì§
# =====================================================

def sample_interarrival():
    """ÎèÑÏ∞© Í∞ÑÍ≤© ÏÉòÌîåÎßÅ"""
    return INTERARRIVAL_DIST.rvs(*INTERARRIVAL_PARAMS)

def sample_service_time(scenario="normal"):
    """ÏÑúÎπÑÏä§ ÏãúÍ∞Ñ ÏÉòÌîåÎßÅ"""
    if scenario == "normal":
        return SERVICE_NORMAL_DIST.rvs(*SERVICE_NORMAL_PARAMS)
    elif scenario == "peak":
        return SERVICE_PEAK_DIST.rvs(*SERVICE_PEAK_PARAMS)
    else:
        raise ValueError("scenarioÎäî 'normal' ÎòêÎäî 'peak'Îßå Í∞ÄÎä•Ìï©ÎãàÎã§.")


# =====================================================
# 3. Í≥†Í∞ù ÌîÑÎ°úÏÑ∏Ïä§
# =====================================================

def customer(env, name, counter, scenario, stats_recorder):
    arrival_time = env.now

    with counter.request() as req:
        yield req
        start = env.now
        waiting = start - arrival_time

        service_time = sample_service_time(scenario)
        yield env.timeout(service_time)
        end = env.now

    system_time = end - arrival_time

    stats_recorder["waiting_times"].append(waiting)
    stats_recorder["service_times"].append(service_time)
    stats_recorder["system_times"].append(system_time)


def customer_generator(env, counter, scenario, stats_recorder):
    i = 0
    while True:
        ia = sample_interarrival()
        yield env.timeout(ia)

        i += 1
        env.process(customer(env, f"Cust {i}", counter, scenario, stats_recorder))


# =====================================================
# 4. ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ (1ÏãúÍ∞Ñ 30Î∂Ñ = 5400Ï¥à)
# =====================================================

def run_simulation(scenario="normal", num_counters=2,
                   sim_time=5400, seed=42):
    """
    ÌïòÎÇòÏùò ÏãúÎÇòÎ¶¨Ïò§ + Ïπ¥Ïö¥ÌÑ∞ ÏàòÏóê ÎåÄÌï¥ 1Ìöå ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
    Í≤∞Í≥º: ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ, ÏãúÏä§ÌÖúÏ≤¥Î•òÏãúÍ∞Ñ Îì± Î¶¨ÌÑ¥
    """
    random.seed(seed)
    np.random.seed(seed)

    env = simpy.Environment()
    counter = simpy.Resource(env, capacity=num_counters)

    stats_recorder = {
        "waiting_times": [],
        "service_times": [],
        "system_times": []
    }

    env.process(customer_generator(env, counter, scenario, stats_recorder))
    env.run(until=sim_time)

    wt = np.array(stats_recorder["waiting_times"])
    st = np.array(stats_recorder["service_times"])
    sys = np.array(stats_recorder["system_times"])

    results = {
        "scenario": scenario,
        "num_counters": num_counters,
        "num_customers": len(wt),
        "avg_waiting": wt.mean() if len(wt) else np.nan,
        "avg_service": st.mean() if len(st) else np.nan,
        "avg_system": sys.mean() if len(sys) else np.nan
    }
    return results


# =====================================================
# 5. ÏµúÏ†Å Ïπ¥Ïö¥ÌÑ∞ Ïàò Ï∞æÎäî Ìï®Ïàò Ï∂îÍ∞Ä ‚úÖ
# =====================================================

def find_optimal_counters(scenario, candidate_counters,
                          sim_time=5400, seed=42,
                          wait_threshold=None, verbose=True):
    """
    - scenario: "normal" ÎòêÎäî "peak"
    - candidate_counters: [1,2,3,4] Ïù¥Îü∞ ÏãùÏùò ÌõÑÎ≥¥ Î¶¨Ïä§Ìä∏
    - wait_threshold: ÌóàÏö© Í∞ÄÎä•Ìïú 'ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ(Ï¥à)' Í∏∞Ï§Ä
        * Ïòà: 60 Ïù¥Î©¥ "ÌèâÍ∑† ÎåÄÍ∏∞ <= 60Ï¥à"Î•º ÎßåÏ°±ÌïòÎäî
              Í∞ÄÏû• ÏûëÏùÄ Ïπ¥Ïö¥ÌÑ∞ ÏàòÎ•º ÏµúÏ†ÅÏù¥ÎùºÍ≥† ÌåêÎã®
        * None Ïù¥Î©¥ 'avg_waitingÏù¥ ÏµúÏÜå'Ïù∏ Í≤ΩÏö∞ ÏÑ†ÌÉù
    - Î∞òÌôòÍ∞í: (best_result, all_results Î¶¨Ïä§Ìä∏)
    """
    all_results = []

    for n in candidate_counters:
        res = run_simulation(scenario=scenario,
                             num_counters=n,
                             sim_time=sim_time,
                             seed=seed)
        all_results.append(res)

        if verbose:
            print(f"[{scenario}] counters={n}, "
                  f"N={res['num_customers']}, "
                  f"Avg Wait={res['avg_waiting']:.2f}s, "
                  f"Avg System={res['avg_system']:.2f}s")

    # 1) Í∏∞Ï§Ä ÎåÄÍ∏∞ÏãúÍ∞ÑÏù¥ Ï£ºÏñ¥Ï°åÎã§Î©¥ ‚Üí Í∑∏Í±∏ ÎßåÏ°±ÌïòÎäî ÏµúÏÜå Ïπ¥Ïö¥ÌÑ∞ ÏÑ†ÌÉù
    if wait_threshold is not None:
        eligible = [r for r in all_results if r["avg_waiting"] <= wait_threshold]
        if len(eligible) > 0:
            best = min(eligible, key=lambda r: r["num_counters"])
        else:
            # Í∏∞Ï§ÄÏùÑ ÎßåÏ°±ÌïòÎäî Í≤å ÏóÜÎã§Î©¥, Í∑∏ÎÉ• Í∞ÄÏû• ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞ÑÏù¥ ÏßßÏùÄ ÏºÄÏù¥Ïä§ ÏÑ†ÌÉù
            best = min(all_results, key=lambda r: r["avg_waiting"])
    else:
        # 2) Í∏∞Ï§ÄÏù¥ ÏóÜÎã§Î©¥ ‚Üí avg_waiting ÏµúÏÜåÏù∏ ÏºÄÏù¥Ïä§ ÏÑ†ÌÉù
        best = min(all_results, key=lambda r: r["avg_waiting"])

    if verbose:
        print("\n>>> [", scenario, "] ÏµúÏ†Å Ïπ¥Ïö¥ÌÑ∞ Ïàò ÏÑ†ÌÉù Í≤∞Í≥º <<<")
        print(f"ÏµúÏ†Å counters = {best['num_counters']}Í∞ú")
        print(f"  ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ = {best['avg_waiting']:.2f}Ï¥à")
        print(f"  ÌèâÍ∑† ÏãúÏä§ÌÖúÏ≤¥Î•òÏãúÍ∞Ñ = {best['avg_system']:.2f}Ï¥à")
        print(f"  Ï≤òÎ¶¨ Í≥†Í∞ù Ïàò = {best['num_customers']}Î™Ö\n")

    return best, all_results


# =====================================================
# 6. ÌÖåÏä§Ìä∏ Ïã§Ìñâ ÏòàÏãú (ÏµúÏ†Å Ïπ¥Ïö¥ÌÑ∞ ÏûêÎèô Í≥ÑÏÇ∞)
# =====================================================

if __name__ == "__main__":
    # ÏòàÏãú Í∏∞Ï§Ä:
    # - ÏùºÎ∞òÏùº: ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ 60Ï¥à Ïù¥Ìïò Î™©Ìëú
    # - ÌîºÌÅ¨Ïùº: ÌèâÍ∑† ÎåÄÍ∏∞ÏãúÍ∞Ñ 30Ï¥à Ïù¥Ìïò Î™©Ìëú (Ï¢Ä Îçî ÌÉÄÏù¥Ìä∏ÌïòÍ≤å)
    NORMAL_WAIT_TARGET = 60.0   # 1Î∂Ñ
    PEAK_WAIT_TARGET   = 30.0   # 30Ï¥à

    print("=== Normal Day ÏãúÎÇòÎ¶¨Ïò§ ===")
    best_normal, normal_res = find_optimal_counters(
        scenario="normal",
        candidate_counters=[1, 2, 3, 4],
        sim_time=5400,
        seed=123,
        wait_threshold=NORMAL_WAIT_TARGET,
        verbose=True
    )

    print("=== Peak Day ÏãúÎÇòÎ¶¨Ïò§ ===")
    best_peak, peak_res = find_optimal_counters(
        scenario="peak",
        candidate_counters=[2, 3, 4, 5],
        sim_time=5400,
        seed=123,
        wait_threshold=PEAK_WAIT_TARGET,
        verbose=True
    )

    print(">>> Ï¢ÖÌï© ÏöîÏïΩ")
    print(f"ÏùºÎ∞òÏùº ÏµúÏ†Å Ïπ¥Ïö¥ÌÑ∞ Ïàò: {best_normal['num_counters']}Í∞ú "
          f"(ÌèâÍ∑† ÎåÄÍ∏∞ {best_normal['avg_waiting']:.2f}Ï¥à)")
    print(f"ÌîºÌÅ¨Ïùº ÏµúÏ†Å Ïπ¥Ïö¥ÌÑ∞ Ïàò: {best_peak['num_counters']}Í∞ú "
          f"(ÌèâÍ∑† ÎåÄÍ∏∞ {best_peak['avg_waiting']:.2f}Ï¥à)")
```
