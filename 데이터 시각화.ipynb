```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, lognorm

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    data       : 1D Í¥ÄÏ∏° Îç∞Ïù¥ÌÑ∞
    dist       : scipy.stats Î∂ÑÌè¨ Í∞ùÏ≤¥ (gamma, lognorm Îì±)
    dist_params: dist.fit() Í≤∞Í≥º ÌäúÌîå (shape1, shape2..., loc, scale)
    """
    data = np.asarray(data)
    data = data[data > 0]
    n = len(data)
    
    # ÎßàÏßÄÎßâ Îëê Í∞úÎäî loc, scale, ÎÇòÎ®∏ÏßÄÎäî shape ÌååÎùºÎØ∏ÌÑ∞Î°ú Ï≤òÎ¶¨
    *shape_params, loc, scale = dist_params
    
    # Îç∞Ïù¥ÌÑ∞ Ï†ïÎ†¨
    sorted_data = np.sort(data)
    
    plt.figure(figsize=(12, 5))
    
    # ---------- P-P Plot ----------
    plt.subplot(1, 2, 1)
    emp_cdf = (np.arange(1, n+1)) / (n+0.0)
    theo_cdf = dist.cdf(sorted_data, *shape_params, loc=loc, scale=scale)
    
    plt.plot(theo_cdf, emp_cdf, 'o', alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel('Theoretical CDF')
    plt.ylabel('Empirical CDF')
    plt.title(f'{title_prefix} P-P Plot')
    plt.grid(True)
    
    # ---------- Q-Q Plot ----------
    plt.subplot(1, 2, 2)
    # Ïù¥Î°†Ï†Å Î∂ÑÏúÑÏàò (0~1 ÏÇ¨Ïù¥ Í∑†Îì± Î∂ÑÌï†)
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)
    
    plt.scatter(theo_q, sorted_data, alpha=0.6)
    min_q = min(theo_q.min(), sorted_data.min())
    max_q = max(theo_q.max(), sorted_data.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Empirical Quantiles')
    plt.title(f'{title_prefix} Q-Q Plot')
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()


# ÌååÎùºÎØ∏ÌÑ∞ Ï∂îÏ†ï (loc=0 Í≥†Ï†ï)
params_c1         = gamma.fit(c1_all,     floc=0)
params_c2_nonpeak = gamma.fit(c2_nonpeak, floc=0)
params_c2_peak_ln = lognorm.fit(c2_peak,  floc=0)

# P-P & Q-Q Plot Í∑∏Î¶¨Í∏∞
plot_pp_qq(c1_all,     gamma,   params_c1,         'Counter 1 (All, Gamma)')
plot_pp_qq(c2_nonpeak, gamma,   params_c2_nonpeak,'Counter 2 Non-Peak (Gamma)')
plot_pp_qq(c2_peak,    lognorm, params_c2_peak_ln,'Counter 2 Peak (Lognormal)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import gamma, lognorm, expon, weibull_min

# =====================================================
# 0. Í≥µÌÜµ ÎèÑÏö∞ÎØ∏ Ìï®ÏàòÎì§
# =====================================================

def hhmmss_to_sec(t):
    """'HH:MM:SS' Î¨∏ÏûêÏó¥ÏùÑ Ï¥à Îã®ÏúÑ Ïà´ÏûêÎ°ú Î≥ÄÌôò"""
    h, m, s = map(int, t.split(':'))
    return h*3600 + m*60 + s

def to_seconds_list(str_list):
    """ÏãúÍ∞Ñ Î¨∏ÏûêÏó¥ Î¶¨Ïä§Ìä∏Î•º Ï¥à Îã®ÏúÑ numpy Î∞∞Ïó¥Î°ú Î≥ÄÌôò"""
    return np.array([hhmmss_to_sec(t) for t in str_list], dtype=float)

def tukey_bounds(data, k=1.5):
    """Tukey IQR Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò Í≤ΩÍ≥Ñ Í≥ÑÏÇ∞ (Q1-1.5IQR, Q3+1.5IQR)"""
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

def fit_and_compare_distributions(data, label=""):
    """
    Ïó¨Îü¨ Î∂ÑÌè¨(expon, gamma, lognorm, weibull_min)Î•º ÌîºÌåÖÌïòÍ≥†
    KS-test, loglikÎ•º Ï∂úÎ†•. p-value Í∏∞Ï§Ä Ï†ïÎ†¨ÌïòÏó¨ Î∞òÌôò.
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]  # 0 Ïù¥Ìïò Ï†úÍ±∞
    candidates = {
        "expon": expon,
        "gamma": gamma,
        "lognorm": lognorm,
        "weibull_min": weibull_min
    }

    results = []
    print(f"\n===== [{label}] Î∂ÑÌè¨ Ï†ÅÌï© Í≤∞Í≥º =====")
    for name, dist in candidates.items():
        # ÏÑúÎπÑÏä§/ÎèÑÏ∞©ÏãúÍ∞ÑÏùÄ 0 Ïù¥ÏÉÅÏù¥ÎØÄÎ°ú loc=0 Í≥†Ï†ï
        params = dist.fit(data, floc=0)
        ks, p = stats.kstest(data, dist.cdf, args=params)
        loglik = np.sum(dist.logpdf(data, *params))
        results.append({
            "name": name,
            "dist": dist,
            "params": params,
            "ks": ks,
            "p": p,
            "loglik": loglik
        })
        print(f"{name:10s} | KS={ks:.4f}, p={p:.6f}, loglik={loglik:.1f}")
        print(f"  params = {params}\n")

    results.sort(key=lambda x: x["p"], reverse=True)

    print(f"\n>>> [{label}] p-value Ïàú(Ï†ÅÌï©ÎèÑ Ïàú) <<<")
    for r in results:
        print(f"{r['name']:10s}  p={r['p']:.6f}, KS={r['ks']:.4f}")

    # ÏµúÏ†Å Î∂ÑÌè¨ ÌûàÏä§ÌÜ†Í∑∏Îû® + PDF ÏãúÍ∞ÅÌôî
    best = results[0]
    best_name = best["name"]
    best_dist = best["dist"]
    best_params = best["params"]

    x = np.linspace(data.min(), data.max(), 400)
    pdf = best_dist.pdf(x, *best_params)

    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=30, density=True, alpha=0.6,
             edgecolor='black', label='Observed')
    plt.plot(x, pdf, 'r-', lw=2, label=f'Best fit: {best_name}')
    plt.title(f"[{label}] Best Fitted: {best_name}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    return results

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    ÏÑ†ÌÉùÎêú Î∂ÑÌè¨Ïóê ÎåÄÌï¥ P-P plot, Q-Q plotÏùÑ Í∑∏Î†§Ï£ºÎäî Ìï®Ïàò.
    dist_paramsÎäî dist.fit Í≤∞Í≥º ÌäúÌîå (shape..., loc, scale)
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]
    data_sorted = np.sort(data)
    n = len(data_sorted)

    *shape_params, loc, scale = dist_params

    # -------- P-P Plot --------
    emp_cdf = (np.arange(1, n+1)) / (n + 0.0)
    theo_cdf = dist.cdf(data_sorted, *shape_params, loc=loc, scale=scale)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(theo_cdf, emp_cdf, alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel("Theoretical CDF")
    plt.ylabel("Empirical CDF")
    plt.title(f"{title_prefix} P-P Plot")
    plt.grid(True)

    # -------- Q-Q Plot --------
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)

    plt.subplot(1, 2, 2)
    plt.scatter(theo_q, data_sorted, alpha=0.6)
    min_q = min(theo_q.min(), data_sorted.min())
    max_q = max(theo_q.max(), data_sorted.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Empirical Quantiles")
    plt.title(f"{title_prefix} Q-Q Plot")
    plt.grid(True)

    plt.tight_layout()
    plt.show()


# =====================================================
# 1. ÏÑúÎπÑÏä§ÌÉÄÏûÑ / ÎèÑÏ∞©ÏãúÍ∞Ñ ‚Üí Ï¥à Îã®ÏúÑ Î∞∞Ïó¥Î°ú Î≥ÄÌôò
# =====================================================

# --- Counter 1 ÏÑúÎπÑÏä§ÌÉÄÏûÑ (Ï¥à) ---
c1_1006_service_secs = to_seconds_list(counter1_1006_service_time)
c1_1007_service_secs = to_seconds_list(counter1_1007_service_time)
c1_1008_service_secs = to_seconds_list(counter1_1008_service_time)
c1_1015_service_secs = to_seconds_list(counter1_1015_service_time)

# --- Counter 2 ÏÑúÎπÑÏä§ÌÉÄÏûÑ (Ï¥à) ---
c2_1006_service_secs = to_seconds_list(counter2_1006_service_time)
c2_1007_service_secs = to_seconds_list(counter2_1007_service_time)
c2_1008_service_secs = to_seconds_list(counter2_1008_service_time)
c2_1015_service_secs = to_seconds_list(counter2_1015_service_time)

# --- Counter 1 ÎèÑÏ∞©ÏãúÍ∞Ñ (Ï¥à) ---
a1_1006_secs = np.sort(to_seconds_list(counter1_1006_arrival_time))
a1_1007_secs = np.sort(to_seconds_list(counter1_1007_arrival_time))
a1_1008_secs = np.sort(to_seconds_list(counter1_1008_arrival_time))
a1_1015_secs = np.sort(to_seconds_list(counter1_1015_arrival_time))

# --- Counter 2 ÎèÑÏ∞©ÏãúÍ∞Ñ (Ï¥à) ---
a2_1006_secs = np.sort(to_seconds_list(counter2_1006_arrival_time))
a2_1007_secs = np.sort(to_seconds_list(counter2_1007_arrival_time))
a2_1008_secs = np.sort(to_seconds_list(counter2_1008_arrival_time))
a2_1015_secs = np.sort(to_seconds_list(counter2_1015_arrival_time))

# --- Interarrival time (Ï¥à) Í≥ÑÏÇ∞: Í∞ôÏùÄ Ïπ¥Ïö¥ÌÑ∞ ÎÇ¥ÏóêÏÑú Ïó∞ÏÜç ÎèÑÏ∞© Í∞ÑÍ≤© ---
ia1_1006 = np.diff(a1_1006_secs)
ia1_1007 = np.diff(a1_1007_secs)
ia1_1008 = np.diff(a1_1008_secs)
ia1_1015 = np.diff(a1_1015_secs)

ia2_1006 = np.diff(a2_1006_secs)
ia2_1007 = np.diff(a2_1007_secs)
ia2_1008 = np.diff(a2_1008_secs)
ia2_1015 = np.diff(a2_1015_secs)

# 'ÌïòÎ£® Ï†ÑÏ≤¥' interarrivalÏùÄ Îëê Ïπ¥Ïö¥ÌÑ∞ Ìï©Ïπú Í±∏ ÏÇ¨Ïö© (Îã®Ïàú concat)
ia_day_1006 = np.concatenate([ia1_1006, ia2_1006])
ia_day_1007 = np.concatenate([ia1_1007, ia2_1007])
ia_day_1008 = np.concatenate([ia1_1008, ia2_1008])
ia_day_1015 = np.concatenate([ia1_1015, ia2_1015])

# 4Ïùº Ï†ÑÏ≤¥ interarrival (ÎèôÏßàÏÑ± Í≤ÄÏ†ï ÌõÑ Ìïú Îç©Ïñ¥Î¶¨Î°ú Ïì∏ ÏòàÏ†ï)
ia_all = np.concatenate([ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015])

# =====================================================
# 2. ÏÑúÎπÑÏä§ÌÉÄÏûÑ: ÏùºÎ∞òÏùº vs ÌîºÌÅ¨Ïùº Î∂ÑÎ¶¨
# =====================================================

# ÏùºÎ∞òÏùº = 10/06, 10/07, 10/08 (Îëê Ïπ¥Ïö¥ÌÑ∞ Î™®Îëê)
st_normal = np.concatenate([
    c1_1006_service_secs, c1_1007_service_secs, c1_1008_service_secs,
    c2_1006_service_secs, c2_1007_service_secs, c2_1008_service_secs
])

# ÌîºÌÅ¨Ïùº = 10/15 (Îëê Ïπ¥Ïö¥ÌÑ∞ Î™®Îëê)
st_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# ÏùºÏûêÎ≥Ñ ÏÑúÎπÑÏä§ÌÉÄÏûÑ (ÎèôÏßàÏÑ± Í≤ÄÏ†ïÏö©)
st_day_1006 = np.concatenate([c1_1006_service_secs, c2_1006_service_secs])
st_day_1007 = np.concatenate([c1_1007_service_secs, c2_1007_service_secs])
st_day_1008 = np.concatenate([c1_1008_service_secs, c2_1008_service_secs])
st_day_1015 = np.concatenate([c1_1015_service_secs, c2_1015_service_secs])

# =====================================================
# 3. ÎèÖÎ¶ΩÏÑ± Í≤ÄÏ†ï (ÏÑúÎπÑÏä§ÌÉÄÏûÑ: Counter1 vs Counter2)
#    - ÏÇ∞Ï†êÎèÑ + Spearman ÏÉÅÍ¥ÄÍ≥ÑÏàò
# =====================================================

# ÏòàÏãú 1: ÏùºÎ∞òÏùº(10/07) ÏÑúÎπÑÏä§ÌÉÄÏûÑ ÎèÖÎ¶ΩÏÑ±
min_len_0707 = min(len(c1_1007_service_secs), len(c2_1007_service_secs))
x_0707 = c1_1007_service_secs[:min_len_0707]
y_0707 = c2_1007_service_secs[:min_len_0707]

plt.figure(figsize=(5, 4))
plt.scatter(x_0707, y_0707, alpha=0.6)
plt.xlabel("Counter1 Service Time (s) - 10/07")
plt.ylabel("Counter2 Service Time (s) - 10/07")
plt.title("Scatter: Service Time (Non-Peak Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_0707, p_0707 = stats.spearmanr(x_0707, y_0707)
print(f"[Spearman] 10/07 ServiceTime C1 vs C2: rho={rho_0707:.3f}, p={p_0707:.4f}")

# ÏòàÏãú 2: ÌîºÌÅ¨Ïùº(10/15) ÏÑúÎπÑÏä§ÌÉÄÏûÑ ÎèÖÎ¶ΩÏÑ±
min_len_1015 = min(len(c1_1015_service_secs), len(c2_1015_service_secs))
x_1015 = c1_1015_service_secs[:min_len_1015]
y_1015 = c2_1015_service_secs[:min_len_1015]

plt.figure(figsize=(5, 4))
plt.scatter(x_1015, y_1015, alpha=0.6, color='orange')
plt.xlabel("Counter1 Service Time (s) - 10/15")
plt.ylabel("Counter2 Service Time (s) - 10/15")
plt.title("Scatter: Service Time (Peak Day Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_1015, p_1015 = stats.spearmanr(x_1015, y_1015)
print(f"[Spearman] 10/15 ServiceTime C1 vs C2: rho={rho_1015:.3f}, p={p_1015:.4f}")

# (ÏõêÌïòÎ©¥ interarrivalÏóê ÎåÄÌï¥ÏÑúÎèÑ Í∞ôÏùÄ Î∞©ÏãùÏúºÎ°ú Spearman ÎèåÎ¶¥ Ïàò ÏûàÏùå)


# =====================================================
# 4. ÎèôÏßàÏÑ± Í≤ÄÏ†ï (Ïó¨Îü¨ ÎÇ† Îç∞Ïù¥ÌÑ∞ Ìï©Ï≥êÎèÑ ÎêòÎäîÏßÄ)
#    - Interarrival, Service Îëò Îã§ Kruskal‚ÄìWallis ÏÇ¨Ïö©
# =====================================================

# Interarrival 4Ïùº ÎèôÏßàÏÑ± (Îëê Ïπ¥Ïö¥ÌÑ∞ Ìï©Ïπú ÌïòÎ£® Îã®ÏúÑ)
H_ia, p_ia = stats.kruskal(ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015)
print(f"\n[Interarrival] Kruskal‚ÄìWallis: H={H_ia:.3f}, p={p_ia:.6f}")

# Service 4Ïùº ÎèôÏßàÏÑ±
H_st, p_st = stats.kruskal(st_day_1006, st_day_1007, st_day_1008, st_day_1015)
print(f"[Service] Kruskal‚ÄìWallis: H={H_st:.3f}, p={p_st:.6f}")

# p_ia > 0.05 ÎùºÎ©¥ ‚Üí interarrivalÎäî 4Ïùº Ìï©Ï≥êÏÑú ÏÇ¨Ïö© OK
# p_st << 0.05 ÎùºÎ©¥ ‚Üí ÏÑúÎπÑÏä§ÌÉÄÏûÑÏùÄ ÏùºÎ∞òÏùº vs ÌîºÌÅ¨Ïùº Î∂ÑÎ¶¨ ÌïÑÏöî


# =====================================================
# 5. Tukey boxplotÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌôïÏù∏ (Service Normal / Peak)
# =====================================================

for data, name in [(st_normal, "Normal Service Time"),
                   (st_peak, "Peak Service Time")]:

    low, up = tukey_bounds(data)
    outliers = data[(data < low) | (data > up)]

    print(f"\n[{name}] Tukey bounds: {low:.1f} ~ {up:.1f}")
    print(f"  Ïù¥ÏÉÅÏπò Í∞úÏàò: {len(outliers)}")

    plt.figure(figsize=(4, 5))
    plt.boxplot(data, vert=True, showfliers=True)
    plt.title(f"{name} (Tukey boxplot)")
    plt.ylabel("Service Time (s)")
    plt.tight_layout()
    plt.show()


# =====================================================
# 6. Î∂ÑÌè¨ ÌîºÌåÖ Activity 1
#    - Interarrival Ï†ÑÏ≤¥
#    - Service Normal (10/06~08)
#    - Service Peak (10/15)
# =====================================================

res_ia       = fit_and_compare_distributions(ia_all, label="Interarrival (All days)")
res_st_norm  = fit_and_compare_distributions(st_normal, label="Service Time - Normal Days")
res_st_peak  = fit_and_compare_distributions(st_peak, label="Service Time - Peak Day")

best_ia      = res_ia[0]
best_norm    = res_st_norm[0]
best_peak    = res_st_peak[0]

print("\n[ÏÑ†Ï†ïÎêú ÏµúÏÉÅÏúÑ ÌõÑÎ≥¥ Î∂ÑÌè¨Îì§]")
print("Interarrival :", best_ia["name"],   best_ia["params"])
print("Service Normal:", best_norm["name"], best_norm["params"])
print("Service Peak  :", best_peak["name"], best_peak["params"])


# =====================================================
# 7. ÏÑ†Ï†ïÎêú Î∂ÑÌè¨Ïóê ÎåÄÌïú P-P / Q-Q Plot (Ï†ïÍµêÌïú Ï†ÅÌï©ÎèÑ ÌôïÏù∏)
# =====================================================

# Interarrival
plot_pp_qq(ia_all, best_ia["dist"], best_ia["params"],
           title_prefix=f"Interarrival ({best_ia['name']})")

# Service Normal
plot_pp_qq(st_normal, best_norm["dist"], best_norm["params"],
           title_prefix=f"Service Normal ({best_norm['name']})")

# Service Peak
plot_pp_qq(st_peak, best_peak["dist"], best_peak["params"],
           title_prefix=f"Service Peak ({best_peak['name']})")


# =====================================================
# 8. (Ï∞∏Í≥†) SimPyÏóêÏÑú Ïì∏ Ïàò ÏûàÎäî ÌòïÌÉúÎ°ú ÌååÎùºÎØ∏ÌÑ∞ Ï†ïÎ¶¨ ÏòàÏãú
#    - Ïó¨Í∏∞ÏÑúÎäî Í∑∏ÎÉ• printÎßå, Ïã§Ï†ú SimPy ÏΩîÎìúÏóêÏÑú rvs()Î°ú ÏÇ¨Ïö©
# =====================================================

# Ïòà: interarrivalÍ∞Ä exponÏù¥Î©¥
if best_ia["name"] == "expon":
    loc_ia, scale_ia = best_ia["params"][1], best_ia["params"][2]
    print("\n[SimPyÏö© Interarrival ÏòàÏãú] stats.expon.rvs(loc=%.2f, scale=%.2f)" % (loc_ia, scale_ia))

# Ïòà: ÏùºÎ∞òÏùº ÏÑúÎπÑÏä§ÌÉÄÏûÑÏù¥ gammaÏù∏ Í≤ΩÏö∞
if best_norm["name"] == "gamma":
    shape_n, loc_n, scale_n = best_norm["params"]
    print("[SimPyÏö© Service Normal ÏòàÏãú] stats.gamma.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (shape_n, loc_n, scale_n))

# Ïòà: ÌîºÌÅ¨Ïùº ÏÑúÎπÑÏä§ÌÉÄÏûÑÏù¥ lognormÏù∏ Í≤ΩÏö∞
if best_peak["name"] == "lognorm":
    s_p, loc_p, scale_p = best_peak["params"]
    print("[SimPyÏö© Service Peak ÏòàÏãú] stats.lognorm.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (s_p, loc_p, scale_p))
```
