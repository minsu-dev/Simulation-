```python
import numpy as np
from scipy.stats import kruskal
import scikit_posthocs as sp

# -------------------------------
# ÏãúÍ∞Ñ -> Ï¥à Î≥ÄÌôò Ìï®Ïàò
# -------------------------------
def t2s(t):
    h, m, s = map(int, t.split(':'))
    return h*3600 + m*60 + s

def conv(lst):
    return np.array([t2s(x) for x in lst])

# -------------------------------
# Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò (Counter 1)
# -------------------------------
c1_06 = conv(counter1_1006_service_time)
c1_07 = conv(counter1_1007_service_time)
c1_08 = conv(counter1_1008_service_time)
c1_15 = conv(counter1_1015_service_time)

# -------------------------------
# Kruskal‚ÄìWallis Í≤ÄÏ†ï
# -------------------------------
stat, p = kruskal(c1_06, c1_07, c1_08, c1_15)

print("=== Counter 1 Ï†ÑÏ≤¥ 4Ïùº ÎèôÏßàÏÑ± Í≤ÄÏ†ï ===")
print("Statistic:", stat)
print("p-value :", p)

# -------------------------------
# Dunn's test (Bonferroni Î≥¥Ï†ï)
# -------------------------------
import pandas as pd

# Îç∞Ïù¥ÌÑ∞ Ìï©ÏπòÍ∏∞
all_data = np.concatenate([c1_06, c1_07, c1_08, c1_15])
groups = ['1006']*len(c1_06) + ['1007']*len(c1_07) + ['1008']*len(c1_08) + ['1015']*len(c1_15)

df = pd.DataFrame({'time_sec': all_data, 'date': groups})

dunn = sp.posthoc_dunn(df, val_col='time_sec', group_col='date', p_adjust='bonferroni')

print("\n=== Dunn's Test (Bonferroni corrected p-values) ===")
print(dunn)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# -------------------------------
# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
# -------------------------------
# Counter 1 Ï†ÑÏ≤¥
c1_all = np.concatenate([c1_06, c1_07, c1_08, c1_15])  

# Counter 2 ÎπÑÌîºÌÅ¨: 10/07, 10/08
c2_nonpeak = np.concatenate([c2_07, c2_08])

# Counter 2 ÌîºÌÅ¨: 10/06, 10/15
c2_peak = np.concatenate([c2_06, c2_15])

# ÏÉòÌîå Í∏∏Ïù¥Î•º ÎßûÏ∂∞Ï£ºÍ∏∞ (shortest length Í∏∞Ï§Ä)
min_len_nonpeak = min(len(c1_all), len(c2_nonpeak))
min_len_peak = min(len(c1_all), len(c2_peak))

c1_nonpeak = c1_all[:min_len_nonpeak]
c2_nonpeak = c2_nonpeak[:min_len_nonpeak]

c1_peak = c1_all[:min_len_peak]
c2_peak = c2_peak[:min_len_peak]

# -------------------------------
# 1Ô∏è‚É£ ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
# -------------------------------
plt.figure(figsize=(12,5))

# ÎπÑÌîºÌÅ¨
plt.subplot(1,2,1)
plt.scatter(c1_nonpeak, c2_nonpeak, alpha=0.6)
plt.xlabel('Counter 1 Service Time (s)')
plt.ylabel('Counter 2 Non-Peak Service Time (s)')
plt.title('Scatter: Counter 1 vs Counter 2 Non-Peak')

# ÌîºÌÅ¨
plt.subplot(1,2,2)
plt.scatter(c1_peak, c2_peak, alpha=0.6, color='red')
plt.xlabel('Counter 1 Service Time (s)')
plt.ylabel('Counter 2 Peak Service Time (s)')
plt.title('Scatter: Counter 1 vs Counter 2 Peak')

plt.tight_layout()
plt.show()

# -------------------------------
# 2Ô∏è‚É£ ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
# -------------------------------
rho_nonpeak, pval_nonpeak = spearmanr(c1_nonpeak, c2_nonpeak)
rho_peak, pval_peak = spearmanr(c1_peak, c2_peak)

print("=== Spearman Correlation ===")
print(f"Non-Peak: rho={rho_nonpeak:.3f}, p-value={pval_nonpeak:.5f}")
print(f"Peak    : rho={rho_peak:.3f}, p-value={pval_peak:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Ïòà: ÏÑúÎπÑÏä§ ÌÉÄÏûÑ Îç∞Ïù¥ÌÑ∞ (numpy array)
data = c1_all  # ÎòêÎäî c2_nonpeak, c2_peak Îì± ÏõêÌïòÎäî Îç∞Ïù¥ÌÑ∞Î°ú Î≥ÄÍ≤Ω

# 1) Í∞êÎßàÎ∂ÑÌè¨ ÌååÎùºÎØ∏ÌÑ∞ Ï∂îÏ†ï (shape, loc, scale)
shape, loc, scale = gamma.fit(data, floc=0)  # loc=0 Í≥†Ï†ïÌï¥ÏÑú fitting (ÏÑúÎπÑÏä§ÌÉÄÏûÑÏùÄ 0 Ïù¥ÏÉÅ)

# 2) ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Observed data')

# 3) Í∞êÎßàÎ∂ÑÌè¨ PDF Í∑∏Î¶¨Í∏∞
x = np.linspace(min(data), max(data), 1000)
pdf_fitted = gamma.pdf(x, shape, loc=loc, scale=scale)
plt.plot(x, pdf_fitted, 'r-', lw=2, label='Fitted Gamma PDF')

plt.xlabel('Service Time (s)')
plt.ylabel('Density')
plt.title('Histogram and Fitted Gamma Distribution')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Í∑∏Î£πÎ≥Ñ Îç∞Ïù¥ÌÑ∞ Î¶¨Ïä§Ìä∏ÏôÄ Ï†úÎ™©
groups = [
    (c1_all, 'Counter 1 (All)'),
    (c2_nonpeak, 'Counter 2 Non-Peak'),
    (c2_peak, 'Counter 2 Peak')
]

plt.figure(figsize=(18, 5))

for i, (data, title) in enumerate(groups, 1):
    # Í∞êÎßàÎ∂ÑÌè¨ ÌååÎùºÎØ∏ÌÑ∞ Ï∂îÏ†ï (loc=0 Í≥†Ï†ï)
    shape, loc, scale = gamma.fit(data, floc=0)
    
    # ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞ (Î∞ÄÎèÑ Í∏∞Ï§Ä)
    plt.subplot(1, 3, i)
    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Observed data')
    
    # Í∞êÎßàÎ∂ÑÌè¨ PDF Í≥ÑÏÇ∞
    x = np.linspace(min(data), max(data), 1000)
    pdf_fitted = gamma.pdf(x, shape, loc=loc, scale=scale)
    plt.plot(x, pdf_fitted, 'r-', lw=2, label='Fitted Gamma PDF')
    
    plt.title(title)
    plt.xlabel('Service Time (s)')
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```
