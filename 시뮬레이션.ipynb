```python
# ================== simulation_normal_30runs.py ==================
import simpy
import numpy as np
from scipy.stats import gamma

# ======================================================
# 0. 분포 파라미터 (분석 결과 반영)
# ======================================================

# Interarrival time 4일 전체 통합 (Gamma)
IA_SHAPE = 0.6852387150017172
IA_LOC   = -13.117809568004514
IA_SCALE = 67.64591758869439

# Service time (Normal Days: 10/06~10/08)
SV_SHAPE = 0.35278743478169966
SV_LOC   = -20.992822911041664
SV_SCALE = 79.05467986513663

# 시뮬레이션 시간 (3시간 = 10,800초)
SIM_TIME = 5400

# 반복 횟수
N_REP = 30


# ======================================================
# 1. 고객 객체 (대기시간/체류시간 저장용)
# ======================================================

class Customer:
    def __init__(self, name, arrival_time):
        self.name = name
        self.arrival_time = arrival_time
        self.service_start = None
        self.service_end = None

    @property
    def waiting_time(self):
        return self.service_start - self.arrival_time

    @property
    def system_time(self):
        return self.service_end - self.arrival_time


# ======================================================
# 2. 고객 프로세스 (수업시간 코드 구조 동일)
# ======================================================

def customer(env, name, server, service_frozen, finished):
    arrival_time = env.now
    cust = Customer(name, arrival_time)

    # 자원 요청
    with server.request() as req:
        yield req

        # 서비스 시작
        cust.service_start = env.now
        service_time = service_frozen.rvs()

        if service_time <= 0:
            service_time = 0.1

        yield env.timeout(service_time)

        # 종료 처리
        cust.service_end = env.now
        finished.append(cust)


# ======================================================
# 3. 고객 생성 프로세스 (interarrival 분포 기반)
# ======================================================

def customer_generator(env, server, ia_frozen, service_frozen, finished):
    i = 0
    while True:
        inter = ia_frozen.rvs()
        if inter <= 0:
            inter = 0.1
        yield env.timeout(inter)

        i += 1
        env.process(customer(env, f"Customer {i}", server,
                             service_frozen, finished))


# ======================================================
# 4. 시뮬레이션 1회 실행 함수
# ======================================================

def run_one_rep(seed):

    env = simpy.Environment()
    server = simpy.Resource(env, capacity=1)

    finished = []

    # "동결된" 분포 만들기 (rv_frozen 객체)
    ia_frozen      = gamma(IA_SHAPE, IA_LOC, IA_SCALE)
    service_frozen = gamma(SV_SHAPE, SV_LOC, SV_SCALE)

    # 고객 생성 프로세스 실행
    env.process(customer_generator(env, server, ia_frozen,
                                   service_frozen, finished))

    env.run(until=SIM_TIME)

    waits = [c.waiting_time for c in finished]
    sys_times = [c.system_time for c in finished]

    return {
        "n": len(waits),
        "avg_wait": np.mean(waits),
        "avg_sys": np.mean(sys_times),
        "max_wait": np.max(waits)
    }


# ======================================================
# 5. 30회 반복 실행 (Normal)
# ======================================================

if __name__ == "__main__":
    results = []

    for r in range(N_REP):
        # seed number가 42에 고정되지 않고, 매 반복마다 바뀌도록 설정
        seed = r + 1          # 1, 2, 3, ..., 30
        res = run_one_rep(seed)
        results.append(res)

        print(f"[Normal] Rep {r+1:2d} → "
              f"완료고객={res['n']:4d}, "
              f"평균대기={res['avg_wait']:7.2f}, "
              f"평균체류={res['avg_sys']:7.2f}, "
              f"최대대기={res['max_wait']:7.2f}")


    # 전체 요약
    print("\n===== Normal(10/06~10/08) 요약 =====")
    print(f"평균 완료고객     : {np.mean([r['n'] for r in results]):.2f}")
    print(f"평균 대기시간     : {np.mean([r['avg_wait'] for r in results]):.2f}")
    print(f"평균 체류시간     : {np.mean([r['avg_sys'] for r in results]):.2f}")
    print(f"평균 최대대기시간 : {np.mean([r['max_wait'] for r in results]):.2f}")
```

# 피크일 일때의 시뮬레이션 모델

```python
# ================== simulation_peak_30runs.py ==================
import simpy
import numpy as np
from scipy.stats import gamma

# ======================================================
# 0. 분포 파라미터 (분석 결과 반영)
# ======================================================

# Interarrival time 4일 전체 통합 (Gamma)
IA_SHAPE = 0.6852387150017172
IA_LOC   = -13.117809568004514
IA_SCALE = 67.64591758869439

# Service time (Peak Day: 10/15)
SV_SHAPE = 0.5054710018301837
SV_LOC   = -9.275558523962127
SV_SCALE = 57.957740322769766


# 시뮬레이션 시간 (3시간 = 10,800초)
SIM_TIME = 3 * 60 * 60

# 반복 횟수
N_REP = 30


# ======================================================
# 1. 고객 객체 (대기시간/체류시간 저장용)
# ======================================================

class Customer:
    def __init__(self, name, arrival_time):
        self.name = name
        self.arrival_time = arrival_time
        self.service_start = None
        self.service_end = None

    @property
    def waiting_time(self):
        return self.service_start - self.arrival_time

    @property
    def system_time(self):
        return self.service_end - self.arrival_time


# ======================================================
# 2. 고객 프로세스 (수업시간 코드 구조 동일)
# ======================================================

def customer(env, name, server, service_frozen, finished):
    arrival_time = env.now
    cust = Customer(name, arrival_time)

    # 서버 자원 요청
    with server.request() as req:
        yield req

        # 서비스 시작
        cust.service_start = env.now
        service_time = service_frozen.rvs()

        if service_time <= 0:
            service_time = 0.1

        # 서비스 수행
        yield env.timeout(service_time)

        # 종료 처리
        cust.service_end = env.now
        finished.append(cust)


# ======================================================
# 3. 고객 생성 프로세스 (interarrival 분포 기반)
# ======================================================

def customer_generator(env, server, ia_frozen, service_frozen, finished):
    i = 0
    while True:
        inter = ia_frozen.rvs()
        if inter <= 0:
            inter = 0.1

        # 다음 고객 도착까지 대기
        yield env.timeout(inter)

        i += 1
        env.process(customer(env, f"Customer {i}", server,
                             service_frozen, finished))


# ======================================================
# 4. 시뮬레이션 1회 실행 함수
# ======================================================

def run_one_rep(seed):

    env = simpy.Environment()
    server = simpy.Resource(env, capacity=1)

    finished = []

    # "동결된" 분포 만들기 (rv_frozen 객체)
    ia_frozen      = gamma(IA_SHAPE, IA_LOC, IA_SCALE)
    service_frozen = gamma(SV_SHAPE, SV_LOC, SV_SCALE)

    # 고객 생성 프로세스 실행
    env.process(customer_generator(env, server, ia_frozen,
                                   service_frozen, finished))

    env.run(until=SIM_TIME)

    waits = [c.waiting_time for c in finished]
    sys_times = [c.system_time for c in finished]

    return {
        "n": len(waits),
        "avg_wait": np.mean(waits),
        "avg_sys": np.mean(sys_times),
        "max_wait": np.max(waits)
    }


# ======================================================
# 5. 30회 반복 실행 (Peak Day)
# ======================================================

if __name__ == "__main__":
    results = []

    for r in range(N_REP):
        seed = r + 100   # Normal과 겹치지 않게 100,101,... 로 줘도 됨
        res = run_one_rep(seed)
        results.append(res)

        print(f"[Peak] Rep {r+1:2d} → "
              f"완료고객={res['n']:4d}, "
              f"평균대기={res['avg_wait']:7.2f}, "
              f"평균체류={res['avg_sys']:7.2f}, "
              f"최대대기={res['max_wait']:7.2f}")

    # 전체 요약
    print("\n===== Peak(10/15) 요약 =====")
    print(f"평균 완료고객     : {np.mean([r['n'] for r in results]):.2f}")
    print(f"평균 대기시간     : {np.mean([r['avg_wait'] for r in results]):.2f}")
    print(f"평균 체류시간     : {np.mean([r['avg_sys'] for r in results]):.2f}")
    print(f"평균 최대대기시간 : {np.mean([r['max_wait'] for r in results]):.2f}")
```
