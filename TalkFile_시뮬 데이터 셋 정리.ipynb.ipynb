1. 공통 함수 + interarrival 계산 함수

```python
from scipy.stats import kruskal

# HH:MM:SS → 초 단위로 변환
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 도착시간 리스트 → 정렬 → interarrival time (초) 리스트 반환
def get_interarrival(arrival_time_list):
    # 1) 초단위로 변환
    secs = [hhmmss_to_sec(t) for t in arrival_time_list]
    # 2) 혹시 순서가 뒤섞였을 수 있으니 정렬
    secs_sorted = sorted(secs)
    # 3) 인접 차이 = interarrival
    interarrival = [
        secs_sorted[i] - secs_sorted[i-1]
        for i in range(1, len(secs_sorted))
    ]
    return interarrival
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# =====================================
# 1. HH:MM:SS → 초 변환 함수
# =====================================
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# =====================================
# 2. Interarrival 계산 함수
# =====================================
def interarrival_from_times(time_list):
    """
    time_list: ['14:32:36', '14:35:11', ...]
    return: 연속 도착 간격(초) np.array, 길이는 n-1
    """
    secs = np.array([hhmmss_to_sec(t) for t in time_list], dtype=float)
    ia = np.diff(secs)   # 앞뒤 시간 차이
    return ia

# =====================================
# 3. 기술통계 함수
# =====================================
def describe_data(data, label="Data"):
    data = np.array(data, dtype=float)
    data = data[~np.isnan(data)]

    n = len(data)
    if n == 0:
        print(f"\n[{label}] 데이터 없음")
        return

    mean = np.mean(data)
    std = np.std(data, ddof=1)          # 표본 표준편차
    min_v = np.min(data)
    q1 = np.percentile(data, 25)
    median = np.percentile(data, 50)
    q3 = np.percentile(data, 75)
    max_v = np.max(data)
    iqr = q3 - q1
    cv = std / mean if mean != 0 else np.nan
    skew = stats.skew(data)
    kurt = stats.kurtosis(data)         # excess kurtosis

    print(f"\n===== [{label}] 기술통계 =====")
    print(f"N            : {n}")
    print(f"Mean         : {mean:8.3f}")
    print(f"Std          : {std:8.3f}")
    print(f"Min          : {min_v:8.3f}")
    print(f"Q1 (25%)     : {q1:8.3f}")
    print(f"Median (50%) : {median:8.3f}")
    print(f"Q3 (75%)     : {q3:8.3f}")
    print(f"Max          : {max_v:8.3f}")
    print(f"IQR          : {iqr:8.3f}")
    print(f"CV           : {cv:8.3f}")
    print(f"Skewness     : {skew:8.3f}")
    print(f"Kurtosis     : {kurt:8.3f}")

# =====================================
# 4. 서비스시간을 초로 변환
# =====================================

# 10/06
c1_1006_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1006_service_time])
c2_1006_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1006_service_time])

# 10/07
c1_1007_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1007_service_time])
c2_1007_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1007_service_time])

# 10/08
c1_1008_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1008_service_time])
c2_1008_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1008_service_time])

# 10/15
c1_1015_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1015_service_time])
c2_1015_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1015_service_time])

# =====================================
# 5. Interarrival Time 계산 (도착시간 기반)
# =====================================

# 10/06
ia_c1_1006 = interarrival_from_times(counter1_1006_arrival_time)
ia_c2_1006 = interarrival_from_times(counter2_1006_arrival_time)

# 10/07
ia_c1_1007 = interarrival_from_times(counter1_1007_arrival_time)
ia_c2_1007 = interarrival_from_times(counter2_1007_arrival_time)

# 10/08
ia_c1_1008 = interarrival_from_times(counter1_1008_arrival_time)
ia_c2_1008 = interarrival_from_times(counter2_1008_arrival_time)

# 10/15
ia_c1_1015 = interarrival_from_times(counter1_1015_arrival_time)
ia_c2_1015 = interarrival_from_times(counter2_1015_arrival_time)

# =====================================
# 6. 분석용 그룹 묶기
# =====================================

# 6-1. 서비스시간: 일반일(10/06~10/08) vs 피크일(10/15)
service_normal = np.concatenate([
    c1_1006_service_secs, c2_1006_service_secs,
    c1_1007_service_secs, c2_1007_service_secs,
    c1_1008_service_secs, c2_1008_service_secs
])

service_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# 6-2. Interarrival: 전체 / 일반일 / 피크일
ia_normal = np.concatenate([
    ia_c1_1006, ia_c2_1006,
    ia_c1_1007, ia_c2_1007,
    ia_c1_1008, ia_c2_1008
])

ia_peak = np.concatenate([
    ia_c1_1015, ia_c2_1015
])

ia_all = np.concatenate([ia_normal, ia_peak])

# =====================================
# 7. 기술통계 출력
# =====================================

# --- 서비스타임 ---
describe_data(service_normal, "Service Time - Normal Days (10/06~10/08, C1+C2)")
describe_data(service_peak,   "Service Time - Peak Day (10/15, C1+C2)")

describe_data(c1_1006_service_secs, "Service Time - 10/06 Counter1")
describe_data(c2_1006_service_secs, "Service Time - 10/06 Counter2")

# --- Interarrival Time ---
describe_data(ia_all,    "Interarrival - All Days (C1+C2)")
describe_data(ia_normal, "Interarrival - Normal Days (10/06~10/08)")
describe_data(ia_peak,   "Interarrival - Peak Day (10/15)")

# =====================================
# 8. 시각화 함수 (히스토그램 + 박스플롯)
# =====================================

def plot_histogram(data_list, labels, title, xlabel):
    """
    data_list: [array1, array2, ...]
    labels   : ['normal', 'peak', ...]
    """
    plt.figure(figsize=(10, 5))
    for d, lab in zip(data_list, labels):
        plt.hist(d, bins=30, alpha=0.5, density=True, label=lab)
    plt.xlabel(xlabel)
    plt.ylabel("Density")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_boxplot(data_list, labels, title, ylabel):
    plt.figure(figsize=(8, 5))
    plt.boxplot(data_list, labels=labels, showfliers=True)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    plt.show()

# =====================================
# 9. 실제 그래프 출력
# =====================================

# 9-1. 서비스타임 히스토그램 & 박스플롯 (일반일 vs 피크일)
plot_histogram(
    [service_normal, service_peak],
    ["Normal Days", "Peak Day"],
    "Service Time Distribution (Normal vs Peak)",
    "Service Time (seconds)"
)

plot_boxplot(
    [service_normal, service_peak],
    ["Normal", "Peak"],
    "Service Time Boxplot (Normal vs Peak)",
    "Service Time (seconds)"
)

# 9-2. Interarrival 히스토그램 & 박스플롯 (일반일 vs 피크일)
plot_histogram(
    [ia_normal, ia_peak],
    ["Normal Days", "Peak Day"],
    "Interarrival Time Distribution (Normal vs Peak)",
    "Interarrival Time (seconds)"
)

plot_boxplot(
    [ia_normal, ia_peak],
    ["Normal", "Peak"],
    "Interarrival Time Boxplot (Normal vs Peak)",
    "Interarrival Time (seconds)"
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import gamma, lognorm, expon, weibull_min

# =====================================================
# 0. 공통 도우미 함수들
# =====================================================

def hhmmss_to_sec(t):
    """'HH:MM:SS' 문자열을 초 단위 숫자로 변환"""
    h, m, s = map(int, t.split(':'))
    return h*3600 + m*60 + s

def to_seconds_list(str_list):
    """시간 문자열 리스트를 초 단위 numpy 배열로 변환"""
    return np.array([hhmmss_to_sec(t) for t in str_list], dtype=float)

def tukey_bounds(data, k=1.5):
    """Tukey IQR 기준으로 이상치 경계 계산 (Q1-1.5IQR, Q3+1.5IQR)"""
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

def fit_and_compare_distributions(data, label=""):
    """
    여러 분포(expon, gamma, lognorm, weibull_min)를 피팅하고
    KS-test, loglik를 출력. p-value 기준 정렬하여 반환.
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]  # 0 이하 제거
    candidates = {
        "expon": expon,
        "gamma": gamma,
        "lognorm": lognorm,
        "weibull_min": weibull_min
    }

    results = []
    print(f"\n===== [{label}] 분포 적합 결과 =====")
    for name, dist in candidates.items():
        # 서비스/도착시간은 0 이상이므로 loc=0 고정
        params = dist.fit(data, floc=0)
        ks, p = stats.kstest(data, dist.cdf, args=params)
        loglik = np.sum(dist.logpdf(data, *params))
        results.append({
            "name": name,
            "dist": dist,
            "params": params,
            "ks": ks,
            "p": p,
            "loglik": loglik
        })
        print(f"{name:10s} | KS={ks:.4f}, p={p:.6f}, loglik={loglik:.1f}")
        print(f"  params = {params}\n")

    results.sort(key=lambda x: x["p"], reverse=True)

    print(f"\n>>> [{label}] p-value 순(적합도 순) <<<")
    for r in results:
        print(f"{r['name']:10s}  p={r['p']:.6f}, KS={r['ks']:.4f}")

    # 최적 분포 히스토그램 + PDF 시각화
    best = results[0]
    best_name = best["name"]
    best_dist = best["dist"]
    best_params = best["params"]

    x = np.linspace(data.min(), data.max(), 400)
    pdf = best_dist.pdf(x, *best_params)

    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=30, density=True, alpha=0.6,
             edgecolor='black', label='Observed')
    plt.plot(x, pdf, 'r-', lw=2, label=f'Best fit: {best_name}')
    plt.title(f"[{label}] Best Fitted: {best_name}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()


    return results

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    선택된 분포에 대해 P-P plot, Q-Q plot을 그려주는 함수.
    dist_params는 dist.fit 결과 튜플 (shape..., loc, scale)
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]
    data_sorted = np.sort(data)
    n = len(data_sorted)

    *shape_params, loc, scale = dist_params

    # -------- P-P Plot --------
    emp_cdf = (np.arange(1, n+1)) / (n + 0.0)
    theo_cdf = dist.cdf(data_sorted, *shape_params, loc=loc, scale=scale)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(theo_cdf, emp_cdf, alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel("Theoretical CDF")
    plt.ylabel("Empirical CDF")
    plt.title(f"{title_prefix} P-P Plot")
    plt.grid(True)

    # -------- Q-Q Plot --------
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)

    plt.subplot(1, 2, 2)
    plt.scatter(theo_q, data_sorted, alpha=0.6)
    min_q = min(theo_q.min(), data_sorted.min())
    max_q = max(theo_q.max(), data_sorted.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Empirical Quantiles")
    plt.title(f"{title_prefix} Q-Q Plot")
    plt.grid(True)

    plt.tight_layout()
    plt.show()


# =====================================================
# 1. 서비스타임 / 도착시간 → 초 단위 배열로 변환
# =====================================================

# --- Counter 1 서비스타임 (초) ---
c1_1006_service_secs = to_seconds_list(counter1_1006_service_time)
c1_1007_service_secs = to_seconds_list(counter1_1007_service_time)
c1_1008_service_secs = to_seconds_list(counter1_1008_service_time)
c1_1015_service_secs = to_seconds_list(counter1_1015_service_time)

# --- Counter 2 서비스타임 (초) ---
c2_1006_service_secs = to_seconds_list(counter2_1006_service_time)
c2_1007_service_secs = to_seconds_list(counter2_1007_service_time)
c2_1008_service_secs = to_seconds_list(counter2_1008_service_time)
c2_1015_service_secs = to_seconds_list(counter2_1015_service_time)

# --- Counter 1 도착시간 (초) ---
a1_1006_secs = np.sort(to_seconds_list(counter1_1006_arrival_time))
a1_1007_secs = np.sort(to_seconds_list(counter1_1007_arrival_time))
a1_1008_secs = np.sort(to_seconds_list(counter1_1008_arrival_time))
a1_1015_secs = np.sort(to_seconds_list(counter1_1015_arrival_time))

# --- Counter 2 도착시간 (초) ---
a2_1006_secs = np.sort(to_seconds_list(counter2_1006_arrival_time))
a2_1007_secs = np.sort(to_seconds_list(counter2_1007_arrival_time))
a2_1008_secs = np.sort(to_seconds_list(counter2_1008_arrival_time))
a2_1015_secs = np.sort(to_seconds_list(counter2_1015_arrival_time))

# --- Interarrival time (초) 계산: 같은 카운터 내에서 연속 도착 간격 ---
ia1_1006 = np.diff(a1_1006_secs)
ia1_1007 = np.diff(a1_1007_secs)
ia1_1008 = np.diff(a1_1008_secs)
ia1_1015 = np.diff(a1_1015_secs)

ia2_1006 = np.diff(a2_1006_secs)
ia2_1007 = np.diff(a2_1007_secs)
ia2_1008 = np.diff(a2_1008_secs)
ia2_1015 = np.diff(a2_1015_secs)

# '하루 전체' interarrival은 두 카운터 합친 걸 사용 (단순 concat)
ia_day_1006 = np.concatenate([ia1_1006, ia2_1006])
ia_day_1007 = np.concatenate([ia1_1007, ia2_1007])
ia_day_1008 = np.concatenate([ia1_1008, ia2_1008])
ia_day_1015 = np.concatenate([ia1_1015, ia2_1015])

# 4일 전체 interarrival (동질성 검정 후 한 덩어리로 쓸 예정)
ia_all = np.concatenate([ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015])

# =====================================================
# 2. 서비스타임: 일반일 vs 피크일 분리
# =====================================================

# 일반일 = 10/06, 10/07, 10/08 (두 카운터 모두)
st_normal = np.concatenate([
    c1_1006_service_secs, c1_1007_service_secs, c1_1008_service_secs,
    c2_1006_service_secs, c2_1007_service_secs, c2_1008_service_secs
])

# 피크일 = 10/15 (두 카운터 모두)
st_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# 일자별 서비스타임 (동질성 검정용)
st_day_1006 = np.concatenate([c1_1006_service_secs, c2_1006_service_secs])
st_day_1007 = np.concatenate([c1_1007_service_secs, c2_1007_service_secs])
st_day_1008 = np.concatenate([c1_1008_service_secs, c2_1008_service_secs])
st_day_1015 = np.concatenate([c1_1015_service_secs, c2_1015_service_secs])

# =====================================================
# 3. 독립성 검정 (서비스타임: Counter1 vs Counter2)
#    - 산점도 + Spearman 상관계수
# =====================================================

# 예시 1: 일반일(10/07) 서비스타임 독립성
min_len_0707 = min(len(c1_1007_service_secs), len(c2_1007_service_secs))
x_0707 = c1_1007_service_secs[:min_len_0707]
y_0707 = c2_1007_service_secs[:min_len_0707]

plt.figure(figsize=(5, 4))
plt.scatter(x_0707, y_0707, alpha=0.6)
plt.xlabel("Counter1 Service Time (s) - 10/07")
plt.ylabel("Counter2 Service Time (s) - 10/07")
plt.title("Scatter: Service Time (Non-Peak Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_0707, p_0707 = stats.spearmanr(x_0707, y_0707)
print(f"[Spearman] 10/07 ServiceTime C1 vs C2: rho={rho_0707:.3f}, p={p_0707:.4f}")

# 예시 2: 피크일(10/15) 서비스타임 독립성
min_len_1015 = min(len(c1_1015_service_secs), len(c2_1015_service_secs))
x_1015 = c1_1015_service_secs[:min_len_1015]
y_1015 = c2_1015_service_secs[:min_len_1015]

plt.figure(figsize=(5, 4))
plt.scatter(x_1015, y_1015, alpha=0.6, color='orange')
plt.xlabel("Counter1 Service Time (s) - 10/15")
plt.ylabel("Counter2 Service Time (s) - 10/15")
plt.title("Scatter: Service Time (Peak Day Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_1015, p_1015 = stats.spearmanr(x_1015, y_1015)
print(f"[Spearman] 10/15 ServiceTime C1 vs C2: rho={rho_1015:.3f}, p={p_1015:.4f}")

# =====================================================
# 9-b. Lagged Correlation Plot for Non-Peak & Peak Days
# =====================================================

from statsmodels.tsa.stattools import acf
import numpy as np
import matplotlib.pyplot as plt

# -------- Interarrival 분리 --------
ia_nonpeak = np.concatenate([ia_day_1006, ia_day_1007, ia_day_1008])
ia_peak    = ia_day_1015

def plot_acf_custom(data, nlags, title):
    acf_vals = acf(data, nlags=nlags, fft=True)
    lags = np.arange(nlags + 1)
    N = len(data)
    conf = 1.96 / np.sqrt(N)   # Confidence interval

    plt.figure(figsize=(10, 4))

    # Vertical lines + dots (stem 대체)
    plt.vlines(lags, 0, acf_vals, lw=1.3, alpha=0.8)
    plt.scatter(lags, acf_vals, s=28)

    # Confidence interval
    plt.hlines([0, conf, -conf], xmin=0, xmax=nlags,
               colors=['black', 'red', 'red'],
               linestyles=['solid', 'dashed', 'dashed'], linewidth=0.9)
    plt.fill_between(lags, conf, -conf, color='red', alpha=0.05)

    plt.title(title)
    plt.xlabel("Lag")
    plt.ylabel("Autocorrelation")
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Plot 실행
plot_acf_custom(ia_nonpeak, nlags=50,
                title="Lagged Correlation (ACF) - Non-Peak Days (10/06, 10/07, 10/08)")

plot_acf_custom(ia_peak, nlags=50,
                title="Lagged Correlation (ACF) - Peak Day (10/15)")

# (원하면 interarrival에 대해서도 같은 방식으로 Spearman 돌릴 수 있음)


# =====================================================
# 4. 동질성 검정 (여러 날 데이터 합쳐도 되는지)
#    - Interarrival, Service 둘 다 Kruskal–Wallis 사용
# =====================================================

# Interarrival 4일 동질성 (두 카운터 합친 하루 단위)
H_ia, p_ia = stats.kruskal(ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015)
print(f"\n[Interarrival] Kruskal–Wallis: H={H_ia:.3f}, p={p_ia:.6f}")

# =====================================================
# Interarrival 전체 산점도
# =====================================================
plt.figure(figsize=(10, 5))
x = np.arange(len(ia_all))
plt.scatter(x, ia_all, s=10, alpha=0.6)
plt.title("Scatter Plot of Interarrival Time (All Days)")
plt.xlabel("Arrival Index")
plt.ylabel("Interarrival Time (seconds)")
plt.grid(True)
plt.tight_layout()
plt.show()

# =====================================================
# 9. Lagged Correlation Plot (Autocorrelation)
# =====================================================

from statsmodels.tsa.stattools import acf
import numpy as np
import matplotlib.pyplot as plt

nlags = 50
acf_values = acf(ia_all, nlags=nlags, fft=True)
lags = np.arange(nlags + 1)

# 신뢰구간 (대략) : ±1.96/sqrt(N)
N = len(ia_all)
conf = 1.96 / np.sqrt(N)

plt.figure(figsize=(10, 4))

# 막대 느낌의 stem 대체: 수직선 + 점
plt.vlines(lags, 0, acf_values, lw=1.5, alpha=0.8)
plt.scatter(lags, acf_values, s=30, alpha=0.9)

# 신뢰구간 표시
plt.hlines([0, conf, -conf], xmin=0, xmax=nlags, colors=['black','red','red'],
           linestyles=['solid','dashed','dashed'], linewidth=0.8)
plt.fill_between(lags, conf, -conf, color='red', alpha=0.05)

plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.title("Lagged Correlation (ACF) Plot - Interarrival (All Days)")
plt.xlim(-0.5, nlags + 0.5)
plt.ylim(min(-1, acf_values.min()-0.05), 1.05)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# Service 4일 동질성
H_st, p_st = stats.kruskal(st_day_1006, st_day_1007, st_day_1008, st_day_1015)
print(f"[Service] Kruskal–Wallis: H={H_st:.3f}, p={p_st:.6f}")

# p_ia > 0.05 라면 → interarrival는 4일 합쳐서 사용 OK
# p_st << 0.05 라면 → 서비스타임은 일반일 vs 피크일 분리 필요


# =====================================================
# 5. Tukey boxplot으로 이상치 확인 (Service Normal / Peak)
# =====================================================

for data, name in [(st_normal, "Normal Service Time"),
                   (st_peak, "Peak Service Time")]:

    low, up = tukey_bounds(data)
    outliers = data[(data < low) | (data > up)]

    print(f"\n[{name}] Tukey bounds: {low:.1f} ~ {up:.1f}")
    print(f"  이상치 개수: {len(outliers)}")

    plt.figure(figsize=(4, 5))
    plt.boxplot(data, vert=True, showfliers=True)
    plt.title(f"{name} (Tukey boxplot)")
    plt.ylabel("Service Time (s)")
    plt.tight_layout()
    plt.show()


# =====================================================
# 6. 분포 피팅 Activity 1
#    - Interarrival 전체
#    - Service Normal (10/06~08)
#    - Service Peak (10/15)
# =====================================================

res_ia       = fit_and_compare_distributions(ia_all, label="Interarrival (All days)")
res_st_norm  = fit_and_compare_distributions(st_normal, label="Service Time - Normal Days")
res_st_peak  = fit_and_compare_distributions(st_peak, label="Service Time - Peak Day")

best_ia      = res_ia[0]
best_norm    = res_st_norm[0]
best_peak    = res_st_peak[0]

print("\n[선정된 최상위 후보 분포들]")
print("Interarrival :", best_ia["name"],   best_ia["params"])
print("Service Normal:", best_norm["name"], best_norm["params"])
print("Service Peak  :", best_peak["name"], best_peak["params"])


# =====================================================
# 7. 선정된 분포에 대한 P-P / Q-Q Plot (정교한 적합도 확인)
# =====================================================

# Interarrival
plot_pp_qq(ia_all, best_ia["dist"], best_ia["params"],
           title_prefix=f"Interarrival ({best_ia['name']})")

# Service Normal
plot_pp_qq(st_normal, best_norm["dist"], best_norm["params"],
           title_prefix=f"Service Normal ({best_norm['name']})")

# Service Peak
plot_pp_qq(st_peak, best_peak["dist"], best_peak["params"],
           title_prefix=f"Service Peak ({best_peak['name']})")


# =====================================================
# 8. (참고) SimPy에서 쓸 수 있는 형태로 파라미터 정리 예시
#    - 여기서는 그냥 print만, 실제 SimPy 코드에서 rvs()로 사용
# =====================================================

# 예: interarrival가 expon이면
if best_ia["name"] == "expon":
    loc_ia, scale_ia = best_ia["params"][1], best_ia["params"][2]
    print("\n[SimPy용 Interarrival 예시] stats.expon.rvs(loc=%.2f, scale=%.2f)" % (loc_ia, scale_ia))

# 예: 일반일 서비스타임이 gamma인 경우
if best_norm["name"] == "gamma":
    shape_n, loc_n, scale_n = best_norm["params"]
    print("[SimPy용 Service Normal 예시] stats.gamma.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (shape_n, loc_n, scale_n))

# 예: 피크일 서비스타임이 lognorm인 경우
if best_peak["name"] == "lognorm":
    s_p, loc_p, scale_p = best_peak["params"]
    print("[SimPy용 Service Peak 예시] stats.lognorm.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (s_p, loc_p, scale_p))
```

```python
import simpy
import numpy as np
from scipy import stats
import random

# =====================================================
# 1. (중요) 분포 파라미터 채워넣기 — 네 분석 결과로 바꿔줘야 하는 부분
# =====================================================

# 예시: 실제 best_ia, best_norm, best_peak에서 나온 값으로 수정하면 됨
INTERARRIVAL_DIST_NAME = "expon"
INTERARRIVAL_PARAMS    = (0.0, 30.0)       # 예: (loc, scale)

SERVICE_NORMAL_DIST_NAME = "gamma"
SERVICE_NORMAL_PARAMS    = (3.8869, 0.0, 16.4619)   # 예: (shape, loc, scale)

SERVICE_PEAK_DIST_NAME = "lognorm"
SERVICE_PEAK_PARAMS    = (0.6802, 0.0, 39.7381)     # 예: (s, loc, scale)

DIST_MAP = {
    "expon": stats.expon,
    "gamma": stats.gamma,
    "lognorm": stats.lognorm,
    "weibull_min": stats.weibull_min
}

INTERARRIVAL_DIST   = DIST_MAP[INTERARRIVAL_DIST_NAME]
SERVICE_NORMAL_DIST = DIST_MAP[SERVICE_NORMAL_DIST_NAME]
SERVICE_PEAK_DIST   = DIST_MAP[SERVICE_PEAK_DIST_NAME]


# =====================================================
# 2. 샘플링 함수들
# =====================================================

def sample_interarrival():
    """도착 간격 샘플링"""
    return INTERARRIVAL_DIST.rvs(*INTERARRIVAL_PARAMS)

def sample_service_time(scenario="normal"):
    """서비스 시간 샘플링"""
    if scenario == "normal":
        return SERVICE_NORMAL_DIST.rvs(*SERVICE_NORMAL_PARAMS)
    elif scenario == "peak":
        return SERVICE_PEAK_DIST.rvs(*SERVICE_PEAK_PARAMS)
    else:
        raise ValueError("scenario는 'normal' 또는 'peak'만 가능합니다.")


# =====================================================
# 3. 고객 프로세스
# =====================================================

def customer(env, name, counter, scenario, stats_recorder):
    arrival_time = env.now

    with counter.request() as req:
        yield req
        start = env.now
        waiting = start - arrival_time

        service_time = sample_service_time(scenario)
        yield env.timeout(service_time)
        end = env.now

    system_time = end - arrival_time

    stats_recorder["waiting_times"].append(waiting)
    stats_recorder["service_times"].append(service_time)
    stats_recorder["system_times"].append(system_time)


def customer_generator(env, counter, scenario, stats_recorder):
    i = 0
    while True:
        ia = sample_interarrival()
        yield env.timeout(ia)

        i += 1
        env.process(customer(env, f"Cust {i}", counter, scenario, stats_recorder))


# =====================================================
# 4. 시뮬레이션 실행 (1시간 30분 = 5400초)
# =====================================================

def run_simulation(scenario="normal", num_counters=2,
                   sim_time=5400, seed=42):
    """
    하나의 시나리오 + 카운터 수에 대해 1회 시뮬레이션 실행
    결과: 평균 대기시간, 시스템체류시간 등 리턴
    """
    random.seed(seed)
    np.random.seed(seed)

    env = simpy.Environment()
    counter = simpy.Resource(env, capacity=num_counters)

    stats_recorder = {
        "waiting_times": [],
        "service_times": [],
        "system_times": []
    }

    env.process(customer_generator(env, counter, scenario, stats_recorder))
    env.run(until=sim_time)

    wt = np.array(stats_recorder["waiting_times"])
    st = np.array(stats_recorder["service_times"])
    sys = np.array(stats_recorder["system_times"])

    results = {
        "scenario": scenario,
        "num_counters": num_counters,
        "num_customers": len(wt),
        "avg_waiting": wt.mean() if len(wt) else np.nan,
        "avg_service": st.mean() if len(st) else np.nan,
        "avg_system": sys.mean() if len(sys) else np.nan
    }
    return results


# =====================================================
# 5. 최적 카운터 수 찾는 함수 추가 ✅
# =====================================================

def find_optimal_counters(scenario, candidate_counters,
                          sim_time=5400, seed=42,
                          wait_threshold=None, verbose=True):
    """
    - scenario: "normal" 또는 "peak"
    - candidate_counters: [1,2,3,4] 이런 식의 후보 리스트
    - wait_threshold: 허용 가능한 '평균 대기시간(초)' 기준
        * 예: 60 이면 "평균 대기 <= 60초"를 만족하는
              가장 작은 카운터 수를 최적이라고 판단
        * None 이면 'avg_waiting이 최소'인 경우 선택
    - 반환값: (best_result, all_results 리스트)
    """
    all_results = []

    for n in candidate_counters:
        res = run_simulation(scenario=scenario,
                             num_counters=n,
                             sim_time=sim_time,
                             seed=seed)
        all_results.append(res)

        if verbose:
            print(f"[{scenario}] counters={n}, "
                  f"N={res['num_customers']}, "
                  f"Avg Wait={res['avg_waiting']:.2f}s, "
                  f"Avg System={res['avg_system']:.2f}s")

    # 1) 기준 대기시간이 주어졌다면 → 그걸 만족하는 최소 카운터 선택
    if wait_threshold is not None:
        eligible = [r for r in all_results if r["avg_waiting"] <= wait_threshold]
        if len(eligible) > 0:
            best = min(eligible, key=lambda r: r["num_counters"])
        else:
            # 기준을 만족하는 게 없다면, 그냥 가장 평균 대기시간이 짧은 케이스 선택
            best = min(all_results, key=lambda r: r["avg_waiting"])
    else:
        # 2) 기준이 없다면 → avg_waiting 최소인 케이스 선택
        best = min(all_results, key=lambda r: r["avg_waiting"])

    if verbose:
        print("\n>>> [", scenario, "] 최적 카운터 수 선택 결과 <<<")
        print(f"최적 counters = {best['num_counters']}개")
        print(f"  평균 대기시간 = {best['avg_waiting']:.2f}초")
        print(f"  평균 시스템체류시간 = {best['avg_system']:.2f}초")
        print(f"  처리 고객 수 = {best['num_customers']}명\n")

    return best, all_results


# =====================================================
# 6. 테스트 실행 예시 (최적 카운터 자동 계산)
# =====================================================

if __name__ == "__main__":
    # 예시 기준:
    # - 일반일: 평균 대기시간 60초 이하 목표
    # - 피크일: 평균 대기시간 30초 이하 목표 (좀 더 타이트하게)
    NORMAL_WAIT_TARGET = 60.0   # 1분
    PEAK_WAIT_TARGET   = 30.0   # 30초

    print("=== Normal Day 시나리오 ===")
    best_normal, normal_res = find_optimal_counters(
        scenario="normal",
        candidate_counters=[1, 2, 3, 4],
        sim_time=5400,
        seed=123,
        wait_threshold=NORMAL_WAIT_TARGET,
        verbose=True
    )

    print("=== Peak Day 시나리오 ===")
    best_peak, peak_res = find_optimal_counters(
        scenario="peak",
        candidate_counters=[2, 3, 4, 5],
        sim_time=5400,
        seed=123,
        wait_threshold=PEAK_WAIT_TARGET,
        verbose=True
    )

    print(">>> 종합 요약")
    print(f"일반일 최적 카운터 수: {best_normal['num_counters']}개 "
          f"(평균 대기 {best_normal['avg_waiting']:.2f}초)")
    print(f"피크일 최적 카운터 수: {best_peak['num_counters']}개 "
          f"(평균 대기 {best_peak['avg_waiting']:.2f}초)")
```

```python
# 10/06(월) - 1번 계산대 (총 58건)
counter1_1006_arrival_time = [
    '14:32:36','14:35:11','14:37:31','14:37:59','14:39:48','14:40:53','14:42:27','14:41:51','14:44:12','14:45:34',
    '14:48:07','14:48:09','14:49:57','14:50:35','14:52:46','14:53:51','14:55:55','14:57:19','14:57:52','15:01:40',
    '15:02:29','15:04:13','15:05:14','15:05:36','15:07:44','15:07:52','15:09:13','15:09:46','15:10:35','15:10:52',
    '15:33:44','15:34:53','15:35:13','15:37:02','15:37:04','15:38:21','15:38:46','15:40:23','15:42:02','15:42:50',
    '15:43:35','15:45:23','15:46:41','15:48:04','15:48:42','15:50:54','15:52:05','15:52:57','15:53:19','15:54:29',
    '15:55:13','15:57:01','15:59:14','15:59:17','16:01:24','16:01:48','16:02:03','16:02:52'
]

counter1_1006_wait_time = [
    '00:01:11','00:00:31','00:00:00','00:00:32','00:00:00','00:00:11','00:00:00','00:01:21','00:01:01','00:02:17',
    '00:01:39','00:02:26','00:01:26','00:01:33','00:01:04','00:00:56','00:00:00','00:00:00','00:00:28','00:00:00',
    '00:00:00','00:00:11','00:00:09','00:00:53','00:00:00','00:01:00','00:00:31','00:00:26','00:00:24','00:01:12',
    '00:00:00','00:00:00','00:00:48','00:00:56','00:01:59','00:01:30','00:02:02','00:01:18','00:01:03','00:02:05',
    '00:02:28','00:01:55','00:01:30','00:01:31','00:02:34','00:01:19','00:01:28','00:01:44','00:02:40','00:02:57',
    '00:02:51','00:01:40','00:00:02','00:01:51','00:00:40','00:01:37','00:02:33','00:03:27'
]

counter1_1006_service_time = [
    '00:01:15','00:01:13','00:01:00','00:00:38','00:01:16','00:00:42','00:00:45','00:02:01','00:02:38','00:01:55',
    '00:00:49','00:00:48','00:00:45','00:01:42','00:00:57','00:00:41','00:01:13','00:01:01','00:01:32','00:00:34',
    '00:01:55','00:00:59','00:01:06','00:00:50','00:01:08','00:00:52','00:00:28','00:00:47','00:01:05','00:01:17',
    '00:01:05','00:01:08','00:01:57','00:01:05','00:00:48','00:00:57','00:00:53','00:01:24','00:01:50','00:01:08',
    '00:01:15','00:00:53','00:01:24','00:01:41','00:00:57','00:01:20','00:01:08','00:01:18','00:01:27','00:00:38',
    '00:00:37','00:00:35','00:01:52','00:00:56','00:01:21','00:01:11','00:01:43','00:01:19'
]

counter1_1006_service_end_time = [
    '14:35:02','14:36:55','14:38:31','14:39:09','14:41:04','14:41:46','14:43:12','14:45:13','14:47:51','14:49:46',
    '14:50:35','14:51:23','14:52:08','14:53:50','14:54:47','14:55:28','14:57:08','14:58:20','14:59:52','15:02:14',
    '15:04:24','15:05:23','15:06:29','15:07:19','15:08:52','15:09:44','15:10:12','15:10:59','15:12:04','15:13:21',
    '15:34:49','15:36:01','15:37:58','15:39:03','15:39:51','15:40:48','15:41:41','15:43:05','15:44:55','15:46:03',
    '15:47:18','15:48:11','15:49:35','15:51:16','15:52:13','15:53:33','15:54:41','15:55:59','15:57:26','15:58:04',
    '15:58:41','15:59:16','16:01:08','16:02:04','16:03:25','16:04:36','16:06:19','16:07:38'
]

# 10/06(월) - 2번 계산대 (총 65건)
counter2_1006_arrival_time = [
    '14:32:48','14:33:25','14:34:58','14:37:24','14:38:15','14:40:11','14:42:48','14:44:02','14:44:54','14:45:23',
    '14:46:23','14:46:54','14:48:57','14:49:48','14:49:57','14:50:10','14:50:17','14:52:16','14:55:42','14:57:17',
    '14:59:38','15:00:18','15:03:59','15:04:58','15:06:32','15:07:53','15:08:54','15:10:29','15:11:23','15:11:57',
    '15:12:01','15:12:57','15:13:11','15:17:32','15:19:07','15:20:35','15:21:04','15:21:07','15:24:29','15:24:39',
    '15:26:42','15:27:40','15:29:11','15:29:17','15:29:35','15:30:30','15:32:51','15:33:09','15:34:10','15:35:07',
    '15:35:16','15:36:08','15:36:30','15:38:03','15:39:15','15:41:30','15:42:09','15:42:14','15:42:19','15:45:38',
    '15:48:17','15:48:30','15:49:50','15:50:02','15:52:51'
]
counter2_1006_wait_time = [
    '00:00:54','00:00:55','00:01:15','00:00:10','00:00:08','00:00:13','00:00:08','00:00:28','00:00:10','00:00:22',
    '00:01:30','00:01:13','00:00:22','00:00:46','00:00:54','00:01:11','00:01:40','00:00:20','00:00:15','00:00:29',
    '00:00:13','00:00:35','00:00:15','00:00:42','00:00:16','00:00:27','00:00:55','00:00:09','00:00:07','00:00:37',
    '00:01:47','00:02:04','00:02:45','00:00:22','00:00:17','00:00:13','00:00:25','00:01:35','00:00:09','00:00:41',
    '00:00:35','00:00:16','00:00:10','00:00:50','00:01:19','00:01:28','00:00:47','00:01:12','00:00:47','00:00:40',
    '00:01:59','00:01:54','00:02:56','00:02:19','00:01:41','00:00:37','00:02:11','00:02:40','00:03:50','00:01:07',
    '00:00:18','00:00:56','00:02:00','00:03:08','00:01:08'
]

counter2_1006_service_time = [
    '00:00:37','00:00:19','00:00:19','00:00:21','00:00:28','00:00:53','00:00:34','00:00:05','00:00:29','00:01:57',
    '00:00:09','00:01:08','00:00:59','00:00:10','00:00:15','00:00:25','00:00:26','00:00:46','00:00:47','00:00:22',
    '00:00:51','00:00:51','00:01:23','00:00:35','00:01:20','00:00:49','00:00:30','00:00:40','00:00:58','00:00:50',
    '00:01:08','00:00:49','00:00:34','00:00:43','00:01:01','00:00:32','00:01:01','00:00:38','00:00:34','00:01:33',
    '00:00:21','00:01:12','00:00:40','00:00:45','00:00:55','00:01:28','00:00:24','00:00:21','00:00:35','00:01:09',
    '00:00:29','00:00:48','00:00:29','00:00:25','00:00:26','00:02:08','00:00:20','00:01:00','00:00:21','00:00:32',
    '00:00:24','00:02:09','00:01:14','00:00:18','00:00:38'
]

counter2_1006_service_end_time = [
    '14:34:19','14:34:39','14:36:32','14:37:55','14:38:51','14:41:17','14:43:30','14:44:35','14:45:33','14:47:42',
    '14:48:02','14:49:15','14:50:18','14:50:44','14:51:06','14:51:46','14:52:23','14:53:22','14:56:44','14:58:08',
    '15:00:42','15:01:44','15:05:37','15:06:15','15:08:08','15:09:09','15:10:19','15:11:18','15:12:28','15:13:24',
    '15:14:56','15:15:50','15:16:30','15:18:37','15:20:25','15:21:20','15:22:30','15:23:20','15:25:12','15:26:53',
    '15:27:38','15:29:08','15:30:01','15:30:52','15:31:49','15:33:26','15:34:02','15:34:42','15:35:32','15:36:56',
    '15:37:44','15:38:50','15:39:55','15:40:47','15:41:22','15:44:15','15:44:40','15:45:54','15:46:30','15:47:17',
    '15:48:59','15:51:35','15:53:04','15:53:28','15:54:37'
]


# 10/07(화) - 1번 계산대 (총 74건, 38번 제외 -> 직원 교대칸)

counter1_1007_arrival_time = [
    '14:34:22','14:36:51','14:37:30','14:39:03','14:39:40','14:43:04','14:43:48','14:45:27','14:46:04','14:46:51',
    '14:48:14','14:49:49','14:50:26','14:51:41','14:54:20','14:54:26','14:56:36','14:58:11','14:58:12','14:58:56',
    '14:59:58','15:00:01','15:01:01','15:01:45','15:02:17','15:02:48','15:03:33','15:03:59','15:05:19','15:07:40',
    '15:07:50','15:08:00','15:09:04','15:09:47','15:11:00','15:11:47','15:12:10','15:15:08','15:15:09','15:17:16',
    '15:18:08','15:18:13','15:19:22','15:20:55','15:22:09','15:23:04','15:23:47','15:24:21','15:24:58','15:26:56',
    '15:29:10','15:30:38','15:32:10','15:32:46','15:34:48','15:35:22','15:35:58','15:36:30','15:37:31','15:39:30',
    '15:39:36','15:40:53','15:43:39','15:45:34','15:45:53','15:47:09','15:47:37','15:48:02','15:49:56','15:50:33',
    '15:51:01','15:54:48','15:55:21','15:55:39','15:56:33'
]

counter1_1007_wait_time = [
    '00:01:04','00:00:38','00:00:49','00:00:00','00:00:31','00:00:00','00:00:00','00:00:09','00:00:35','00:01:00',
    '00:01:33','00:01:37','00:01:24','00:01:02','00:01:36','00:02:04','00:00:28','00:00:00','00:00:45','00:00:31',
    '00:00:01','00:00:43','00:00:02','00:00:00','00:00:00','00:00:09','00:00:56','00:01:42','00:01:47','00:00:16',
    '00:01:06','00:02:08','00:02:15','00:02:50','00:02:00','00:01:37','00:01:49',
    '00:00:52','00:00:00','00:00:23','00:00:43','00:00:36','00:00:52','00:00:09','00:00:50','00:01:26','00:01:54',
    '00:01:59','00:00:38','00:00:00','00:00:02','00:00:00','00:00:10','00:00:04','00:01:09','00:00:57','00:00:48',
    '00:01:11','00:02:50','00:03:04','00:02:08','00:01:17','00:00:38','00:00:57','00:00:29','00:01:21','00:03:08',
    '00:02:11','00:02:10','00:02:47','00:00:20','00:01:06','00:01:38','00:01:37'
]

counter1_1007_service_time = [
    '00:02:03','00:00:50','00:00:43','00:01:08','00:01:21','00:00:32','00:01:48','00:01:03','00:01:11','00:01:56',
    '00:01:39','00:00:24','00:00:53','00:03:13','00:00:24','00:00:34','00:01:06','00:00:45','00:00:30','00:00:31',
    '00:00:45','00:00:17','00:00:30','00:00:26','00:00:40','00:01:32','00:01:11','00:01:22','00:00:49','00:00:59',
    '00:01:12','00:01:11','00:01:18','00:00:23','00:00:24','00:00:34','00:01:08',
    '00:01:14','00:01:15','00:00:24','00:01:01','00:01:49','00:00:31','00:00:35','00:01:18','00:01:01','00:00:41',
    '00:00:37','00:00:59','00:01:28','00:00:58','00:00:44','00:01:55','00:01:38','00:00:23','00:00:23','00:01:23',
    '00:03:36','00:00:19','00:00:20','00:01:54','00:01:16','00:00:36','00:00:47','00:01:17','00:02:12','00:00:56',
    '00:00:35','00:01:05','00:01:18','00:00:43','00:00:48','00:00:52','00:01:08'
]

counter1_1007_service_end_time = [
    '14:37:29','14:38:19','14:39:02','14:40:11','14:41:32','14:43:36','14:45:36','14:46:39','14:47:50','14:49:47',
    '14:51:26','14:51:50','14:52:43','14:55:56','14:56:20','14:57:04','14:58:10','14:58:56','14:59:27','14:59:58',
    '15:00:44','15:01:01','15:01:33','15:02:11','15:02:57','15:04:29','15:05:40','15:07:03','15:07:55','15:08:55',
    '15:10:08','15:11:19','15:12:37','15:13:00','15:13:24','15:13:58','15:15:07',
    '15:17:15','15:18:31','15:18:55','15:19:57','15:21:47','15:22:18','15:22:53','15:25:12','15:26:14','15:26:56',
    '15:27:34','15:28:33','15:30:38','15:31:38','15:32:54','15:34:51','15:36:30','15:36:54','15:37:18','15:38:41',
    '15:42:18','15:42:39','15:43:00','15:44:55','15:46:12','15:46:48','15:47:37','15:48:55','15:51:10','15:52:06',
    '15:52:42','15:53:48','15:55:06','15:55:51','15:57:15','15:58:09','15:59:18'
]

# 10/07(화) - 2번 계산대 (총 72건)

counter2_1007_arrival_time = [
    '14:34:15','14:34:33','14:35:08','14:36:43','14:37:10','14:39:47','14:42:20','14:44:04','14:46:17','14:46:35',
    '14:47:56','14:48:54','14:49:56','14:50:17','14:50:31','14:51:38','14:52:56','14:55:23','14:57:54','15:00:22',
    '15:01:36','15:02:53','15:03:44','15:04:58','15:06:23','15:06:55','15:07:15','15:08:24','15:09:12','15:10:02',
    '15:15:51','15:16:12','15:16:14','15:17:55','15:18:28','15:19:30','15:21:27','15:22:10','15:23:07','15:24:31',
    '15:25:59','15:26:28','15:29:10','15:29:37','15:30:18','15:33:04','15:34:07','15:34:34','15:34:56','15:35:58',
    '15:36:22','15:36:37','15:37:32','15:37:38','15:38:52','15:39:17','15:39:33','15:40:18','15:41:39','15:43:19',
    '15:45:46','15:47:08','15:47:29','15:48:11','15:50:06','15:53:05','15:54:04','15:54:21','15:55:24','15:56:54',
    '15:57:23','15:58:08'
]

counter2_1007_wait_time = [
    '00:00:00','00:01:22','00:01:23','00:01:10','00:02:49','00:01:29','00:01:12','00:00:48','00:00:00','00:00:43',
    '00:00:41','00:00:23','00:00:20','00:00:58','00:01:37','00:01:24','00:01:16','00:00:00','00:00:00','00:00:00',
    '00:00:13','00:00:58','00:01:17','00:02:19','00:01:48','00:02:17','00:03:26','00:03:39','00:03:59','00:03:55',
    '00:00:23','00:01:26','00:02:15','00:01:36','00:02:31','00:02:27','00:01:21','00:01:44','00:01:58','00:01:17',
    '00:01:13','00:01:47','00:00:01','00:01:07','00:02:09','00:00:10','00:00:42','00:01:24','00:01:56','00:01:51',
    '00:02:42','00:03:11','00:03:00','00:03:47','00:03:50','00:04:15','00:04:43','00:04:30','00:04:06','00:03:52',
    '00:03:00','00:02:58','00:03:54','00:03:58','00:02:57','00:00:46','00:00:55','00:01:53','00:02:10','00:01:19',
    '00:01:33','00:01:56'
]

counter2_1007_service_time = [
    '00:01:40','00:00:36','00:01:22','00:02:06','00:01:17','00:02:16','00:01:20','00:00:40','00:01:01','00:01:19',
    '00:00:40','00:00:59','00:00:59','00:00:53','00:00:54','00:01:10','00:01:10','00:01:56','00:01:17','00:01:27',
    '00:02:02','00:01:10','00:02:16','00:00:54','00:01:01','00:01:29','00:01:22','00:01:08','00:00:46','00:02:17',
    '00:01:24','00:00:51','00:01:02','00:01:28','00:00:58','00:00:51','00:01:06','00:01:11','00:00:43','00:01:24',
    '00:01:03','00:00:56','00:01:33','00:01:43','00:00:47','00:01:35','00:01:09','00:00:54','00:00:57','00:01:15',
    '00:00:44','00:00:44','00:01:26','00:01:17','00:00:50','00:00:44','00:00:32','00:00:57','00:01:26','00:01:35',
    '00:01:20','00:01:17','00:00:46','00:00:54','00:00:48','00:01:08','00:01:15','00:01:20','00:00:39','00:00:43',
    '00:01:08','00:01:07'
]

counter2_1007_service_end_time = [
    '14:35:55','14:36:31','14:37:53','14:39:59','14:41:16','14:43:32','14:44:52','14:45:32','14:47:18','14:48:37',
    '14:49:17','14:50:16','14:51:15','14:52:08','14:53:02','14:54:12','14:55:22','14:57:19','14:59:11','15:01:49',
    '15:03:51','15:05:01','15:07:17','15:08:11','15:09:12','15:10:41','15:12:03','15:13:11','15:13:57','15:16:14',
    '15:17:38','15:18:29','15:19:31','15:20:59','15:21:57','15:22:48','15:23:54','15:25:05','15:25:48','15:27:12',
    '15:28:15','15:29:11','15:30:44','15:32:27','15:33:14','15:34:49','15:35:58','15:36:52','15:37:49','15:39:04',
    '15:39:48','15:40:32','15:41:25','15:42:42','15:43:32','15:44:16','15:44:48','15:45:45','15:47:11','15:48:46',
    '15:50:06','15:51:23','15:52:09','15:53:03','15:53:51','15:54:59','15:56:14','15:57:34','15:58:13','15:58:56',
    '16:00:04','16:01:11'
]


# 10월 8일 (수) 1번 계산대(74개)

counter1_1008_arrival_time = [
    '14:28:30','14:30:05','14:31:30','14:32:25','14:33:27','14:33:59','14:35:25','14:36:21','14:37:20','14:38:16',
    '14:39:40','14:40:29','14:42:58','14:43:03','14:44:54','14:45:14','14:45:30','14:46:12','14:47:30','14:48:00',
    '14:48:48','14:51:43','14:52:13','14:53:06','14:53:52','14:55:29','14:56:02','14:55:39','14:57:54','14:59:05',
    '14:59:47','15:00:31','15:01:21','15:02:08','15:02:48','15:08:03','15:11:01','15:12:02','15:12:48','15:14:04',
    '15:14:31','15:16:11','15:16:58','15:20:52','15:21:34','15:21:35','15:22:30','15:24:39','15:25:08','15:26:10',
    '15:26:11','15:27:27','15:29:10','15:29:30','15:29:44','15:29:57','15:31:20','15:32:05','15:34:20','15:36:01',
    '15:36:32','15:44:03','15:47:11','15:47:38','15:48:07','15:48:45','15:50:26','15:53:23','15:53:46','15:55:07',
    '15:56:00','15:56:29','15:58:59','15:59:01'
]



counter1_1008_wait_time = [
    '00:00:00','00:00:20','00:00:00','00:00:17','00:00:25','00:01:03','00:00:25','00:00:00','00:00:00','00:00:14',
    '00:00:24','00:00:23','00:00:00','00:00:22','00:00:00','00:00:26','00:00:32','00:00:45','00:00:32','00:02:06',
    '00:02:15','00:00:27','00:00:59','00:00:24','00:00:52','00:00:41','00:01:21','00:03:03','00:01:42','00:01:14',
    '00:01:05','00:01:02','00:01:36','00:02:02','00:02:14','00:00:00','00:00:00','00:00:00','00:00:04','00:00:00',
    '00:00:53','00:00:00','00:00:06','00:00:00','00:00:33','00:01:19','00:01:34','00:00:15','00:00:31','00:00:00',
    '00:01:24','00:00:35','00:00:00','00:00:32','00:00:36','00:01:15','00:01:11','00:02:07','00:01:45','00:07:01',
    '00:07:41','00:01:52','00:00:00','00:00:17','00:00:47','00:01:13','00:00:52','00:00:00','00:00:14','00:00:00',
    '00:00:07','00:00:56','00:00:00','00:00:14'
]


counter1_1008_service_time = [
    '00:01:55','00:00:55','00:01:12','00:01:10','00:01:10','00:00:48','00:00:20','00:00:24','00:01:10','00:01:34',
    '00:00:48','00:01:13','00:00:27','00:00:25','00:00:46','00:00:22','00:00:55','00:01:05','00:02:04','00:00:57',
    '00:01:07','00:01:02','00:00:18','00:01:14','00:01:26','00:01:13','00:01:19','00:00:54','00:00:43','00:00:33',
    '00:00:41','00:01:24','00:01:13','00:00:52','00:01:02','00:02:33','00:00:25','00:00:50','00:01:08','00:01:20',
    '00:00:37','00:00:53','00:02:32','00:01:15','00:00:47','00:01:10','00:00:50','00:00:45','00:00:23','00:01:25',
    '00:00:27','00:00:31','00:00:51','00:00:18','00:00:52','00:01:19','00:01:41','00:01:53','00:06:57','00:01:12',
    '00:01:42','00:00:44','00:00:44','00:00:59','00:01:04','00:01:20','00:00:30','00:00:37','00:00:46','00:01:00',
    '00:01:18','00:00:58','00:00:16','00:01:43'
]


counter1_1008_service_end_time = [
    '14:30:25','14:31:20','14:32:42','14:33:52','14:35:02','14:35:50','14:36:10','14:36:45','14:38:30','14:40:04',
    '14:40:52','14:42:05','14:43:25','14:43:50','14:45:40','14:46:02','14:46:57','14:48:02','14:50:06','14:51:03',
    '14:52:10','14:53:12','14:53:30','14:54:44','14:56:10','14:57:23','14:58:42','14:59:36','15:00:19','15:00:52',
    '15:01:33','15:02:57','15:04:10','15:05:02','15:06:04','15:10:36','15:11:26','15:12:52','15:14:00','15:15:24',
    '15:16:01','15:17:04','15:19:36','15:22:07','15:22:54','15:24:04','15:24:54','15:25:39','15:26:02','15:27:35',
    '15:28:02','15:28:33','15:30:01','15:30:20','15:31:12','15:32:31','15:34:12','15:36:05','15:43:02','15:44:14',
    '15:45:55','15:46:39','15:47:55','15:48:54','15:49:58','15:51:18','15:51:48','15:54:00','15:54:46','15:56:07',
    '15:57:25','15:58:23','15:59:15','16:00:58'
]


# 10월 8일 (수) 2번 계산대

counter2_1008_arrival_time = [
    '14:29:35','14:30:41','14:31:37','14:32:24','14:34:11','14:35:23','14:34:36','14:36:57','14:38:33','14:38:50',
    '14:40:04','14:40:19','14:43:36','14:43:57','14:45:29','14:45:52','14:47:19','14:48:50','14:49:40','14:50:22',
    '14:51:51','14:51:52','14:52:27','14:53:12','14:55:56','14:56:14','14:56:29','14:56:43','14:58:26','14:58:37',
    '14:59:40','15:02:09','15:03:02','15:03:34','15:04:29','15:09:54','15:11:10','15:12:20','15:15:16','15:17:19',
    '15:18:12','15:18:39','15:20:55','15:21:21','15:23:29','15:25:04','15:26:43','15:27:40','15:29:21','15:29:38',
    '15:29:55','15:31:44','15:32:37','15:33:01','15:34:03','15:35:17','15:36:16','15:37:24','15:38:20','15:39:06',
    '15:39:44','15:40:21','15:40:38','15:42:52','15:43:22','15:43:29','15:47:45','15:49:12','15:49:41','15:51:26',
    '15:52:57','15:54:57','15:57:06','15:57:53','15:59:16'
]



counter2_1008_wait_time = [
    '00:00:00','00:00:28','00:00:09','00:00:23','00:00:00','00:00:00','00:02:07','00:00:34','00:00:08','00:00:49',
    '00:00:43','00:01:16','00:00:00','00:00:51','00:00:14','00:00:26','00:00:00','00:00:00','00:00:15','00:00:33',
    '00:00:18','00:01:11','00:01:35','00:01:37','00:00:00','00:00:57','00:01:42','00:02:55','00:02:08','00:03:02',
    '00:02:37','00:01:02','00:00:50','00:01:03','00:01:04','00:00:00','00:00:09','00:00:11','00:00:00','00:00:00',
    '00:00:19','00:00:40','00:00:00','00:00:30','00:00:00','00:00:00','00:00:00','00:00:19','00:00:00','00:01:19',
    '00:02:08','00:01:04','00:01:14','00:01:14','00:02:05','00:01:58','00:02:23','00:02:21','00:02:11','00:02:19',
    '00:02:49','00:03:07','00:03:31','00:03:43','00:02:24','00:02:36','00:02:44','00:00:00','00:00:00','00:00:24',
    '00:00:19','00:00:11','00:00:00','00:00:00','00:00:19','00:00:38'
]

counter2_1008_service_time = [
    '00:01:34','00:00:37','00:01:01','00:01:17','00:00:42','00:01:20','00:00:48','00:01:10','00:00:58','00:01:08',
    '00:00:48','00:01:34','00:01:12','00:00:55','00:00:35','00:00:51','00:01:25','00:01:05','00:01:00','00:01:14',
    '00:00:54','00:00:59','00:00:47','00:00:32','00:01:15','00:01:00','00:01:27','00:00:56','00:01:05','00:00:38',
    '00:00:54','00:00:41','00:00:45','00:00:56','00:01:38','00:01:25','00:01:12','00:02:24','00:00:57','00:01:12',
    '00:00:48','00:01:08','00:00:56','00:01:30','00:01:28','00:01:18','00:01:16','00:00:58','00:01:36','00:01:06',
    '00:00:45','00:01:03','00:01:15','00:00:55','00:01:39','00:00:57','00:00:58','00:01:04','00:01:16','00:00:56',
    '00:01:01','00:00:29','00:00:55','00:00:42','00:00:15','00:01:09','00:01:06','00:01:06','00:00:53','00:01:40',
    '00:01:23','00:00:48','00:01:30','00:01:06','00:01:42','00:00:57'
]

counter2_1008_service_end_time = [
    '14:31:09','14:31:46','14:32:47','14:34:04','14:34:53','14:36:43','14:37:31','14:38:41','14:39:39','14:40:47',
    '14:41:35','14:43:09','14:44:48','14:45:43','14:46:18','14:47:09','14:48:44','14:49:55','14:50:55','14:52:09',
    '14:53:03','14:54:02','14:54:49','14:55:21','14:57:11','14:58:11','14:59:38','15:00:34','15:01:39','15:02:17',
    '15:03:11','15:03:52','15:04:37','15:05:33','15:07:11','15:11:19','15:12:31','15:14:55','15:16:13','15:18:31',
    '15:19:19','15:20:27','15:21:51','15:23:21','15:24:57','15:26:22','15:27:59','15:28:57','15:30:57','15:32:03',
    '15:32:48','15:33:51','15:35:06','15:36:01','15:37:40','15:38:37','15:39:35','15:40:39','15:41:55','15:42:51',
    '15:43:52','15:44:21','15:45:16','15:45:58','15:46:13','15:47:22','15:48:51','15:50:05','15:51:45','15:53:08',
    '15:53:56','15:56:27','15:58:12','15:59:54','16:00:51'
]

# 10월 15일(수) 1번 계산대(85개)


counter1_1015_arrival_time = [
    '15:00:08','15:01:36','15:01:38','15:02:51','15:04:43','15:05:52','15:06:13','15:06:14','15:07:08','15:09:19',
    '15:10:36','15:11:20','15:11:50','15:12:11','15:13:11','15:15:08','15:15:20','15:16:02','15:16:50','15:17:59',
    '15:20:13','15:20:20','15:20:55','15:21:03','15:22:45','15:23:55','15:26:11','15:26:27','15:26:43','15:27:04',
    '15:27:43','15:28:15','15:29:57','15:31:31','15:31:32','15:31:35','15:32:02','15:33:05','15:34:09','15:34:14',
    '15:34:46','15:35:24','15:35:37','15:37:09','15:38:28','15:38:31','15:39:05','15:42:20','15:42:42','15:43:51',
    '15:44:32','15:45:09','15:46:00','15:46:02','15:46:35','15:48:43','15:50:02','15:52:34','15:52:41','15:53:02',
    '15:53:47','15:54:13','15:57:40','15:58:30','16:00:20','16:03:40','16:05:21','16:05:57','16:07:10','16:07:47',
    '16:11:37','16:14:24','16:15:13','16:15:58','16:16:30','16:17:28','16:18:43','16:20:28','16:20:56','16:22:03',
    '16:22:14','16:22:27','16:25:20','16:28:32','16:28:58'
]


counter1_1015_wait_time = [
    '00:00:00','00:00:00','00:01:04','00:02:11','00:00:46','00:01:48','00:01:44','00:02:18','00:02:30','00:00:40',
    '00:00:00','00:00:17','00:00:30','00:00:35','00:01:28','00:00:29','00:01:33','00:01:45','00:02:28','00:01:55',
    '00:00:33','00:01:53','00:02:29','00:03:04','00:02:10','00:01:29','00:00:00','00:00:57','00:01:21','00:01:59',
    '00:01:41','00:01:27','00:00:42','00:00:33','00:02:07','00:03:22','00:03:51','00:03:33','00:03:21','00:03:52',
    '00:04:02','00:04:51','00:06:03','00:04:56','00:04:45','00:05:11','00:05:11','00:02:52','00:03:21','00:03:03',
    '00:03:30','00:03:47','00:03:43','00:04:34','00:04:59','00:04:51','00:05:26','00:05:00','00:05:57','00:06:25',
    '00:06:55','00:07:01','00:04:10','00:04:49','00:05:03','00:02:14','00:01:37','00:01:56','00:01:44','00:03:34',
    '00:00:12','00:00:00','00:00:00','00:01:11','00:01:41','00:02:08','00:01:44','00:00:58','00:01:03','00:00:47',
    '00:01:00','00:01:24','00:00:00','00:00:00','00:00:27'
]

counter1_1015_service_time = [
    '00:00:18','00:01:06','00:02:20','00:00:27','00:02:11','00:00:17','00:00:35','00:01:06','00:00:21','00:00:31',
    '00:01:01','00:00:43','00:00:26','00:01:53','00:00:58','00:01:16','00:00:54','00:01:31','00:00:36','00:00:52',
    '00:01:27','00:01:11','00:00:43','00:00:48','00:00:29','00:00:23','00:01:13','00:00:40','00:00:59','00:00:21',
    '00:00:18','00:00:57','00:01:25','00:01:35','00:01:18','00:00:56','00:00:45','00:00:52','00:00:36','00:00:42',
    '00:01:27','00:01:25','00:00:26','00:01:08','00:00:29','00:00:34','00:00:56','00:00:51','00:00:51','00:01:08',
    '00:00:54','00:00:47','00:00:53','00:00:58','00:02:00','00:01:54','00:02:06','00:01:04','00:00:49','00:01:15',
    '00:00:32','00:00:36','00:01:29','00:02:04','00:00:31','00:01:04','00:00:55','00:01:01','00:02:27','00:00:28',
    '00:01:35','00:00:47','00:01:56','00:01:02','00:01:25','00:00:51','00:00:59','00:00:23','00:00:51','00:00:24',
    '00:00:37','00:00:24','00:01:08','00:00:53','00:00:51'
]

counter1_1015_service_end_time = [
    '15:00:26','15:02:42','15:05:02','15:05:29','15:07:40','15:07:57','15:08:32','15:09:38','15:09:59','15:10:30',
    '15:11:37','15:12:20','15:12:46','15:14:39','15:15:37','15:16:53','15:17:47','15:19:18','15:19:54','15:20:46',
    '15:22:13','15:23:24','15:24:07','15:24:55','15:25:24','15:25:47','15:27:24','15:28:04','15:29:03','15:29:24',
    '15:29:42','15:30:39','15:32:04','15:33:39','15:34:57','15:35:53','15:36:38','15:37:30','15:38:06','15:38:48',
    '15:40:15','15:41:40','15:42:06','15:43:13','15:43:42','15:44:16','15:45:12','15:46:03','15:46:54','15:48:02',
    '15:48:56','15:49:43','15:50:36','15:51:34','15:53:34','15:55:28','15:57:34','15:58:38','15:59:27','16:00:42',
    '16:01:14','16:01:50','16:03:19','16:05:23','16:05:54','16:06:58','16:07:53','16:08:54','16:11:21','16:11:49',
    '16:13:24','16:15:11','16:17:09','16:18:11','16:19:36','16:20:27','16:21:26','16:21:49','16:22:50','16:23:14',
    '16:23:51','16:24:15','16:26:28','16:29:25','16:30:16'
]

# 10월 15일 (수) 2번 계산대 (76개)

counter2_1015_arrival_time= [
    '15:00:55','15:02:30','15:03:36','15:04:07','15:04:40','15:05:16','15:11:33','15:11:49','15:13:49','15:15:18',
    '15:16:08','15:17:28','15:18:11','15:18:58','15:19:17','15:20:33','15:20:58','15:21:21','15:23:13','15:24:37',
    '15:25:10','15:27:10','15:27:42','15:28:36','15:30:07','15:31:04','15:31:15','15:31:35','15:32:26','15:33:33',
    '15:33:58','15:34:07','15:34:15','15:35:02','15:36:16','15:37:35','15:38:31','15:39:04','15:39:38','15:40:00',
    '15:41:18','15:43:07','15:44:11','15:45:19','15:45:44','15:46:02','15:46:19','15:48:02','15:49:37','15:50:40',
    '15:52:36','15:53:00','15:55:30','15:57:09','15:58:46','16:00:01','16:02:30','16:02:40','16:05:37','16:07:30',
    '16:07:41','16:10:36','16:14:44','16:14:56','16:15:40','16:16:25','16:17:01','16:18:33','16:19:41','16:21:06',
    '16:21:46','16:23:01','16:24:54','16:25:07','16:28:16','16:28:41'
]

counter2_1015_wait_time = [
    '00:01:47','00:00:45','00:00:11','00:00:26','00:01:10','00:02:26','00:02:15','00:02:35','00:01:22','00:01:04',
    '00:01:00','00:00:31','00:00:38','00:00:56','00:03:06','00:02:57','00:03:09','00:03:05','00:01:30','00:01:54',
    '00:02:40','00:01:47','00:01:44','00:01:24','00:01:19','00:01:01','00:01:49','00:02:12','00:02:22','00:02:21',
    '00:03:07','00:03:48','00:03:59','00:04:21','00:04:10','00:03:22','00:03:02','00:03:33','00:04:04','00:04:33',
    '00:03:36','00:02:13','00:04:06','00:03:20','00:03:41','00:04:15','00:07:33','00:06:50','00:05:47','00:06:10',
    '00:05:01','00:05:10','00:03:08','00:02:30','00:01:57','00:01:48','00:00:42','00:01:15','00:01:45','00:01:11',
    '00:02:37','00:00:30','00:00:34','00:00:37','00:01:09','00:01:15','00:00:51','00:01:02','00:01:03','00:01:15',
    '00:01:12','00:01:25','00:01:07','00:01:36','00:00:00','00:00:18'
]

counter2_1015_service_time = [
    '00:00:18','00:00:30','00:00:28','00:01:07','00:01:07','00:06:05','00:00:34','00:00:38','00:01:09','00:00:39',
    '00:00:49','00:00:36','00:01:02','00:02:24','00:00:57','00:00:30','00:00:12','00:00:14','00:01:44','00:01:02',
    '00:00:53','00:00:19','00:00:29','00:00:57','00:00:33','00:00:45','00:00:16','00:00:58','00:00:59','00:01:07',
    '00:00:41','00:00:16','00:01:01','00:00:48','00:00:24','00:00:34','00:00:56','00:00:58','00:00:43','00:00:19',
    '00:00:14','00:02:30','00:00:14','00:00:34','00:00:50','00:03:32','00:00:57','00:00:27','00:00:06','00:00:43',
    '00:00:21','00:00:20','00:00:49','00:00:57','00:01:04','00:01:01','00:00:32','00:00:53','00:01:17','00:01:26',
    '00:00:41','00:00:41','00:00:13','00:01:07','00:00:49','00:00:04','00:01:33','00:01:05','00:01:32','00:00:19',
    '00:01:14','00:01:10','00:00:38','00:00:41','00:00:39','00:01:20'
]

counter2_1015_service_end_time = [
    '15:03:00','15:03:45','15:04:15','15:05:40','15:06:57','15:13:47','15:14:22','15:15:02','15:16:20','15:17:01',
    '15:17:57','15:18:35','15:19:51','15:22:18','15:23:20','15:24:00','15:24:19','15:24:40','15:26:27','15:27:33',
    '15:28:43','15:29:16','15:29:55','15:30:57','15:31:59','15:32:50','15:33:20','15:34:45','15:35:47','15:37:01',
    '15:37:46','15:38:11','15:39:15','15:40:11','15:40:50','15:41:31','15:42:29','15:43:35','15:44:25','15:44:52',
    '15:45:08','15:47:50','15:48:31','15:49:13','15:50:15','15:53:49','15:54:49','15:55:19','15:55:30','15:57:33',
    '15:57:58','15:58:30','15:59:27','16:00:36','16:01:47','16:02:50','16:03:44','16:04:48','16:08:39','16:10:07',
    '16:10:59','16:11:47','16:15:31','16:16:40','16:17:38','16:17:44','16:19:25','16:20:40','16:22:16','16:22:40',
    '16:24:12','16:25:36','16:26:39','16:27:24','16:28:55','16:30:19'
]

from scipy.stats import kruskal

# HH:MM:SS → 초 단위로 변환
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 도착시간 리스트 → 정렬 → interarrival time (초) 리스트 반환
def get_interarrival(arrival_time_list):
    # 1) 초단위로 변환
    secs = [hhmmss_to_sec(t) for t in arrival_time_list]
    # 2) 혹시 순서가 뒤섞였을 수 있으니 정렬
    secs_sorted = sorted(secs)
    # 3) 인접 차이 = interarrival
    interarrival = [
        secs_sorted[i] - secs_sorted[i-1]
        for i in range(1, len(secs_sorted))
    ]
    return interarrival
# ------------------------
# Counter 1: 날짜별 interarrival
# ------------------------
c1_1006_ia = get_interarrival(counter1_1006_arrival_time)
c1_1007_ia = get_interarrival(counter1_1007_arrival_time)
c1_1008_ia = get_interarrival(counter1_1008_arrival_time)
c1_1015_ia = get_interarrival(counter1_1015_arrival_time)

print("[Counter 1] interarrival 개수")
print("10/06:", len(c1_1006_ia))
print("10/07:", len(c1_1007_ia))
print("10/08:", len(c1_1008_ia))
print("10/15:", len(c1_1015_ia))

# ------------------------
# Counter 2: 날짜별 interarrival
# ------------------------
c2_1006_ia = get_interarrival(counter2_1006_arrival_time)
c2_1007_ia = get_interarrival(counter2_1007_arrival_time)
c2_1008_ia = get_interarrival(counter2_1008_arrival_time)
c2_1015_ia = get_interarrival(counter2_1015_arrival_time)

print("\n[Counter 2] interarrival 개수")
print("10/06:", len(c2_1006_ia))
print("10/07:", len(c2_1007_ia))
print("10/08:", len(c2_1008_ia))
print("10/15:", len(c2_1015_ia))

alpha = 0.05  # 유의수준


# ==============================
# ① Counter 1 : 4일 동질성 검정
# ==============================
stat_c1, p_c1 = kruskal(c1_1006_ia, c1_1007_ia, c1_1008_ia, c1_1015_ia)

print("\n==============================")
print("[Counter 1] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat_c1:.4f}")
print(f"p-value  = {p_c1:.6f}")

if p_c1 > alpha:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각 불가")
    print("   ⇒ Counter 1의 4일 interarrival 분포는 동질적이라고 볼 수 있음.")
    print("   ⇒ Counter 1은 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각")
    print("   ⇒ Counter 1의 4일 중 적어도 하루의 interarrival 분포가 다름.")
    print("   ⇒ Counter 1은 날짜를 나누거나 원인 분석 필요.")


# ==============================
# ② Counter 2 : 4일 동질성 검정
# ==============================
stat_c2, p_c2 = kruskal(c2_1006_ia, c2_1007_ia, c2_1008_ia, c2_1015_ia)

print("\n==============================")
print("[Counter 2] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat_c2:.4f}")
print(f"p-value  = {p_c2:.6f}")

if p_c2 > alpha:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각 불가")
    print("   ⇒ Counter 2의 4일 interarrival 분포는 동질적이라고 볼 수 있음.")
    print("   ⇒ Counter 2도 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각")
    print("   ⇒ Counter 2의 4일 중 적어도 하루의 interarrival 분포가 다름.")
    print("   ⇒ Counter 2 역시 날짜를 나누거나 원인 분석 필요.")

import numpy as np

# HH:MM:SS → 초로 변환
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 도착시간 리스트 두 개(1번, 2번)를 합쳐서 → 정렬 → interarrival(초) 리스트
def get_daily_interarrival(counter1_times, counter2_times):
    # 1,2번 합쳐서 초 단위로 변환
    secs = [hhmmss_to_sec(t) for t in counter1_times] + \
           [hhmmss_to_sec(t) for t in counter2_times]
    # 시간순 정렬 (리스트에 약간 순서 꼬인 게 있어서 꼭 정렬해줘야 함!)
    secs_sorted = sorted(secs)
    # 인접 고객 도착 간격 = interarrival time
    inter = [secs_sorted[i] - secs_sorted[i-1] for i in range(1, len(secs_sorted))]
    return inter

# --- 1) 4일치 interarrival(일자별) 계산 ---
ia_1006 = get_daily_interarrival(counter1_1006_arrival_time, counter2_1006_arrival_time)
ia_1007 = get_daily_interarrival(counter1_1007_arrival_time, counter2_1007_arrival_time)
ia_1008 = get_daily_interarrival(counter1_1008_arrival_time, counter2_1008_arrival_time)
ia_1015 = get_daily_interarrival(counter1_1015_arrival_time, counter2_1015_arrival_time)

print("10/06 interarrival 개수:", len(ia_1006))
print("10/07 interarrival 개수:", len(ia_1007))
print("10/08 interarrival 개수:", len(ia_1008))
print("10/15 interarrival 개수:", len(ia_1015))


# --- 2) interarrival 값 자체 출력 (초 단위 & 분 단위) ---
def print_interarrivals(label, ia_list):
    print(f"\n=== {label} (n={len(ia_list)}) ===")
    print("초 단위 리스트:")
    print(ia_list)   # 이게 네가 말한 '값 자체'

    mins = [x/60 for x in ia_list]
    print("\n분 단위 리스트(=초/60):")
    print(mins)

    print("\n요약 통계:")
    print("  평균(초):", np.mean(ia_list))
    print("  표준편차(초):", np.std(ia_list))
    print("  최소/최대(초):", np.min(ia_list), "/", np.max(ia_list))

print_interarrivals("10/06 전체(1+2번)", ia_1006)
print_interarrivals("10/07 전체(1+2번)", ia_1007)
print_interarrivals("10/08 전체(1+2번)", ia_1008)
print_interarrivals("10/15 전체(1+2번)", ia_1015)

def get_interarrival_single_counter(arrival_times):
    secs = sorted(hhmmss_to_sec(t) for t in arrival_times)
    return [secs[i] - secs[i-1] for i in range(1, len(secs))]

c1_1006_ia = get_interarrival_single_counter(counter1_1006_arrival_time)
# 이런 식으로 나머지도 계산 가능

# 위에서 구한 ia_1006, ia_1007, ia_1008, ia_1015 재사용

all_ia = ia_1006 + ia_1007 + ia_1008 + ia_1015  # 4일 전체 interarrival (초 단위)
all_ia = np.array(all_ia)

print("4일 전체 interarrival 개수:", len(all_ia))
print("평균(초):", np.mean(all_ia))
print("표준편차(초):", np.std(all_ia))

import matplotlib.pyplot as plt
from scipy import stats

# 0보다 큰 값만 사용 (혹시 0이 있으면 분포 적합에 문제될 수 있어서 필터링)
all_ia_pos = all_ia[all_ia > 0]

# --- 1) 히스토그램 그리기 ---
plt.hist(all_ia_pos, bins=20, density=True, alpha=0.5)
plt.xlabel("Interarrival time (sec)")
plt.ylabel("Density")
plt.title("Histogram of Interarrival Times (4 days, both counters)")
plt.show()

# --- 2) 여러 분포 피팅해보기 ---
candidates = {
    "expon": stats.expon,          # 지수분포
    "gamma": stats.gamma,          # 감마분포
    "lognorm": stats.lognorm,      # 로그정규
    "weibull_min": stats.weibull_min  # Weibull 최소형
}

fit_results = {}

for name, dist in candidates.items():
    # 파라미터 적합
    params = dist.fit(all_ia_pos)
    fit_results[name] = params

    # K-S 검정으로 적합도 대략 보기
    ks_stat, p_val = stats.kstest(all_ia_pos, name, args=params)
    print(f"\n[{name}]")
    print("  params:", params)
    print("  KS stat:", ks_stat)
    print("  p-value:", p_val)

# --- 3) 히스토그램 위에 분포 곡선 그려보기 (시각적으로 비교) ---
xmin, xmax = all_ia_pos.min(), all_ia_pos.max()
x = np.linspace(xmin, xmax, 200)

plt.hist(all_ia_pos, bins=20, density=True, alpha=0.5, label="Data")

for name, dist in candidates.items():
    params = fit_results[name]
    pdf = dist.pdf(x, *params)
    plt.plot(x, pdf, label=name)

plt.xlabel("Interarrival time (sec)")
plt.ylabel("Density")
plt.title("Fitted Distributions to Interarrival Times")
plt.legend()
plt.show()

from scipy.stats import kruskal
import numpy as np

# HH:MM:SS → 초 변환 함수
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 하루 서비스타임 = 1번 + 2번 합치기
def get_daily_service_times(counter1_service, counter2_service):
    secs_c1 = [hhmmss_to_sec(t) for t in counter1_service]
    secs_c2 = [hhmmss_to_sec(t) for t in counter2_service]
    return secs_c1 + secs_c2


# --- 4일 서비스타임 데이터 준비 ---
st_1006 = get_daily_service_times(counter1_1006_service_time, counter2_1006_service_time)
st_1007 = get_daily_service_times(counter1_1007_service_time, counter2_1007_service_time)
st_1008 = get_daily_service_times(counter1_1008_service_time, counter2_1008_service_time)
st_1015 = get_daily_service_times(counter1_1015_service_time, counter2_1015_service_time)

print("10/06 서비스타임 개수:", len(st_1006))
print("10/07 서비스타임 개수:", len(st_1007))
print("10/08 서비스타임 개수:", len(st_1008))
print("10/15 서비스타임 개수:", len(st_1015))


# --- 동질성 검정(Kruskal–Wallis) ---
alpha = 0.05

stat, p_val = kruskal(st_1006, st_1007, st_1008, st_1015)

print("\n==============================")
print("[서비스타임] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat:.4f}")
print(f"p-value  = {p_val:.6f}")

if p_val > alpha:
    print(f"\n→ p-value > {alpha} 이므로 귀무가설 기각 불가")
    print("   ⇒ 4일 서비스타임 분포는 '동질적'이다.")
    print("   ⇒ 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"\n→ p-value ≤ {alpha} 이므로 귀무가설 기각")
    print("   ⇒ 4일 중 적어도 하루의 서비스타임 분포가 다르다.")
    print("   ⇒ 4일을 합쳐 쓰면 안 되고, 원인 분석 또는 분리 필요.")

import numpy as np

def summary(label, data):
    data = np.array(data)
    print(f"\n=== {label} ===")
    print("n :", len(data))
    print("평균(초):", np.mean(data))
    print("표준편차(초):", np.std(data))
    print("최소/최대(초):", np.min(data), "/", np.max(data))

summary("10/06 서비스타임", st_1006)
summary("10/07 서비스타임", st_1007)
summary("10/08 서비스타임", st_1008)
summary("10/15 서비스타임", st_1015)

from scipy.stats import kruskal

pairs = [
    ("10/06", st_1006, "10/07", st_1007),
    ("10/06", st_1006, "10/08", st_1008),
    ("10/06", st_1006, "10/15", st_1015),
    ("10/07", st_1007, "10/08", st_1008),
    ("10/07", st_1007, "10/15", st_1015),
    ("10/08", st_1008, "10/15", st_1015),
]

alpha = 0.05

print("\n=== 날짜 쌍별 Kruskal–Wallis 검정 (서비스타임) ===")
for a_label, a, b_label, b in pairs:
    stat, p = kruskal(a, b)
    print(f"{a_label} vs {b_label} → H={stat:.3f}, p={p:.6f}", end="  ")
    if p > alpha:
        print("→ 차이 없음 (동질)")
    else:
        print("→ 차이 있음 (이질)")

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# =====================================
# 1. HH:MM:SS → 초 변환 함수
# =====================================
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# =====================================
# 2. Interarrival 계산 함수
# =====================================
def interarrival_from_times(time_list):
    """
    time_list: ['14:32:36', '14:35:11', ...]
    return: 연속 도착 간격(초) np.array, 길이는 n-1
    """
    secs = np.array([hhmmss_to_sec(t) for t in time_list], dtype=float)
    ia = np.diff(secs)   # 앞뒤 시간 차이
    return ia

# =====================================
# 3. 기술통계 함수
# =====================================
def describe_data(data, label="Data"):
    data = np.array(data, dtype=float)
    data = data[~np.isnan(data)]

    n = len(data)
    if n == 0:
        print(f"\n[{label}] 데이터 없음")
        return

    mean = np.mean(data)
    std = np.std(data, ddof=1)          # 표본 표준편차
    min_v = np.min(data)
    q1 = np.percentile(data, 25)
    median = np.percentile(data, 50)
    q3 = np.percentile(data, 75)
    max_v = np.max(data)
    iqr = q3 - q1
    cv = std / mean if mean != 0 else np.nan
    skew = stats.skew(data)
    kurt = stats.kurtosis(data)         # excess kurtosis

    print(f"\n===== [{label}] 기술통계 =====")
    print(f"N            : {n}")
    print(f"Mean         : {mean:8.3f}")
    print(f"Std          : {std:8.3f}")
    print(f"Min          : {min_v:8.3f}")
    print(f"Q1 (25%)     : {q1:8.3f}")
    print(f"Median (50%) : {median:8.3f}")
    print(f"Q3 (75%)     : {q3:8.3f}")
    print(f"Max          : {max_v:8.3f}")
    print(f"IQR          : {iqr:8.3f}")
    print(f"CV           : {cv:8.3f}")
    print(f"Skewness     : {skew:8.3f}")
    print(f"Kurtosis     : {kurt:8.3f}")

# =====================================
# 4. 서비스시간을 초로 변환
# =====================================

# 10/06
c1_1006_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1006_service_time])
c2_1006_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1006_service_time])

# 10/07
c1_1007_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1007_service_time])
c2_1007_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1007_service_time])

# 10/08
c1_1008_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1008_service_time])
c2_1008_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1008_service_time])

# 10/15
c1_1015_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1015_service_time])
c2_1015_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1015_service_time])

# =====================================
# 5. Interarrival Time 계산 (도착시간 기반)
# =====================================

# 10/06
ia_c1_1006 = interarrival_from_times(counter1_1006_arrival_time)
ia_c2_1006 = interarrival_from_times(counter2_1006_arrival_time)

# 10/07
ia_c1_1007 = interarrival_from_times(counter1_1007_arrival_time)
ia_c2_1007 = interarrival_from_times(counter2_1007_arrival_time)

# 10/08
ia_c1_1008 = interarrival_from_times(counter1_1008_arrival_time)
ia_c2_1008 = interarrival_from_times(counter2_1008_arrival_time)

# 10/15
ia_c1_1015 = interarrival_from_times(counter1_1015_arrival_time)
ia_c2_1015 = interarrival_from_times(counter2_1015_arrival_time)

# =====================================
# 6. 분석용 그룹 묶기
# =====================================

# 6-1. 서비스시간: 일반일(10/06~10/08) vs 피크일(10/15)
service_normal = np.concatenate([
    c1_1006_service_secs, c2_1006_service_secs,
    c1_1007_service_secs, c2_1007_service_secs,
    c1_1008_service_secs, c2_1008_service_secs
])

service_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# 6-2. Interarrival: 전체 / 일반일 / 피크일
ia_normal = np.concatenate([
    ia_c1_1006, ia_c2_1006,
    ia_c1_1007, ia_c2_1007,
    ia_c1_1008, ia_c2_1008
])

ia_peak = np.concatenate([
    ia_c1_1015, ia_c2_1015
])

ia_all = np.concatenate([ia_normal, ia_peak])

# =====================================
# 7. 기술통계 출력
# =====================================

# --- 서비스타임 ---
describe_data(service_normal, "Service Time - Normal Days (10/06~10/08, C1+C2)")
describe_data(service_peak,   "Service Time - Peak Day (10/15, C1+C2)")

describe_data(c1_1006_service_secs, "Service Time - 10/06 Counter1")
describe_data(c2_1006_service_secs, "Service Time - 10/06 Counter2")

# --- Interarrival Time ---
describe_data(ia_all,    "Interarrival - All Days (C1+C2)")
describe_data(ia_normal, "Interarrival - Normal Days (10/06~10/08)")
describe_data(ia_peak,   "Interarrival - Peak Day (10/15)")

# =====================================
# 8. 시각화 함수 (히스토그램 + 박스플롯)
# =====================================

def plot_histogram(data_list, labels, title, xlabel):
    """
    data_list: [array1, array2, ...]
    labels   : ['normal', 'peak', ...]
    """
    plt.figure(figsize=(10, 5))
    for d, lab in zip(data_list, labels):
        plt.hist(d, bins=30, alpha=0.5, density=True, label=lab)
    plt.xlabel(xlabel)
    plt.ylabel("Density")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_boxplot(data_list, labels, title, ylabel):
    plt.figure(figsize=(8, 5))
    plt.boxplot(data_list, labels=labels, showfliers=True)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    plt.show()

# =====================================
# 9. 실제 그래프 출력
# =====================================

# 9-1. 서비스타임 히스토그램 & 박스플롯 (일반일 vs 피크일)
plot_histogram(
    [service_normal, service_peak],
    ["Normal Days", "Peak Day"],
    "Service Time Distribution (Normal vs Peak)",
    "Service Time (seconds)"
)

plot_boxplot(
    [service_normal, service_peak],
    ["Normal", "Peak"],
    "Service Time Boxplot (Normal vs Peak)",
    "Service Time (seconds)"
)

# 9-2. Interarrival 히스토그램 & 박스플롯 (일반일 vs 피크일)
plot_histogram(
    [ia_normal, ia_peak],
    ["Normal Days", "Peak Day"],
    "Interarrival Time Distribution (Normal vs Peak)",
    "Interarrival Time (seconds)"
)

plot_boxplot(
    [ia_normal, ia_peak],
    ["Normal", "Peak"],
    "Interarrival Time Boxplot (Normal vs Peak)",
    "Interarrival Time (seconds)"
)

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import gamma, lognorm, expon, weibull_min

# =====================================================
# 0. 공통 도우미 함수들
# =====================================================

def hhmmss_to_sec(t):
    """'HH:MM:SS' 문자열을 초 단위 숫자로 변환"""
    h, m, s = map(int, t.split(':'))
    return h*3600 + m*60 + s

def to_seconds_list(str_list):
    """시간 문자열 리스트를 초 단위 numpy 배열로 변환"""
    return np.array([hhmmss_to_sec(t) for t in str_list], dtype=float)

def tukey_bounds(data, k=1.5):
    """Tukey IQR 기준으로 이상치 경계 계산 (Q1-1.5IQR, Q3+1.5IQR)"""
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

def fit_and_compare_distributions(data, label=""):
    """
    여러 분포(expon, gamma, lognorm, weibull_min)를 피팅하고
    KS-test, loglik를 출력. p-value 기준 정렬하여 반환.
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]  # 0 이하 제거
    candidates = {
        "expon": expon,
        "gamma": gamma,
        "lognorm": lognorm,
        "weibull_min": weibull_min
    }

    results = []
    print(f"\n===== [{label}] 분포 적합 결과 =====")
    for name, dist in candidates.items():
        # 서비스/도착시간은 0 이상이므로 loc=0 고정
        params = dist.fit(data, floc=0)
        ks, p = stats.kstest(data, dist.cdf, args=params)
        loglik = np.sum(dist.logpdf(data, *params))
        results.append({
            "name": name,
            "dist": dist,
            "params": params,
            "ks": ks,
            "p": p,
            "loglik": loglik
        })
        print(f"{name:10s} | KS={ks:.4f}, p={p:.6f}, loglik={loglik:.1f}")
        print(f"  params = {params}\n")

    results.sort(key=lambda x: x["p"], reverse=True)

    print(f"\n>>> [{label}] p-value 순(적합도 순) <<<")
    for r in results:
        print(f"{r['name']:10s}  p={r['p']:.6f}, KS={r['ks']:.4f}")

    # 최적 분포 히스토그램 + PDF 시각화
    best = results[0]
    best_name = best["name"]
    best_dist = best["dist"]
    best_params = best["params"]

    x = np.linspace(data.min(), data.max(), 400)
    pdf = best_dist.pdf(x, *best_params)

    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=30, density=True, alpha=0.6,
             edgecolor='black', label='Observed')
    plt.plot(x, pdf, 'r-', lw=2, label=f'Best fit: {best_name}')
    plt.title(f"[{label}] Best Fitted: {best_name}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    return results

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    선택된 분포에 대해 P-P plot, Q-Q plot을 그려주는 함수.
    dist_params는 dist.fit 결과 튜플 (shape..., loc, scale)
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]
    data_sorted = np.sort(data)
    n = len(data_sorted)

    *shape_params, loc, scale = dist_params

    # -------- P-P Plot --------
    emp_cdf = (np.arange(1, n+1)) / (n + 0.0)
    theo_cdf = dist.cdf(data_sorted, *shape_params, loc=loc, scale=scale)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(theo_cdf, emp_cdf, alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel("Theoretical CDF")
    plt.ylabel("Empirical CDF")
    plt.title(f"{title_prefix} P-P Plot")
    plt.grid(True)

    # -------- Q-Q Plot --------
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)

    plt.subplot(1, 2, 2)
    plt.scatter(theo_q, data_sorted, alpha=0.6)
    min_q = min(theo_q.min(), data_sorted.min())
    max_q = max(theo_q.max(), data_sorted.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Empirical Quantiles")
    plt.title(f"{title_prefix} Q-Q Plot")
    plt.grid(True)

    plt.tight_layout()
    plt.show()


# =====================================================
# 1. 서비스타임 / 도착시간 → 초 단위 배열로 변환
# =====================================================

# --- Counter 1 서비스타임 (초) ---
c1_1006_service_secs = to_seconds_list(counter1_1006_service_time)
c1_1007_service_secs = to_seconds_list(counter1_1007_service_time)
c1_1008_service_secs = to_seconds_list(counter1_1008_service_time)
c1_1015_service_secs = to_seconds_list(counter1_1015_service_time)

# --- Counter 2 서비스타임 (초) ---
c2_1006_service_secs = to_seconds_list(counter2_1006_service_time)
c2_1007_service_secs = to_seconds_list(counter2_1007_service_time)
c2_1008_service_secs = to_seconds_list(counter2_1008_service_time)
c2_1015_service_secs = to_seconds_list(counter2_1015_service_time)

# --- Counter 1 도착시간 (초) ---
a1_1006_secs = np.sort(to_seconds_list(counter1_1006_arrival_time))
a1_1007_secs = np.sort(to_seconds_list(counter1_1007_arrival_time))
a1_1008_secs = np.sort(to_seconds_list(counter1_1008_arrival_time))
a1_1015_secs = np.sort(to_seconds_list(counter1_1015_arrival_time))

# --- Counter 2 도착시간 (초) ---
a2_1006_secs = np.sort(to_seconds_list(counter2_1006_arrival_time))
a2_1007_secs = np.sort(to_seconds_list(counter2_1007_arrival_time))
a2_1008_secs = np.sort(to_seconds_list(counter2_1008_arrival_time))
a2_1015_secs = np.sort(to_seconds_list(counter2_1015_arrival_time))

# --- Interarrival time (초) 계산: 같은 카운터 내에서 연속 도착 간격 ---
ia1_1006 = np.diff(a1_1006_secs)
ia1_1007 = np.diff(a1_1007_secs)
ia1_1008 = np.diff(a1_1008_secs)
ia1_1015 = np.diff(a1_1015_secs)

ia2_1006 = np.diff(a2_1006_secs)
ia2_1007 = np.diff(a2_1007_secs)
ia2_1008 = np.diff(a2_1008_secs)
ia2_1015 = np.diff(a2_1015_secs)

# '하루 전체' interarrival은 두 카운터 합친 걸 사용 (단순 concat)
ia_day_1006 = np.concatenate([ia1_1006, ia2_1006])
ia_day_1007 = np.concatenate([ia1_1007, ia2_1007])
ia_day_1008 = np.concatenate([ia1_1008, ia2_1008])
ia_day_1015 = np.concatenate([ia1_1015, ia2_1015])

# 4일 전체 interarrival (동질성 검정 후 한 덩어리로 쓸 예정)
ia_all = np.concatenate([ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015])

# =====================================================
# 2. 서비스타임: 일반일 vs 피크일 분리
# =====================================================

# 일반일 = 10/06, 10/07, 10/08 (두 카운터 모두)
st_normal = np.concatenate([
    c1_1006_service_secs, c1_1007_service_secs, c1_1008_service_secs,
    c2_1006_service_secs, c2_1007_service_secs, c2_1008_service_secs
])

# 피크일 = 10/15 (두 카운터 모두)
st_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# 일자별 서비스타임 (동질성 검정용)
st_day_1006 = np.concatenate([c1_1006_service_secs, c2_1006_service_secs])
st_day_1007 = np.concatenate([c1_1007_service_secs, c2_1007_service_secs])
st_day_1008 = np.concatenate([c1_1008_service_secs, c2_1008_service_secs])
st_day_1015 = np.concatenate([c1_1015_service_secs, c2_1015_service_secs])

# =====================================================
# 3. 독립성 검정 (서비스타임: Counter1 vs Counter2)
#    - 산점도 + Spearman 상관계수
# =====================================================

# 예시 1: 일반일(10/07) 서비스타임 독립성
min_len_0707 = min(len(c1_1007_service_secs), len(c2_1007_service_secs))
x_0707 = c1_1007_service_secs[:min_len_0707]
y_0707 = c2_1007_service_secs[:min_len_0707]

plt.figure(figsize=(5, 4))
plt.scatter(x_0707, y_0707, alpha=0.6)
plt.xlabel("Counter1 Service Time (s) - 10/07")
plt.ylabel("Counter2 Service Time (s) - 10/07")
plt.title("Scatter: Service Time (Non-Peak Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_0707, p_0707 = stats.spearmanr(x_0707, y_0707)
print(f"[Spearman] 10/07 ServiceTime C1 vs C2: rho={rho_0707:.3f}, p={p_0707:.4f}")

# 예시 2: 피크일(10/15) 서비스타임 독립성
min_len_1015 = min(len(c1_1015_service_secs), len(c2_1015_service_secs))
x_1015 = c1_1015_service_secs[:min_len_1015]
y_1015 = c2_1015_service_secs[:min_len_1015]

plt.figure(figsize=(5, 4))
plt.scatter(x_1015, y_1015, alpha=0.6, color='orange')
plt.xlabel("Counter1 Service Time (s) - 10/15")
plt.ylabel("Counter2 Service Time (s) - 10/15")
plt.title("Scatter: Service Time (Peak Day Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_1015, p_1015 = stats.spearmanr(x_1015, y_1015)
print(f"[Spearman] 10/15 ServiceTime C1 vs C2: rho={rho_1015:.3f}, p={p_1015:.4f}")

# (원하면 interarrival에 대해서도 같은 방식으로 Spearman 돌릴 수 있음)


# =====================================================
# 4. 동질성 검정 (여러 날 데이터 합쳐도 되는지)
#    - Interarrival, Service 둘 다 Kruskal–Wallis 사용
# =====================================================

# Interarrival 4일 동질성 (두 카운터 합친 하루 단위)
H_ia, p_ia = stats.kruskal(ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015)
print(f"\n[Interarrival] Kruskal–Wallis: H={H_ia:.3f}, p={p_ia:.6f}")

# Service 4일 동질성
H_st, p_st = stats.kruskal(st_day_1006, st_day_1007, st_day_1008, st_day_1015)
print(f"[Service] Kruskal–Wallis: H={H_st:.3f}, p={p_st:.6f}")

# p_ia > 0.05 라면 → interarrival는 4일 합쳐서 사용 OK
# p_st << 0.05 라면 → 서비스타임은 일반일 vs 피크일 분리 필요


# =====================================================
# 5. Tukey boxplot으로 이상치 확인 (Service Normal / Peak)
# =====================================================

for data, name in [(st_normal, "Normal Service Time"),
                   (st_peak, "Peak Service Time")]:

    low, up = tukey_bounds(data)
    outliers = data[(data < low) | (data > up)]

    print(f"\n[{name}] Tukey bounds: {low:.1f} ~ {up:.1f}")
    print(f"  이상치 개수: {len(outliers)}")

    plt.figure(figsize=(4, 5))
    plt.boxplot(data, vert=True, showfliers=True)
    plt.title(f"{name} (Tukey boxplot)")
    plt.ylabel("Service Time (s)")
    plt.tight_layout()
    plt.show()


# =====================================================
# 6. 분포 피팅 Activity 1
#    - Interarrival 전체
#    - Service Normal (10/06~08)
#    - Service Peak (10/15)
# =====================================================

res_ia       = fit_and_compare_distributions(ia_all, label="Interarrival (All days)")
res_st_norm  = fit_and_compare_distributions(st_normal, label="Service Time - Normal Days")
res_st_peak  = fit_and_compare_distributions(st_peak, label="Service Time - Peak Day")

best_ia      = res_ia[0]
best_norm    = res_st_norm[0]
best_peak    = res_st_peak[0]

print("\n[선정된 최상위 후보 분포들]")
print("Interarrival :", best_ia["name"],   best_ia["params"])
print("Service Normal:", best_norm["name"], best_norm["params"])
print("Service Peak  :", best_peak["name"], best_peak["params"])


# =====================================================
# 7. 선정된 분포에 대한 P-P / Q-Q Plot (정교한 적합도 확인)
# =====================================================

# Interarrival
plot_pp_qq(ia_all, best_ia["dist"], best_ia["params"],
           title_prefix=f"Interarrival ({best_ia['name']})")

# Service Normal
plot_pp_qq(st_normal, best_norm["dist"], best_norm["params"],
           title_prefix=f"Service Normal ({best_norm['name']})")

# Service Peak
plot_pp_qq(st_peak, best_peak["dist"], best_peak["params"],
           title_prefix=f"Service Peak ({best_peak['name']})")


# =====================================================
# 8. (참고) SimPy에서 쓸 수 있는 형태로 파라미터 정리 예시
#    - 여기서는 그냥 print만, 실제 SimPy 코드에서 rvs()로 사용
# =====================================================

# 예: interarrival가 expon이면
if best_ia["name"] == "expon":
    loc_ia, scale_ia = best_ia["params"][1], best_ia["params"][2]
    print("\n[SimPy용 Interarrival 예시] stats.expon.rvs(loc=%.2f, scale=%.2f)" % (loc_ia, scale_ia))

# 예: 일반일 서비스타임이 gamma인 경우
if best_norm["name"] == "gamma":
    shape_n, loc_n, scale_n = best_norm["params"]
    print("[SimPy용 Service Normal 예시] stats.gamma.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (shape_n, loc_n, scale_n))

# 예: 피크일 서비스타임이 lognorm인 경우
if best_peak["name"] == "lognorm":
    s_p, loc_p, scale_p = best_peak["params"]
    print("[SimPy용 Service Peak 예시] stats.lognorm.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (s_p, loc_p, scale_p))

```

```python

from scipy.stats import kruskal

# HH:MM:SS → 초 단위로 변환
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 도착시간 리스트 → 정렬 → interarrival time (초) 리스트 반환
def get_interarrival(arrival_time_list):
    # 1) 초단위로 변환
    secs = [hhmmss_to_sec(t) for t in arrival_time_list]
    # 2) 혹시 순서가 뒤섞였을 수 있으니 정렬
    secs_sorted = sorted(secs)
    # 3) 인접 차이 = interarrival
    interarrival = [
        secs_sorted[i] - secs_sorted[i-1]
        for i in range(1, len(secs_sorted))
    ]
    return interarrival
# ------------------------
# Counter 1: 날짜별 interarrival
# ------------------------
c1_1006_ia = get_interarrival(counter1_1006_arrival_time)
c1_1007_ia = get_interarrival(counter1_1007_arrival_time)
c1_1008_ia = get_interarrival(counter1_1008_arrival_time)
c1_1015_ia = get_interarrival(counter1_1015_arrival_time)

print("[Counter 1] interarrival 개수")
print("10/06:", len(c1_1006_ia))
print("10/07:", len(c1_1007_ia))
print("10/08:", len(c1_1008_ia))
print("10/15:", len(c1_1015_ia))

# ------------------------
# Counter 2: 날짜별 interarrival
# ------------------------
c2_1006_ia = get_interarrival(counter2_1006_arrival_time)
c2_1007_ia = get_interarrival(counter2_1007_arrival_time)
c2_1008_ia = get_interarrival(counter2_1008_arrival_time)
c2_1015_ia = get_interarrival(counter2_1015_arrival_time)

print("\n[Counter 2] interarrival 개수")
print("10/06:", len(c2_1006_ia))
print("10/07:", len(c2_1007_ia))
print("10/08:", len(c2_1008_ia))
print("10/15:", len(c2_1015_ia))

alpha = 0.05  # 유의수준


# ==============================
# ① Counter 1 : 4일 동질성 검정
# ==============================
stat_c1, p_c1 = kruskal(c1_1006_ia, c1_1007_ia, c1_1008_ia, c1_1015_ia)

print("\n==============================")
print("[Counter 1] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat_c1:.4f}")
print(f"p-value  = {p_c1:.6f}")

if p_c1 > alpha:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각 불가")
    print("   ⇒ Counter 1의 4일 interarrival 분포는 동질적이라고 볼 수 있음.")
    print("   ⇒ Counter 1은 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각")
    print("   ⇒ Counter 1의 4일 중 적어도 하루의 interarrival 분포가 다름.")
    print("   ⇒ Counter 1은 날짜를 나누거나 원인 분석 필요.")


# ==============================
# ② Counter 2 : 4일 동질성 검정
# ==============================
stat_c2, p_c2 = kruskal(c2_1006_ia, c2_1007_ia, c2_1008_ia, c2_1015_ia)

print("\n==============================")
print("[Counter 2] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat_c2:.4f}")
print(f"p-value  = {p_c2:.6f}")

if p_c2 > alpha:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각 불가")
    print("   ⇒ Counter 2의 4일 interarrival 분포는 동질적이라고 볼 수 있음.")
    print("   ⇒ Counter 2도 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각")
    print("   ⇒ Counter 2의 4일 중 적어도 하루의 interarrival 분포가 다름.")
    print("   ⇒ Counter 2 역시 날짜를 나누거나 원인 분석 필요.")

import numpy as np

# HH:MM:SS → 초로 변환
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 도착시간 리스트 두 개(1번, 2번)를 합쳐서 → 정렬 → interarrival(초) 리스트
def get_daily_interarrival(counter1_times, counter2_times):
    # 1,2번 합쳐서 초 단위로 변환
    secs = [hhmmss_to_sec(t) for t in counter1_times] + \
           [hhmmss_to_sec(t) for t in counter2_times]
    # 시간순 정렬 (리스트에 약간 순서 꼬인 게 있어서 꼭 정렬해줘야 함!)
    secs_sorted = sorted(secs)
    # 인접 고객 도착 간격 = interarrival time
    inter = [secs_sorted[i] - secs_sorted[i-1] for i in range(1, len(secs_sorted))]
    return inter

# --- 1) 4일치 interarrival(일자별) 계산 ---
ia_1006 = get_daily_interarrival(counter1_1006_arrival_time, counter2_1006_arrival_time)
ia_1007 = get_daily_interarrival(counter1_1007_arrival_time, counter2_1007_arrival_time)
ia_1008 = get_daily_interarrival(counter1_1008_arrival_time, counter2_1008_arrival_time)
ia_1015 = get_daily_interarrival(counter1_1015_arrival_time, counter2_1015_arrival_time)

print("10/06 interarrival 개수:", len(ia_1006))
print("10/07 interarrival 개수:", len(ia_1007))
print("10/08 interarrival 개수:", len(ia_1008))
print("10/15 interarrival 개수:", len(ia_1015))


# --- 2) interarrival 값 자체 출력 (초 단위 & 분 단위) ---
def print_interarrivals(label, ia_list):
    print(f"\n=== {label} (n={len(ia_list)}) ===")
    print("초 단위 리스트:")
    print(ia_list)   # 이게 네가 말한 '값 자체'

    mins = [x/60 for x in ia_list]
    print("\n분 단위 리스트(=초/60):")
    print(mins)

    print("\n요약 통계:")
    print("  평균(초):", np.mean(ia_list))
    print("  표준편차(초):", np.std(ia_list))
    print("  최소/최대(초):", np.min(ia_list), "/", np.max(ia_list))

print_interarrivals("10/06 전체(1+2번)", ia_1006)
print_interarrivals("10/07 전체(1+2번)", ia_1007)
print_interarrivals("10/08 전체(1+2번)", ia_1008)
print_interarrivals("10/15 전체(1+2번)", ia_1015)

def get_interarrival_single_counter(arrival_times):
    secs = sorted(hhmmss_to_sec(t) for t in arrival_times)
    return [secs[i] - secs[i-1] for i in range(1, len(secs))]

c1_1006_ia = get_interarrival_single_counter(counter1_1006_arrival_time)
# 이런 식으로 나머지도 계산 가능

# 위에서 구한 ia_1006, ia_1007, ia_1008, ia_1015 재사용

all_ia = ia_1006 + ia_1007 + ia_1008 + ia_1015  # 4일 전체 interarrival (초 단위)
all_ia = np.array(all_ia)

print("4일 전체 interarrival 개수:", len(all_ia))
print("평균(초):", np.mean(all_ia))
print("표준편차(초):", np.std(all_ia))

import matplotlib.pyplot as plt
from scipy import stats

# 0보다 큰 값만 사용 (혹시 0이 있으면 분포 적합에 문제될 수 있어서 필터링)
all_ia_pos = all_ia[all_ia > 0]

# --- 1) 히스토그램 그리기 ---
plt.hist(all_ia_pos, bins=20, density=True, alpha=0.5)
plt.xlabel("Interarrival time (sec)")
plt.ylabel("Density")
plt.title("Histogram of Interarrival Times (4 days, both counters)")
plt.show()

# --- 2) 여러 분포 피팅해보기 ---
candidates = {
    "expon": stats.expon,          # 지수분포
    "gamma": stats.gamma,          # 감마분포
    "lognorm": stats.lognorm,      # 로그정규
    "weibull_min": stats.weibull_min  # Weibull 최소형
}

fit_results = {}

for name, dist in candidates.items():
    # 파라미터 적합
    params = dist.fit(all_ia_pos)
    fit_results[name] = params

    # K-S 검정으로 적합도 대략 보기
    ks_stat, p_val = stats.kstest(all_ia_pos, name, args=params)
    print(f"\n[{name}]")
    print("  params:", params)
    print("  KS stat:", ks_stat)
    print("  p-value:", p_val)

# --- 3) 히스토그램 위에 분포 곡선 그려보기 (시각적으로 비교) ---
xmin, xmax = all_ia_pos.min(), all_ia_pos.max()
x = np.linspace(xmin, xmax, 200)

plt.hist(all_ia_pos, bins=20, density=True, alpha=0.5, label="Data")

for name, dist in candidates.items():
    params = fit_results[name]
    pdf = dist.pdf(x, *params)
    plt.plot(x, pdf, label=name)

plt.xlabel("Interarrival time (sec)")
plt.ylabel("Density")
plt.title("Fitted Distributions to Interarrival Times")
plt.legend()
plt.show()

from scipy.stats import kruskal
import numpy as np

# HH:MM:SS → 초 변환 함수
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 하루 서비스타임 = 1번 + 2번 합치기
def get_daily_service_times(counter1_service, counter2_service):
    secs_c1 = [hhmmss_to_sec(t) for t in counter1_service]
    secs_c2 = [hhmmss_to_sec(t) for t in counter2_service]
    return secs_c1 + secs_c2


# --- 4일 서비스타임 데이터 준비 ---
st_1006 = get_daily_service_times(counter1_1006_service_time, counter2_1006_service_time)
st_1007 = get_daily_service_times(counter1_1007_service_time, counter2_1007_service_time)
st_1008 = get_daily_service_times(counter1_1008_service_time, counter2_1008_service_time)
st_1015 = get_daily_service_times(counter1_1015_service_time, counter2_1015_service_time)

print("10/06 서비스타임 개수:", len(st_1006))
print("10/07 서비스타임 개수:", len(st_1007))
print("10/08 서비스타임 개수:", len(st_1008))
print("10/15 서비스타임 개수:", len(st_1015))


# --- 동질성 검정(Kruskal–Wallis) ---
alpha = 0.05

stat, p_val = kruskal(st_1006, st_1007, st_1008, st_1015)

print("\n==============================")
print("[서비스타임] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat:.4f}")
print(f"p-value  = {p_val:.6f}")

if p_val > alpha:
    print(f"\n→ p-value > {alpha} 이므로 귀무가설 기각 불가")
    print("   ⇒ 4일 서비스타임 분포는 '동질적'이다.")
    print("   ⇒ 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"\n→ p-value ≤ {alpha} 이므로 귀무가설 기각")
    print("   ⇒ 4일 중 적어도 하루의 서비스타임 분포가 다르다.")
    print("   ⇒ 4일을 합쳐 쓰면 안 되고, 원인 분석 또는 분리 필요.")

import numpy as np

def summary(label, data):
    data = np.array(data)
    print(f"\n=== {label} ===")
    print("n :", len(data))
    print("평균(초):", np.mean(data))
    print("표준편차(초):", np.std(data))
    print("최소/최대(초):", np.min(data), "/", np.max(data))

summary("10/06 서비스타임", st_1006)
summary("10/07 서비스타임", st_1007)
summary("10/08 서비스타임", st_1008)
summary("10/15 서비스타임", st_1015)

from scipy.stats import kruskal

pairs = [
    ("10/06", st_1006, "10/07", st_1007),
    ("10/06", st_1006, "10/08", st_1008),
    ("10/06", st_1006, "10/15", st_1015),
    ("10/07", st_1007, "10/08", st_1008),
    ("10/07", st_1007, "10/15", st_1015),
    ("10/08", st_1008, "10/15", st_1015),
]

alpha = 0.05

print("\n=== 날짜 쌍별 Kruskal–Wallis 검정 (서비스타임) ===")
for a_label, a, b_label, b in pairs:
    stat, p = kruskal(a, b)
    print(f"{a_label} vs {b_label} → H={stat:.3f}, p={p:.6f}", end="  ")
    if p > alpha:
        print("→ 차이 없음 (동질)")
    else:
        print("→ 차이 있음 (이질)")

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# =====================================
# 1. HH:MM:SS → 초 변환 함수
# =====================================
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# =====================================
# 2. Interarrival 계산 함수
# =====================================
def interarrival_from_times(time_list):
    """
    time_list: ['14:32:36', '14:35:11', ...]
    return: 연속 도착 간격(초) np.array, 길이는 n-1
    """
    secs = np.array([hhmmss_to_sec(t) for t in time_list], dtype=float)
    ia = np.diff(secs)   # 앞뒤 시간 차이
    return ia

# =====================================
# 3. 기술통계 함수
# =====================================
def describe_data(data, label="Data"):
    data = np.array(data, dtype=float)
    data = data[~np.isnan(data)]

    n = len(data)
    if n == 0:
        print(f"\n[{label}] 데이터 없음")
        return

    mean = np.mean(data)
    std = np.std(data, ddof=1)          # 표본 표준편차
    min_v = np.min(data)
    q1 = np.percentile(data, 25)
    median = np.percentile(data, 50)
    q3 = np.percentile(data, 75)
    max_v = np.max(data)
    iqr = q3 - q1
    cv = std / mean if mean != 0 else np.nan
    skew = stats.skew(data)
    kurt = stats.kurtosis(data)         # excess kurtosis

    print(f"\n===== [{label}] 기술통계 =====")
    print(f"N            : {n}")
    print(f"Mean         : {mean:8.3f}")
    print(f"Std          : {std:8.3f}")
    print(f"Min          : {min_v:8.3f}")
    print(f"Q1 (25%)     : {q1:8.3f}")
    print(f"Median (50%) : {median:8.3f}")
    print(f"Q3 (75%)     : {q3:8.3f}")
    print(f"Max          : {max_v:8.3f}")
    print(f"IQR          : {iqr:8.3f}")
    print(f"CV           : {cv:8.3f}")
    print(f"Skewness     : {skew:8.3f}")
    print(f"Kurtosis     : {kurt:8.3f}")

# =====================================
# 4. 서비스시간을 초로 변환
# =====================================

# 10/06
c1_1006_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1006_service_time])
c2_1006_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1006_service_time])

# 10/07
c1_1007_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1007_service_time])
c2_1007_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1007_service_time])

# 10/08
c1_1008_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1008_service_time])
c2_1008_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1008_service_time])

# 10/15
c1_1015_service_secs = np.array([hhmmss_to_sec(t) for t in counter1_1015_service_time])
c2_1015_service_secs = np.array([hhmmss_to_sec(t) for t in counter2_1015_service_time])

# =====================================
# 5. Interarrival Time 계산 (도착시간 기반)
# =====================================

# 10/06
ia_c1_1006 = interarrival_from_times(counter1_1006_arrival_time)
ia_c2_1006 = interarrival_from_times(counter2_1006_arrival_time)

# 10/07
ia_c1_1007 = interarrival_from_times(counter1_1007_arrival_time)
ia_c2_1007 = interarrival_from_times(counter2_1007_arrival_time)

# 10/08
ia_c1_1008 = interarrival_from_times(counter1_1008_arrival_time)
ia_c2_1008 = interarrival_from_times(counter2_1008_arrival_time)

# 10/15
ia_c1_1015 = interarrival_from_times(counter1_1015_arrival_time)
ia_c2_1015 = interarrival_from_times(counter2_1015_arrival_time)

# =====================================
# 6. 분석용 그룹 묶기
# =====================================

# 6-1. 서비스시간: 일반일(10/06~10/08) vs 피크일(10/15)
service_normal = np.concatenate([
    c1_1006_service_secs, c2_1006_service_secs,
    c1_1007_service_secs, c2_1007_service_secs,
    c1_1008_service_secs, c2_1008_service_secs
])

service_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# 6-2. Interarrival: 전체 / 일반일 / 피크일
ia_normal = np.concatenate([
    ia_c1_1006, ia_c2_1006,
    ia_c1_1007, ia_c2_1007,
    ia_c1_1008, ia_c2_1008
])

ia_peak = np.concatenate([
    ia_c1_1015, ia_c2_1015
])

ia_all = np.concatenate([ia_normal, ia_peak])

# =====================================
# 7. 기술통계 출력
# =====================================

# --- 서비스타임 ---
describe_data(service_normal, "Service Time - Normal Days (10/06~10/08, C1+C2)")
describe_data(service_peak,   "Service Time - Peak Day (10/15, C1+C2)")

describe_data(c1_1006_service_secs, "Service Time - 10/06 Counter1")
describe_data(c2_1006_service_secs, "Service Time - 10/06 Counter2")

# --- Interarrival Time ---
describe_data(ia_all,    "Interarrival - All Days (C1+C2)")
describe_data(ia_normal, "Interarrival - Normal Days (10/06~10/08)")
describe_data(ia_peak,   "Interarrival - Peak Day (10/15)")

# =====================================
# 8. 시각화 함수 (히스토그램 + 박스플롯)
# =====================================

def plot_histogram(data_list, labels, title, xlabel):
    """
    data_list: [array1, array2, ...]
    labels   : ['normal', 'peak', ...]
    """
    plt.figure(figsize=(10, 5))
    for d, lab in zip(data_list, labels):
        plt.hist(d, bins=30, alpha=0.5, density=True, label=lab)
    plt.xlabel(xlabel)
    plt.ylabel("Density")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_boxplot(data_list, labels, title, ylabel):
    plt.figure(figsize=(8, 5))
    plt.boxplot(data_list, labels=labels, showfliers=True)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    plt.show()

# =====================================
# 9. 실제 그래프 출력
# =====================================

# 9-1. 서비스타임 히스토그램 & 박스플롯 (일반일 vs 피크일)
plot_histogram(
    [service_normal, service_peak],
    ["Normal Days", "Peak Day"],
    "Service Time Distribution (Normal vs Peak)",
    "Service Time (seconds)"
)

plot_boxplot(
    [service_normal, service_peak],
    ["Normal", "Peak"],
    "Service Time Boxplot (Normal vs Peak)",
    "Service Time (seconds)"
)

# 9-2. Interarrival 히스토그램 & 박스플롯 (일반일 vs 피크일)
plot_histogram(
    [ia_normal, ia_peak],
    ["Normal Days", "Peak Day"],
    "Interarrival Time Distribution (Normal vs Peak)",
    "Interarrival Time (seconds)"
)

plot_boxplot(
    [ia_normal, ia_peak],
    ["Normal", "Peak"],
    "Interarrival Time Boxplot (Normal vs Peak)",
    "Interarrival Time (seconds)"
)

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import gamma, lognorm, expon, weibull_min

# =====================================================
# 0. 공통 도우미 함수들
# =====================================================

def hhmmss_to_sec(t):
    """'HH:MM:SS' 문자열을 초 단위 숫자로 변환"""
    h, m, s = map(int, t.split(':'))
    return h*3600 + m*60 + s

def to_seconds_list(str_list):
    """시간 문자열 리스트를 초 단위 numpy 배열로 변환"""
    return np.array([hhmmss_to_sec(t) for t in str_list], dtype=float)

def tukey_bounds(data, k=1.5):
    """Tukey IQR 기준으로 이상치 경계 계산 (Q1-1.5IQR, Q3+1.5IQR)"""
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

def fit_and_compare_distributions(data, label=""):
    """
    여러 분포(expon, gamma, lognorm, weibull_min)를 피팅하고
    KS-test, loglik를 출력. p-value 기준 정렬하여 반환.
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]  # 0 이하 제거
    candidates = {
        "expon": expon,
        "gamma": gamma,
        "lognorm": lognorm,
        "weibull_min": weibull_min
    }

    results = []
    print(f"\n===== [{label}] 분포 적합 결과 =====")
    for name, dist in candidates.items():
        # 서비스/도착시간은 0 이상이므로 loc=0 고정
        params = dist.fit(data, floc=0)
        ks, p = stats.kstest(data, dist.cdf, args=params)
        loglik = np.sum(dist.logpdf(data, *params))
        results.append({
            "name": name,
            "dist": dist,
            "params": params,
            "ks": ks,
            "p": p,
            "loglik": loglik
        })
        print(f"{name:10s} | KS={ks:.4f}, p={p:.6f}, loglik={loglik:.1f}")
        print(f"  params = {params}\n")

    results.sort(key=lambda x: x["p"], reverse=True)

    print(f"\n>>> [{label}] p-value 순(적합도 순) <<<")
    for r in results:
        print(f"{r['name']:10s}  p={r['p']:.6f}, KS={r['ks']:.4f}")

    # 최적 분포 히스토그램 + PDF 시각화
    best = results[0]
    best_name = best["name"]
    best_dist = best["dist"]
    best_params = best["params"]

    x = np.linspace(data.min(), data.max(), 400)
    pdf = best_dist.pdf(x, *best_params)

    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=30, density=True, alpha=0.6,
             edgecolor='black', label='Observed')
    plt.plot(x, pdf, 'r-', lw=2, label=f'Best fit: {best_name}')
    plt.title(f"[{label}] Best Fitted: {best_name}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    return results

def plot_pp_qq(data, dist, dist_params, title_prefix):
    """
    선택된 분포에 대해 P-P plot, Q-Q plot을 그려주는 함수.
    dist_params는 dist.fit 결과 튜플 (shape..., loc, scale)
    """
    data = np.asarray(data, dtype=float)
    data = data[data > 0]
    data_sorted = np.sort(data)
    n = len(data_sorted)

    *shape_params, loc, scale = dist_params

    # -------- P-P Plot --------
    emp_cdf = (np.arange(1, n+1)) / (n + 0.0)
    theo_cdf = dist.cdf(data_sorted, *shape_params, loc=loc, scale=scale)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(theo_cdf, emp_cdf, alpha=0.6)
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel("Theoretical CDF")
    plt.ylabel("Empirical CDF")
    plt.title(f"{title_prefix} P-P Plot")
    plt.grid(True)

    # -------- Q-Q Plot --------
    probs = (np.arange(1, n+1) - 0.5) / n
    theo_q = dist.ppf(probs, *shape_params, loc=loc, scale=scale)

    plt.subplot(1, 2, 2)
    plt.scatter(theo_q, data_sorted, alpha=0.6)
    min_q = min(theo_q.min(), data_sorted.min())
    max_q = max(theo_q.max(), data_sorted.max())
    plt.plot([min_q, max_q], [min_q, max_q], 'r--')
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Empirical Quantiles")
    plt.title(f"{title_prefix} Q-Q Plot")
    plt.grid(True)

    plt.tight_layout()
    plt.show()


# =====================================================
# 1. 서비스타임 / 도착시간 → 초 단위 배열로 변환
# =====================================================

# --- Counter 1 서비스타임 (초) ---
c1_1006_service_secs = to_seconds_list(counter1_1006_service_time)
c1_1007_service_secs = to_seconds_list(counter1_1007_service_time)
c1_1008_service_secs = to_seconds_list(counter1_1008_service_time)
c1_1015_service_secs = to_seconds_list(counter1_1015_service_time)

# --- Counter 2 서비스타임 (초) ---
c2_1006_service_secs = to_seconds_list(counter2_1006_service_time)
c2_1007_service_secs = to_seconds_list(counter2_1007_service_time)
c2_1008_service_secs = to_seconds_list(counter2_1008_service_time)
c2_1015_service_secs = to_seconds_list(counter2_1015_service_time)

# --- Counter 1 도착시간 (초) ---
a1_1006_secs = np.sort(to_seconds_list(counter1_1006_arrival_time))
a1_1007_secs = np.sort(to_seconds_list(counter1_1007_arrival_time))
a1_1008_secs = np.sort(to_seconds_list(counter1_1008_arrival_time))
a1_1015_secs = np.sort(to_seconds_list(counter1_1015_arrival_time))

# --- Counter 2 도착시간 (초) ---
a2_1006_secs = np.sort(to_seconds_list(counter2_1006_arrival_time))
a2_1007_secs = np.sort(to_seconds_list(counter2_1007_arrival_time))
a2_1008_secs = np.sort(to_seconds_list(counter2_1008_arrival_time))
a2_1015_secs = np.sort(to_seconds_list(counter2_1015_arrival_time))

# --- Interarrival time (초) 계산: 같은 카운터 내에서 연속 도착 간격 ---
ia1_1006 = np.diff(a1_1006_secs)
ia1_1007 = np.diff(a1_1007_secs)
ia1_1008 = np.diff(a1_1008_secs)
ia1_1015 = np.diff(a1_1015_secs)

ia2_1006 = np.diff(a2_1006_secs)
ia2_1007 = np.diff(a2_1007_secs)
ia2_1008 = np.diff(a2_1008_secs)
ia2_1015 = np.diff(a2_1015_secs)

# '하루 전체' interarrival은 두 카운터 합친 걸 사용 (단순 concat)
ia_day_1006 = np.concatenate([ia1_1006, ia2_1006])
ia_day_1007 = np.concatenate([ia1_1007, ia2_1007])
ia_day_1008 = np.concatenate([ia1_1008, ia2_1008])
ia_day_1015 = np.concatenate([ia1_1015, ia2_1015])

# 4일 전체 interarrival (동질성 검정 후 한 덩어리로 쓸 예정)
ia_all = np.concatenate([ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015])

# =====================================================
# 2. 서비스타임: 일반일 vs 피크일 분리
# =====================================================

# 일반일 = 10/06, 10/07, 10/08 (두 카운터 모두)
st_normal = np.concatenate([
    c1_1006_service_secs, c1_1007_service_secs, c1_1008_service_secs,
    c2_1006_service_secs, c2_1007_service_secs, c2_1008_service_secs
])

# 피크일 = 10/15 (두 카운터 모두)
st_peak = np.concatenate([
    c1_1015_service_secs, c2_1015_service_secs
])

# 일자별 서비스타임 (동질성 검정용)
st_day_1006 = np.concatenate([c1_1006_service_secs, c2_1006_service_secs])
st_day_1007 = np.concatenate([c1_1007_service_secs, c2_1007_service_secs])
st_day_1008 = np.concatenate([c1_1008_service_secs, c2_1008_service_secs])
st_day_1015 = np.concatenate([c1_1015_service_secs, c2_1015_service_secs])

# =====================================================
# 3. 독립성 검정 (서비스타임: Counter1 vs Counter2)
#    - 산점도 + Spearman 상관계수
# =====================================================

# 예시 1: 일반일(10/07) 서비스타임 독립성
min_len_0707 = min(len(c1_1007_service_secs), len(c2_1007_service_secs))
x_0707 = c1_1007_service_secs[:min_len_0707]
y_0707 = c2_1007_service_secs[:min_len_0707]

plt.figure(figsize=(5, 4))
plt.scatter(x_0707, y_0707, alpha=0.6)
plt.xlabel("Counter1 Service Time (s) - 10/07")
plt.ylabel("Counter2 Service Time (s) - 10/07")
plt.title("Scatter: Service Time (Non-Peak Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_0707, p_0707 = stats.spearmanr(x_0707, y_0707)
print(f"[Spearman] 10/07 ServiceTime C1 vs C2: rho={rho_0707:.3f}, p={p_0707:.4f}")

# 예시 2: 피크일(10/15) 서비스타임 독립성
min_len_1015 = min(len(c1_1015_service_secs), len(c2_1015_service_secs))
x_1015 = c1_1015_service_secs[:min_len_1015]
y_1015 = c2_1015_service_secs[:min_len_1015]

plt.figure(figsize=(5, 4))
plt.scatter(x_1015, y_1015, alpha=0.6, color='orange')
plt.xlabel("Counter1 Service Time (s) - 10/15")
plt.ylabel("Counter2 Service Time (s) - 10/15")
plt.title("Scatter: Service Time (Peak Day Example)")
plt.grid(True)
plt.tight_layout()
plt.show()

rho_1015, p_1015 = stats.spearmanr(x_1015, y_1015)
print(f"[Spearman] 10/15 ServiceTime C1 vs C2: rho={rho_1015:.3f}, p={p_1015:.4f}")

# (원하면 interarrival에 대해서도 같은 방식으로 Spearman 돌릴 수 있음)


# =====================================================
# 4. 동질성 검정 (여러 날 데이터 합쳐도 되는지)
#    - Interarrival, Service 둘 다 Kruskal–Wallis 사용
# =====================================================

# Interarrival 4일 동질성 (두 카운터 합친 하루 단위)
H_ia, p_ia = stats.kruskal(ia_day_1006, ia_day_1007, ia_day_1008, ia_day_1015)
print(f"\n[Interarrival] Kruskal–Wallis: H={H_ia:.3f}, p={p_ia:.6f}")

# Service 4일 동질성
H_st, p_st = stats.kruskal(st_day_1006, st_day_1007, st_day_1008, st_day_1015)
print(f"[Service] Kruskal–Wallis: H={H_st:.3f}, p={p_st:.6f}")

# p_ia > 0.05 라면 → interarrival는 4일 합쳐서 사용 OK
# p_st << 0.05 라면 → 서비스타임은 일반일 vs 피크일 분리 필요


# =====================================================
# 5. Tukey boxplot으로 이상치 확인 (Service Normal / Peak)
# =====================================================

for data, name in [(st_normal, "Normal Service Time"),
                   (st_peak, "Peak Service Time")]:

    low, up = tukey_bounds(data)
    outliers = data[(data < low) | (data > up)]

    print(f"\n[{name}] Tukey bounds: {low:.1f} ~ {up:.1f}")
    print(f"  이상치 개수: {len(outliers)}")

    plt.figure(figsize=(4, 5))
    plt.boxplot(data, vert=True, showfliers=True)
    plt.title(f"{name} (Tukey boxplot)")
    plt.ylabel("Service Time (s)")
    plt.tight_layout()
    plt.show()


# =====================================================
# 6. 분포 피팅 Activity 1
#    - Interarrival 전체
#    - Service Normal (10/06~08)
#    - Service Peak (10/15)
# =====================================================

res_ia       = fit_and_compare_distributions(ia_all, label="Interarrival (All days)")
res_st_norm  = fit_and_compare_distributions(st_normal, label="Service Time - Normal Days")
res_st_peak  = fit_and_compare_distributions(st_peak, label="Service Time - Peak Day")

best_ia      = res_ia[0]
best_norm    = res_st_norm[0]
best_peak    = res_st_peak[0]

print("\n[선정된 최상위 후보 분포들]")
print("Interarrival :", best_ia["name"],   best_ia["params"])
print("Service Normal:", best_norm["name"], best_norm["params"])
print("Service Peak  :", best_peak["name"], best_peak["params"])


# =====================================================
# 7. 선정된 분포에 대한 P-P / Q-Q Plot (정교한 적합도 확인)
# =====================================================

# Interarrival
plot_pp_qq(ia_all, best_ia["dist"], best_ia["params"],
           title_prefix=f"Interarrival ({best_ia['name']})")

# Service Normal
plot_pp_qq(st_normal, best_norm["dist"], best_norm["params"],
           title_prefix=f"Service Normal ({best_norm['name']})")

# Service Peak
plot_pp_qq(st_peak, best_peak["dist"], best_peak["params"],
           title_prefix=f"Service Peak ({best_peak['name']})")


# =====================================================
# 8. (참고) SimPy에서 쓸 수 있는 형태로 파라미터 정리 예시
#    - 여기서는 그냥 print만, 실제 SimPy 코드에서 rvs()로 사용
# =====================================================

# 예: interarrival가 expon이면
if best_ia["name"] == "expon":
    loc_ia, scale_ia = best_ia["params"][1], best_ia["params"][2]
    print("\n[SimPy용 Interarrival 예시] stats.expon.rvs(loc=%.2f, scale=%.2f)" % (loc_ia, scale_ia))

# 예: 일반일 서비스타임이 gamma인 경우
if best_norm["name"] == "gamma":
    shape_n, loc_n, scale_n = best_norm["params"]
    print("[SimPy용 Service Normal 예시] stats.gamma.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (shape_n, loc_n, scale_n))

# 예: 피크일 서비스타임이 lognorm인 경우
if best_peak["name"] == "lognorm":
    s_p, loc_p, scale_p = best_peak["params"]
    print("[SimPy용 Service Peak 예시] stats.lognorm.rvs(%.3f, loc=%.2f, scale=%.2f)" %
          (s_p, loc_p, scale_p))

```
