```python
alpha = 0.05  # 유의수준

# ==============================
# ① Counter 1 : 4일 동질성 검정
# ==============================
stat_c1, p_c1 = kruskal(c1_1006_ia, c1_1007_ia, c1_1008_ia, c1_1015_ia)

print("\n==============================")
print("[Counter 1] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat_c1:.4f}")
print(f"p-value  = {p_c1:.6f}")

if p_c1 > alpha:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각 불가")
    print("   ⇒ Counter 1의 4일 interarrival 분포는 동질적이라고 볼 수 있음.")
    print("   ⇒ Counter 1은 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각")
    print("   ⇒ Counter 1의 4일 중 적어도 하루의 interarrival 분포가 다름.")
    print("   ⇒ Counter 1은 날짜를 나누거나 원인 분석 필요.")


# ==============================
# ② Counter 2 : 4일 동질성 검정
# ==============================
stat_c2, p_c2 = kruskal(c2_1006_ia, c2_1007_ia, c2_1008_ia, c2_1015_ia)

print("\n==============================")
print("[Counter 2] 4일 Kruskal–Wallis 동질성 검정")
print(f"H 통계량 = {stat_c2:.4f}")
print(f"p-value  = {p_c2:.6f}")

if p_c2 > alpha:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각 불가")
    print("   ⇒ Counter 2의 4일 interarrival 분포는 동질적이라고 볼 수 있음.")
    print("   ⇒ Counter 2도 4일 데이터를 합쳐서 하나의 모집단으로 사용 가능.")
else:
    print(f"→ 유의수준 {alpha}에서 귀무가설 기각")
    print("   ⇒ Counter 2의 4일 중 적어도 하루의 interarrival 분포가 다름.")
    print("   ⇒ Counter 2 역시 날짜를 나누거나 원인 분석 필요.")

import numpy as np

# HH:MM:SS → 초로 변환
def hhmmss_to_sec(t):
    h, m, s = map(int, t.split(':'))
    return h * 3600 + m * 60 + s

# 도착시간 리스트 두 개(1번, 2번)를 합쳐서 → 정렬 → interarrival(초) 리스트
def get_daily_interarrival(counter1_times, counter2_times):
    # 1,2번 합쳐서 초 단위로 변환
    secs = [hhmmss_to_sec(t) for t in counter1_times] + \
           [hhmmss_to_sec(t) for t in counter2_times]
    # 시간순 정렬 (리스트에 약간 순서 꼬인 게 있어서 꼭 정렬해줘야 함!)
    secs_sorted = sorted(secs)
    # 인접 고객 도착 간격 = interarrival time
    inter = [secs_sorted[i] - secs_sorted[i-1] for i in range(1, len(secs_sorted))]
    return inter

# --- 1) 4일치 interarrival(일자별) 계산 ---
ia_1006 = get_daily_interarrival(counter1_1006_arrival_time, counter2_1006_arrival_time)
ia_1007 = get_daily_interarrival(counter1_1007_arrival_time, counter2_1007_arrival_time)
ia_1008 = get_daily_interarrival(counter1_1008_arrival_time, counter2_1008_arrival_time)
ia_1015 = get_daily_interarrival(counter1_1015_arrival_time, counter2_1015_arrival_time)

print("10/06 interarrival 개수:", len(ia_1006))
print("10/07 interarrival 개수:", len(ia_1007))
print("10/08 interarrival 개수:", len(ia_1008))
print("10/15 interarrival 개수:", len(ia_1015))


# --- 2) interarrival 값 자체 출력 (초 단위 & 분 단위) ---
def print_interarrivals(label, ia_list):
    print(f"\n=== {label} (n={len(ia_list)}) ===")
    print("초 단위 리스트:")
    print(ia_list)   # 이게 네가 말한 '값 자체'

    mins = [x/60 for x in ia_list]
    print("\n분 단위 리스트(=초/60):")
    print(mins)

    print("\n요약 통계:")
    print("  평균(초):", np.mean(ia_list))
    print("  표준편차(초):", np.std(ia_list))
    print("  최소/최대(초):", np.min(ia_list), "/", np.max(ia_list))

print_interarrivals("10/06 전체(1+2번)", ia_1006)
print_interarrivals("10/07 전체(1+2번)", ia_1007)
print_interarrivals("10/08 전체(1+2번)", ia_1008)
print_interarrivals("10/15 전체(1+2번)", ia_1015)

def get_interarrival_single_counter(arrival_times):
    secs = sorted(hhmmss_to_sec(t) for t in arrival_times)
    return [secs[i] - secs[i-1] for i in range(1, len(secs))]

c1_1006_ia = get_interarrival_single_counter(counter1_1006_arrival_time)
# 이런 식으로 나머지도 계산 가능

# 위에서 구한 ia_1006, ia_1007, ia_1008, ia_1015 재사용

all_ia = ia_1006 + ia_1007 + ia_1008 + ia_1015  # 4일 전체 interarrival (초 단위)
all_ia = np.array(all_ia)

print("4일 전체 interarrival 개수:", len(all_ia))
print("평균(초):", np.mean(all_ia))
print("표준편차(초):", np.std(all_ia))

import matplotlib.pyplot as plt
from scipy import stats

# 0보다 큰 값만 사용 (혹시 0이 있으면 분포 적합에 문제될 수 있어서 필터링)
all_ia_pos = all_ia[all_ia > 0]

# --- 1) 히스토그램 그리기 ---
plt.hist(all_ia_pos, bins=20, density=True, alpha=0.5)
plt.xlabel("Interarrival time (sec)")
plt.ylabel("Density")
plt.title("Histogram of Interarrival Times (4 days, both counters)")
plt.show()

# --- 2) 여러 분포 피팅해보기 ---
candidates = {
    "expon": stats.expon,          # 지수분포
    "gamma": stats.gamma,          # 감마분포
    "lognorm": stats.lognorm,      # 로그정규
    "weibull_min": stats.weibull_min  # Weibull 최소형
}

fit_results = {}

for name, dist in candidates.items():
    # 파라미터 적합
    params = dist.fit(all_ia_pos)
    fit_results[name] = params

    # K-S 검정으로 적합도 대략 보기
    ks_stat, p_val = stats.kstest(all_ia_pos, name, args=params)
    print(f"\n[{name}]")
    print("  params:", params)
    print("  KS stat:", ks_stat)
    print("  p-value:", p_val)

# --- 3) 히스토그램 위에 분포 곡선 그려보기 (시각적으로 비교) ---
xmin, xmax = all_ia_pos.min(), all_ia_pos.max()
x = np.linspace(xmin, xmax, 200)

plt.hist(all_ia_pos, bins=20, density=True, alpha=0.5, label="Data")

for name, dist in candidates.items():
    params = fit_results[name]
    pdf = dist.pdf(x, *params)
    plt.plot(x, pdf, label=name)

plt.xlabel("Interarrival time (sec)")
plt.ylabel("Density")
plt.title("Fitted Distributions to Interarrival Times")
plt.legend()
plt.show()
```
